```python
# modified_4
# multiple loaded models and one prediction per day_with validity data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import random
import timeit
import os
import yfinance as yf
import copy
import plotly.graph_objs as go
import plotly.io as pio
import re
# import FinanceDataReader as fdr
import datetime
import calendar
import warnings
import pandas_datareader as pdr
import pandas_datareader.data as web
import gc
# import cudf

from torch.nn.modules.transformer import TransformerEncoderLayer, TransformerEncoder
from torch.nn.modules.transformer import TransformerDecoder, TransformerDecoderLayer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from plotly.subplots import make_subplots
from datetime import datetime, date, timedelta
from sklearn.ensemble import BaggingClassifier

# from pandas.tseries.offsets import CustomBusinessDay
# from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, USMartinLutherKingJr, USPresidentsDay, GoodFriday, USMemorialDay, USLaborDay, USThanksgivingDay

warnings.filterwarnings("ignore", category=FutureWarning)

def create_sequences(data, seq_len):
    X = []
    y = []
    data = data.values  
    for i in range(seq_len, data.shape[0]):
        X.append(data[i-seq_len:i, :])
        y.append(data[i:i+1, -1])  

    X = np.array(X)
    y = np.array(y)
    return X, y

def create_sequences_all_features(data, seq_len):
    X = []
    y = []
    data = data.values  
    for i in range(seq_len, data.shape[0]):
        X.append(data[i-seq_len:i, :])
        y.append(data[i:i+1, :])  

    X = np.array(X)
    y = np.array(y)
    return X, y

def calculate_indicators_ver3(data, interval=None, seq_len=None):
    # Adjust parameters based on frequency
    if interval == '1wk':  # Weekly data
        period = 6
        rsi_period = 9
        short_ema_period = 8
        long_ema_period = 16
        signal_period = 5
        vol_period = 14
        bband_window_size = 12
    elif interval == '1mo':  # Monthly data
        period = 4
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 12
        signal_period = 4
        vol_period = 10
        bband_window_size = 8
    elif interval == '1d':  # Monthly data
        period = 10
        rsi_period = 14
        short_ema_period = 12
        long_ema_period = 26
        signal_period = 9
        vol_period = 26
        bband_window_size = 20
    elif interval == None:  # not specified
        period = 5
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 13
        signal_period = 4
        vol_period = 13
        bband_window_size = 10

    if data.shape[0] < long_ema_period + seq_len:
        return None

    # Check if 'Close' column exists in the data
    if 'Close' in data.columns:
        # Calculate daily returns
        data['Return'] = data['Close'].pct_change()
        # Calculate daily Direction
        data["Direction_Signal"] = np.where((data['Return']<-0.001), 1, np.where((data['Return']>0.001), 3, 2))
        # Calculate VPT (Volume-Price Trend)
        data['VPT'] = (data['Volume'] * data['Return']).cumsum()
        # Price Range
        data['Price Range'] = data['High'] - data['Low']
        # Candle Body Size
        data['Body Size'] = data['Open'] - data['Close']
        # Upper Shadow Size
        data['Upper Shadow'] = data.apply(lambda row: row.High - max(row.Open, row.Close), axis=1)
        # Lower Shadow Size
        data['Lower Shadow'] = data.apply(lambda row: min(row.Open, row.Close) - row.Low, axis=1)

        lag = (period - 1) // 2
        # Calculate adjusted price
        adjusted_price = 2 * data['Close'] - data['Close'].shift(lag)

        # Calculate ZLEMA directly using adjusted price
        zlema = adjusted_price.ewm(span=period).mean()
        data['ZLEMA'] = zlema

        # Define tolerance as a percentage of the close price
        tolerance = 0.015

        # Initialize column with 0s
        data['ZLEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * ZLEMA
        data.loc[data['Close'] > data['ZLEMA'] * (1 + tolerance), 'ZLEMA_Signal'] = 3

        # Sell signal when Close is below (1 - tolerance) * ZLEMA
        data.loc[data['Close'] < data['ZLEMA'] * (1 - tolerance), 'ZLEMA_Signal'] = 1

        # Calculate the first exponential moving average (EMA1).
        ema1 = data['Close'].ewm(span=period, min_periods=period).mean()
        # Calculate the second exponential moving average (EMA2).
        ema2 = ema1.ewm(span=period, min_periods=period).mean()
        # Calculate the third exponential moving average (EMA3).
        ema3 = ema2.ewm(span=period, min_periods=period).mean()
        # Calculate the TEMA.
        tema = 3 * ema1 - 3 * ema2 + ema3

        # Convert the TEMA to a Pandas Series.
        data['TEMA'] = tema

        # Initialize column with 0s
        data['TEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * TEMA
        data.loc[data ['Close']> data ['TEMA']* (1 + tolerance), 'TEMA_Signal']= 3

        # Sell signal when Close is below (1 - tolerance) * TEMA
        data.loc[data ['Close']< data ['TEMA']*(1-tolerance), 'TEMA_Signal']= 1

        # Calculate RSI and its signal
        delta = data['Close'].diff()
        up, down = delta.copy(), delta.copy()

        up[up < 0] = 0
        down[down > 0] = 0

        average_gain = up.rolling(window=rsi_period).mean()
        average_loss = abs(down.rolling(window=rsi_period).mean())

        epsilon = 1e-10  # A small constant to prevent division by zero
        rs = average_gain / (average_loss + epsilon)

        data['RSI'] =100 - (100 / (1 + rs))
        data["RSI_Signal"] =2
        data.loc[data.RSI < 30,"RSI_Signal"] =3
        data.loc[data.RSI > 70,"RSI_Signal"] =1

        # # Calculate MACD Line: (12-day EMA - 26-day EMA)
        EMA_short = data['Close'].ewm(span=short_ema_period).mean()
        EMA_long = data['Close'].ewm(span=long_ema_period).mean()
        data['MACD_Line'] = EMA_short - EMA_long

        # Calculate Signal Line: a n-day MA of MACD Line
        data['Signal_Line'] = data["MACD_Line"].ewm(span=signal_period).mean()

        # Generate MACD signals based on crossovers
        data["MACD_Signal"] = 2 # Neutral
        data.loc[(data.MACD_Line > data.Signal_Line) & (data.MACD_Line.shift() < data.Signal_Line.shift()),"MACD_Signal"] = 3 # Buy
        data.loc[(data.MACD_Line < data.Signal_Line) & (data.MACD_Line.shift() > data.Signal_Line.shift()),"MACD_Signal"] = 1 # Sell

        # Generate EMA signals based on crossovers
        data["EMA_Short"]=EMA_short
        data["EMA_Long"]=EMA_long

        # Define a small percentage as a threshold
        ema_diff_threshold = 0.01 # 1% difference

        # Calculate absolute percent difference between short and long EMA
        data['EMA_Diff'] = (EMA_short - EMA_long) / EMA_long

        data["EMA_Signal"] = np.where((data["EMA_Short"] > data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 3, 
                             np.where((data["EMA_Short"] < data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 1, 2))

        # Calculate Volatility as rolling standard deviation of log returns
        data["Log_Return"] =(np.log(data.Close).diff())
        data["Volatility"] =(data.Log_Return.rolling(window=vol_period, center=False).std())

        # Generate Volatility signals based on high or low volatility
        high_vol_threshold = data["Volatility"].quantile(0.85)
        low_vol_threshold = data["Volatility"].quantile(0.15)
        
        data['Volatility_Signal'] = 2
        data.loc[data.Volatility > high_vol_threshold, 'Volatility_Signal'] = 3
        data.loc[data.Volatility < low_vol_threshold, 'Volatility_Signal'] = 1

        # Calculate Bollinger Bands
        data["Middle_Band"] = data["Close"].rolling(window=bband_window_size, center=False).mean()
        data["Std_Dev"] = data["Close"].rolling(window=bband_window_size, center=False).std()
        data["Upper_Band"] = data["Middle_Band"] + data["Std_Dev"] * 2
        data["Lower_Band"] = data["Middle_Band"] - data["Std_Dev"] * 2

        # Identify shifting points
        data["BBand_Signal"] = 2  # neutral

        # Create a copy of the "Close" column
        close_shifted = data["Close"].copy()

        # Check for crossing Upper Band
        cross_upper_band = (close_shifted.shift(1) <= data["Upper_Band"].shift(1)) & (data["Close"] > data["Upper_Band"])
        data.loc[cross_upper_band, "BBand_Signal"] = 3  # buy

        # Check for crossing Lower Band
        cross_lower_band = (close_shifted.shift(1) >= data["Lower_Band"].shift(1)) & (data["Close"] < data["Lower_Band"])
        data.loc[cross_lower_band, "BBand_Signal"] = 1  # sell

        # Drop unnecessary columns
        data = data.drop(columns=["Middle_Band", "Std_Dev", 'Log_Return', 'Open', 'Low', 'High', 'Volume'])

    else:
        print("'Close' column is not present in the input dataframe.")

    return data

def calculate_indicators_ver4(data, interval=None, seq_len=None):
    # Adjust parameters based on frequency
    if interval == '1wk':  # Weekly data
        period = 6
        rsi_period = 9
        short_ema_period = 8
        long_ema_period = 16
        signal_period = 5
        vol_period = 14
        bband_window_size = 12
    elif interval == '1mo':  # Monthly data
        period = 4
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 12
        signal_period = 4
        vol_period = 10
        bband_window_size = 8
    elif interval == '1d':  # Monthly data
        period = 10
        rsi_period = 14
        short_ema_period = 12
        long_ema_period = 26
        signal_period = 9
        vol_period = 26
        bband_window_size = 20
    elif interval == None:  # not specified
        period = 5
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 13
        signal_period = 4
        vol_period = 13
        bband_window_size = 10

    if data.shape[0] < long_ema_period + seq_len:
        return None

    # Check if 'Close' column exists in the data
    if 'Close' in data.columns:
        # Calculate daily returns
        data['Return'] = data['Close'].pct_change()
        # Calculate daily Direction
        data["Direction_Signal"] = np.where((data['Return']<-0.001), 1, np.where((data['Return']>0.001), 3, 2))
        # Calculate VPT (Volume-Price Trend)
        data['VPT'] = (data['Volume'] * data['Return']).cumsum() 

        # Calculate the percentage change for the price range
        data['Price Range %'] = ((data['High'] - data['Low']) / data['Close']) * 100
        # Calculate the percentage change for the candle body size
        data['Body Size %'] = ((data['Open'] - data['Close']) / data['Close']) * 100
        # Calculate the percentage change for the upper shadow size
        data['Upper Shadow %'] = data.apply(lambda row: (row.High - max(row.Open, row.Close)) / row.Close * 100, axis=1)
        # Calculate the percentage change for the lower shadow size
        data['Lower Shadow %'] = data.apply(lambda row: (min(row.Open, row.Close) - row.Low) / row.Close * 100, axis=1)

        lag = (period - 1) // 2
        # Calculate adjusted price
        adjusted_price = 2 * data['Close'] - data['Close'].shift(lag)

        # Calculate ZLEMA directly using adjusted price
        zlema = adjusted_price.ewm(span=period).mean()
        data['ZLEMA'] = zlema

        data['ZLEMA %'] = zlema.pct_change()

        # Define tolerance as a percentage of the close price
        tolerance = 0.015

        # Initialize column with 0s
        data['ZLEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * ZLEMA
        data.loc[data['Close'] > data['ZLEMA'] * (1 + tolerance), 'ZLEMA_Signal'] = 3

        # Sell signal when Close is below (1 - tolerance) * ZLEMA
        data.loc[data['Close'] < data['ZLEMA'] * (1 - tolerance), 'ZLEMA_Signal'] = 1

        # Calculate the first exponential moving average (EMA1).
        ema1 = data['Close'].ewm(span=period, min_periods=period).mean()
        # Calculate the second exponential moving average (EMA2).
        ema2 = ema1.ewm(span=period, min_periods=period).mean()
        # Calculate the third exponential moving average (EMA3).
        ema3 = ema2.ewm(span=period, min_periods=period).mean()
        # Calculate the TEMA.
        tema = 3 * ema1 - 3 * ema2 + ema3

        # # Convert the TEMA to a Pandas Series.
        data['TEMA'] = tema

        data['TEMA %'] = tema.pct_change()

        # Initialize column with 0s
        data['TEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * TEMA
        data.loc[data ['Close']> data ['TEMA']* (1 + tolerance), 'TEMA_Signal']= 3

        # Sell signal when Close is below (1 - tolerance) * TEMA
        data.loc[data ['Close']< data ['TEMA']*(1-tolerance), 'TEMA_Signal']= 1

        # Calculate RSI and its signal
        delta = data['Close'].diff()
        up, down = delta.copy(), delta.copy()

        up[up < 0] = 0
        down[down > 0] = 0

        average_gain = up.rolling(window=rsi_period).mean()
        average_loss = abs(down.rolling(window=rsi_period).mean())

        epsilon = 1e-10  # A small constant to prevent division by zero
        rs = average_gain / (average_loss + epsilon)

        data['RSI'] =100 - (100 / (1 + rs))
        data['RSI_Diff'] = data['RSI'].diff()

        data["RSI_Signal"] =2
        data.loc[data.RSI < 30,"RSI_Signal"] =3
        data.loc[data.RSI > 70,"RSI_Signal"] =1

        # Calculate MACD Line: (12-day EMA - 26-day EMA)
        EMA_short = data['Close'].ewm(span=short_ema_period).mean()
        EMA_long = data['Close'].ewm(span=long_ema_period).mean()
        data['MACD_Line'] = EMA_short - EMA_long

        # Calculate Signal Line: a n-day MA of MACD Line
        data['Signal_Line'] = data["MACD_Line"].ewm(span=signal_period).mean()

        # Generate MACD signals based on crossovers
        data["MACD_Signal"] = 2 # Neutral
        data.loc[(data.MACD_Line > data.Signal_Line) & (data.MACD_Line.shift() < data.Signal_Line.shift()),"MACD_Signal"] = 3 # Buy
        data.loc[(data.MACD_Line < data.Signal_Line) & (data.MACD_Line.shift() > data.Signal_Line.shift()),"MACD_Signal"] = 1 # Sell

        # Generate EMA signals based on crossovers
        data["EMA_Short"]=EMA_short
        data["EMA_Long"]=EMA_long

        # Define a small percentage as a threshold
        ema_diff_threshold = 0.01 # 1% difference

        # Calculate absolute percent difference between short and long EMA
        data['EMA_Diff'] = (EMA_short - EMA_long)/ EMA_long


        data["EMA_Signal"] = np.where((data["EMA_Short"] > data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 3, 
                             np.where((data["EMA_Short"] < data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 1, 2))

        # Calculate Volatility as rolling standard deviation of log returns
        data["Log_Return"] =(np.log(data.Close).diff())
        data["Volatility"] =(data.Log_Return.rolling(window=vol_period, center=False).std())

        # Generate Volatility signals based on high or low volatility
        high_vol_threshold = data["Volatility"].quantile(0.85)
        low_vol_threshold = data["Volatility"].quantile(0.15)
        
        data['Volatility_Signal'] = 2
        data.loc[data.Volatility > high_vol_threshold, 'Volatility_Signal'] = 3
        data.loc[data.Volatility < low_vol_threshold, 'Volatility_Signal'] = 1

        # Calculate Bollinger Bands
        data["Middle_Band"] = data["Close"].rolling(window=bband_window_size, center=False).mean()
        data["Std_Dev"] = data["Close"].rolling(window=bband_window_size, center=False).std()
        data["Upper_Band"] = data["Middle_Band"] + data["Std_Dev"] * 2
        data["Lower_Band"] = data["Middle_Band"] - data["Std_Dev"] * 2

        # Identify shifting points
        data["BBand_Signal"] = 2  # neutral

        # Create a copy of the "Close" column
        close_shifted = data["Close"].copy()

        # Check for crossing Upper Band
        cross_upper_band = (close_shifted.shift(1) <= data["Upper_Band"].shift(1)) & (data["Close"] > data["Upper_Band"])
        data.loc[cross_upper_band, "BBand_Signal"] = 3  # buy

        # Check for crossing Lower Band
        cross_lower_band = (close_shifted.shift(1) >= data["Lower_Band"].shift(1)) & (data["Close"] < data["Lower_Band"])
        data.loc[cross_lower_band, "BBand_Signal"] = 1  # sell

        # Drop unnecessary columns
        data.drop(["Middle_Band", "Std_Dev", "Log_Return", "Open", 'Low', 'High', 'Volume'], axis=1, inplace=True)
        data.dropna(inplace=True)

    else:
        print("'Close' column is not present in the input dataframe.")

    return data

def calculate_indicators_ver58(data, interval=None, seq_len=None):
    # Adjust parameters based on frequency
    if interval == '1wk':  # Weekly data
        period = 6
        rsi_period = 9
        short_ema_period = 8
        long_ema_period = 16
        signal_period = 5
        vol_period = 14
        bband_window_size = 12
    elif interval == '1mo':  # Monthly data
        period = 4
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 12
        signal_period = 4
        vol_period = 10
        bband_window_size = 8
    elif interval == '1d':  # Monthly data
        period = 10
        rsi_period = 14
        short_ema_period = 12
        long_ema_period = 26
        signal_period = 9
        vol_period = 26
        bband_window_size = 20
    elif interval == None:  # not specified
        period = 5
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 13
        signal_period = 4
        vol_period = 13
        bband_window_size = 10

    if data.shape[0] < long_ema_period + seq_len:
        return None

    # Check if 'Close' column exists in the data
    if 'Close' in data.columns:
        # Calculate daily returns
        data['Return'] = data['Close'].pct_change()

        # Calculate daily Direction
        data["Direction_Signal"] = np.where((data['Return']<-0.001), -1, np.where((data['Return']>0.001), 1, 0))

        # Calculate VPT (Volume-Price Trend)
        data['VPT'] = (data['Volume'] * data['Return']).cumsum()
        data['VPT %'] = data['VPT'].diff()

        # Calculate the percentage change for the price range
        data['Price Range %'] = ((data['High'] - data['Low']) / data['Close']) * 100
        # Calculate the percentage change for the candle body size
        data['Body Size %'] = ((data['Open'] - data['Close']) / data['Close']) * 100
        # Calculate the percentage change for the upper shadow size
        data['Upper Shadow %'] = data.apply(lambda row: (row.High - max(row.Open, row.Close)) / row.Close * 100, axis=1)
        # Calculate the percentage change for the lower shadow size
        data['Lower Shadow %'] = data.apply(lambda row: (min(row.Open, row.Close) - row.Low) / row.Close * 100, axis=1)

        lag = (period - 1) // 2
        # Calculate adjusted price
        adjusted_price = 2 * data['Close'] - data['Close'].shift(lag)

        # Calculate ZLEMA directly using adjusted price
        data['ZLEMA'] = adjusted_price.ewm(span=period).mean()

        data['ZLEMA %'] = data['ZLEMA'].pct_change()

        # Define tolerance as a percentage of the close price
        tolerance = 0.015

        # Initialize column with 0s
        data['ZLEMA_Signal'] = 0

        # Buy signal when Close is above (1 + tolerance) * ZLEMA
        data.loc[data['Close'] > data['ZLEMA'] * (1 + tolerance), 'ZLEMA_Signal'] = 1

        # Sell signal when Close is below (1 - tolerance) * ZLEMA
        data.loc[data['Close'] < data['ZLEMA'] * (1 - tolerance), 'ZLEMA_Signal'] = -1

        # Calculate the first exponential moving average (EMA1).
        ema1 = data['Close'].ewm(span=period, min_periods=period).mean()
        # Calculate the second exponential moving average (EMA2).
        ema2 = ema1.ewm(span=period, min_periods=period).mean()
        # Calculate the third exponential moving average (EMA3).
        ema3 = ema2.ewm(span=period, min_periods=period).mean()
        # Calculate the TEMA.
        data['TEMA'] = 3 * ema1 - 3 * ema2 + ema3

        # # Convert the TEMA to a Pandas Series.

        data['TEMA %'] = data['TEMA'].pct_change()

        # Initialize column with 0s
        data['TEMA_Signal'] = 0

        # Buy signal when Close is above (1 + tolerance) * TEMA
        data.loc[data ['Close']> data ['TEMA']* (1 + tolerance), 'TEMA_Signal']= 1

        # Sell signal when Close is below (1 - tolerance) * TEMA
        data.loc[data ['Close']< data ['TEMA']*(1-tolerance), 'TEMA_Signal']= -1

        # Calculate RSI and its signal
        delta = data['Close'].diff()
        up, down = delta.copy(), delta.copy()

        up[up < 0] = 0
        down[down > 0] = 0

        average_gain = up.rolling(window=rsi_period).mean()
        average_loss = abs(down.rolling(window=rsi_period).mean())

        epsilon = 1e-10  # A small constant to prevent division by zero
        rs = average_gain / (average_loss + epsilon)
        data['RSI'] =100 - (100 / (1 + rs))

        data["RSI_Signal"] =0
        data.loc[data.RSI < 30,"RSI_Signal"] =1
        data.loc[data.RSI > 70,"RSI_Signal"] =-1

        # Calculate MACD Line: (12-day EMA - 26-day EMA)
        data['EMA_short'] = data['Close'].ewm(span=short_ema_period).mean()
        data['EMA_long'] = data['Close'].ewm(span=long_ema_period).mean()

        data['MACD_Line'] = data['EMA_short'] - data['EMA_long']

        # Calculate Signal Line: a n-day MA of MACD Line
        data['Signal_Line'] = data["MACD_Line"].ewm(span=signal_period).mean()

        # Generate MACD signals based on crossovers
        data["MACD_Signal"] = 0 # Neutral
        data.loc[(data.MACD_Line > data.Signal_Line) & (data.MACD_Line.shift(1) < data.Signal_Line.shift(1)),"MACD_Signal"] = 1 # Buy
        data.loc[(data.MACD_Line < data.Signal_Line) & (data.MACD_Line.shift(1) > data.Signal_Line.shift(1)),"MACD_Signal"] = -1 # Sell

        # Generate EMA signals based on crossovers
        data["EMA_Short %"]=data['EMA_short'].pct_change()
        data["EMA_Long %"]=data['EMA_long'].pct_change()

        # Define a small percentage as a threshold
        ema_diff_threshold = 0.01 # 1% difference

        # Calculate absolute percent difference between short and long EMA
        data['EMA_Diff'] = (data['EMA_short'] - data['EMA_long'])/ data['EMA_long']
        data["EMA_Signal"] = np.where((data['EMA_short'] > data['EMA_long']) & (abs(data['EMA_Diff']) > ema_diff_threshold), 1, 
                             np.where((data['EMA_short'] < data['EMA_long']) & (abs(data['EMA_Diff']) > ema_diff_threshold), -1, 0))

        # Calculate Volatility as rolling standard deviation of log returns
        data["Log_Return"] =(np.log(data.Close).diff())
        data["Volatility"] =(data.Log_Return.rolling(window=vol_period, center=False).std())

        # Generate Volatility signals based on high or low volatility
        high_vol_threshold = data["Volatility"].quantile(0.85)
        low_vol_threshold = data["Volatility"].quantile(0.15)
        
        data['Volatility_Signal'] = 0
        data.loc[data.Volatility > high_vol_threshold, 'Volatility_Signal'] = 1
        data.loc[data.Volatility < low_vol_threshold, 'Volatility_Signal'] = -1

        # Calculate Bollinger Bands
        middle_band = data["Close"].rolling(window=bband_window_size, center=False).mean()
        std_dev = data["Close"].rolling(window=bband_window_size, center=False).std()
        data["Upper_Band"] = middle_band + std_dev * 2
        data["Lower_Band"] = middle_band - std_dev * 2
        
        data["Upper_Band %"] = data["Lower_Band"].pct_change()
        data["Lower_Band %"] = data["Lower_Band"].pct_change()

        # Identify shifting points
        data["BBand_Signal"] = 0  # neutral

        # Create a copy of the "Close" column
        close_shifted = data["Close"].copy()

        # Check for crossing Upper Band
        cross_upper_band = (close_shifted.shift(1) <= data["Upper_Band"].shift(1)) & (data["Close"] > data["Upper_Band"])
        data.loc[cross_upper_band, "BBand_Signal"] = -1

        # Check for crossing Lower Band
        cross_lower_band = (close_shifted.shift(1) >= data["Lower_Band"].shift(1)) & (data["Close"] < data["Lower_Band"])
        data.loc[cross_lower_band, "BBand_Signal"] = 1  

        # Drop unnecessary columns
        data.drop(["VPT", "ZLEMA", "TEMA", "EMA_short", "EMA_long", "MACD_Line", "Signal_Line","Log_Return", "Upper_Band", "Lower_Band", "Open", 'Low', 'High', 'Volume', 'Close'], axis=1, inplace=True)
        data.dropna(inplace=True)

    else:
        print("'Close' column is not present in the input dataframe.")

    return data

def calculate_indicators_ver7(data, interval=None, seq_len=None):
    # Adjust parameters based on frequency
    if interval == '1wk':  # Weekly data
        period = 6
        rsi_period = 9
        short_ema_period = 8
        long_ema_period = 16
        signal_period = 5
        vol_period = 14
        bband_window_size = 12
    elif interval == '1mo':  # Monthly data
        period = 4
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 12
        signal_period = 4
        vol_period = 10
        bband_window_size = 8
    elif interval == '1d':  # Monthly data
        period = 10
        rsi_period = 14
        short_ema_period = 12
        long_ema_period = 26
        signal_period = 9
        vol_period = 26
        bband_window_size = 20
    elif interval == None:  # not specified
        period = 5
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 13
        signal_period = 4
        vol_period = 13
        bband_window_size = 10

    if data.shape[0] < long_ema_period + seq_len:
        return None

    # Check if 'Close' column exists in the data
    if 'Close' in data.columns:
        # Calculate daily returns
        data['Return'] = data['Close'].pct_change()

        # Calculate daily Direction
        data["Direction_Signal"] = np.where((data['Return']<-0.001), -1, np.where((data['Return']>0.001), 1, 0))

        # Calculate VPT (Volume-Price Trend)
        data['VPT'] = (data['Volume'] * data['Return']).cumsum()

        # Price Range
        data['Price Range'] = data['High'] - data['Low']
        # Candle Body Size
        data['Body Size'] = data['Open'] - data['Close']
        # Upper Shadow Size
        data['Upper Shadow'] = data.apply(lambda row: row.High - max(row.Open, row.Close), axis=1)
        # Lower Shadow Size
        data['Lower Shadow'] = data.apply(lambda row: min(row.Open, row.Close) - row.Low, axis=1)

        # Calculate the percentage change for the price range
        data['Price Range %'] = ((data['High'] - data['Low']) / data['Close']) * 100
        # Calculate the percentage change for the candle body size
        data['Body Size %'] = ((data['Open'] - data['Close']) / data['Close']) * 100
        # Calculate the percentage change for the upper shadow size
        data['Upper Shadow %'] = data.apply(lambda row: (row.High - max(row.Open, row.Close)) / row.Close * 100, axis=1)
        # Calculate the percentage change for the lower shadow size
        data['Lower Shadow %'] = data.apply(lambda row: (min(row.Open, row.Close) - row.Low) / row.Close * 100, axis=1)

        lag = (period - 1) // 2
        # Calculate adjusted price
        adjusted_price = 2 * data['Close'] - data['Close'].shift(lag)

        # Calculate ZLEMA directly using adjusted price
        data['ZLEMA'] = adjusted_price.ewm(span=period).mean()

        data['ZLEMA %'] = data['ZLEMA'].pct_change()

        # Define tolerance as a percentage of the close price
        tolerance = 0.015

        # Initialize column with 0s
        data['ZLEMA_Signal'] = 0

        # Buy signal when Close is above (1 + tolerance) * ZLEMA
        data.loc[data['Close'] > data['ZLEMA'] * (1 + tolerance), 'ZLEMA_Signal'] = 1

        # Sell signal when Close is below (1 - tolerance) * ZLEMA
        data.loc[data['Close'] < data['ZLEMA'] * (1 - tolerance), 'ZLEMA_Signal'] = -1

        # Calculate the first exponential moving average (EMA1).
        ema1 = data['Close'].ewm(span=period, min_periods=period).mean()
        # Calculate the second exponential moving average (EMA2).
        ema2 = ema1.ewm(span=period, min_periods=period).mean()
        # Calculate the third exponential moving average (EMA3).
        ema3 = ema2.ewm(span=period, min_periods=period).mean()
        # Calculate the TEMA.
        data['TEMA'] = 3 * ema1 - 3 * ema2 + ema3

        # # Convert the TEMA to a Pandas Series.

        data['TEMA %'] = data['TEMA'].pct_change()

        # Initialize column with 0s
        data['TEMA_Signal'] = 0

        # Buy signal when Close is above (1 + tolerance) * TEMA
        data.loc[data ['Close']> data ['TEMA']* (1 + tolerance), 'TEMA_Signal']= 1

        # Sell signal when Close is below (1 - tolerance) * TEMA
        data.loc[data ['Close']< data ['TEMA']*(1-tolerance), 'TEMA_Signal']= -1

        # Calculate RSI and its signal
        delta = data['Close'].diff()
        up, down = delta.copy(), delta.copy()

        up[up < 0] = 0
        down[down > 0] = 0

        average_gain = up.rolling(window=rsi_period).mean()
        average_loss = abs(down.rolling(window=rsi_period).mean())

        epsilon = 1e-10  # A small constant to prevent division by zero
        rs = average_gain / (average_loss + epsilon)
        data['RSI'] =100 - (100 / (1 + rs))

        data["RSI_Signal"] =0
        data.loc[data.RSI < 30,"RSI_Signal"] =1
        data.loc[data.RSI > 70,"RSI_Signal"] =-1

        # Calculate MACD Line: (12-day EMA - 26-day EMA)
        data['EMA_short'] = data['Close'].ewm(span=short_ema_period).mean()
        data['EMA_long'] = data['Close'].ewm(span=long_ema_period).mean()

        data['MACD_Line'] = data['EMA_short'] - data['EMA_long']

        # Calculate Signal Line: a n-day MA of MACD Line
        data['Signal_Line'] = data["MACD_Line"].ewm(span=signal_period).mean()

        # Generate MACD signals based on crossovers
        data["MACD_Signal"] = 0 # Neutral
        data.loc[(data.MACD_Line > data.Signal_Line) & (data.MACD_Line.shift(1) < data.Signal_Line.shift(1)),"MACD_Signal"] = 1 # Buy
        data.loc[(data.MACD_Line < data.Signal_Line) & (data.MACD_Line.shift(1) > data.Signal_Line.shift(1)),"MACD_Signal"] = -1 # Sell

        # Generate EMA signals based on crossovers
        data["EMA_Short %"]=data['EMA_short'].pct_change()
        data["EMA_Long %"]=data['EMA_long'].pct_change()

        # Define a small percentage as a threshold
        ema_diff_threshold = 0.01 # 1% difference

        # Calculate absolute percent difference between short and long EMA
        data['EMA_Diff'] = (data['EMA_short'] - data['EMA_long'])/ data['EMA_long']
        data["EMA_Signal"] = np.where((data['EMA_short'] > data['EMA_long']) & (abs(data['EMA_Diff']) > ema_diff_threshold), 1, 
                             np.where((data['EMA_short'] < data['EMA_long']) & (abs(data['EMA_Diff']) > ema_diff_threshold), -1, 0))

        # Calculate Volatility as rolling standard deviation of log returns
        data["Log_Return"] =(np.log(data.Close).diff())
        data["Volatility"] =(data.Log_Return.rolling(window=vol_period, center=False).std())

        # Generate Volatility signals based on high or low volatility
        high_vol_threshold = data["Volatility"].quantile(0.85)
        low_vol_threshold = data["Volatility"].quantile(0.15)
        
        data['Volatility_Signal'] = 0
        data.loc[data.Volatility > high_vol_threshold, 'Volatility_Signal'] = 1
        data.loc[data.Volatility < low_vol_threshold, 'Volatility_Signal'] = -1

        # Calculate Bollinger Bands
        middle_band = data["Close"].rolling(window=bband_window_size, center=False).mean()
        std_dev = data["Close"].rolling(window=bband_window_size, center=False).std()
        data["Upper_Band"] = middle_band + std_dev * 2
        data["Lower_Band"] = middle_band - std_dev * 2
        
        data["Upper_Band %"] = data["Lower_Band"].pct_change()
        data["Lower_Band %"] = data["Lower_Band"].pct_change()

        # Identify shifting points
        data["BBand_Signal"] = 0  # neutral

        # Create a copy of the "Close" column
        close_shifted = data["Close"].copy()

        # Check for crossing Upper Band
        cross_upper_band = (close_shifted.shift(1) <= data["Upper_Band"].shift(1)) & (data["Close"] > data["Upper_Band"])
        data.loc[cross_upper_band, "BBand_Signal"] = -1

        # Check for crossing Lower Band
        cross_lower_band = (close_shifted.shift(1) >= data["Lower_Band"].shift(1)) & (data["Close"] < data["Lower_Band"])
        data.loc[cross_lower_band, "BBand_Signal"] = 1  

        # Drop unnecessary columns
        data.drop(["Log_Return", "Open", 'Low', 'High'], axis=1, inplace=True)
        data.dropna(inplace=True)

    else:
        print("'Close' column is not present in the input dataframe.")

    return data

def fetch_data(symbol_list, start_date, interval):    
   
    # Initialize an empty DataFrame
    all_data = []
    today = datetime.datetime.now()

    for symbol in symbol_list:
    
        data = yf.download(symbol, start=start_date, end=today, interval=interval)
        data = data.copy()

        # Convert the index to datetime
        data.index = pd.to_datetime(data.index)
        # print(data.index[-1])
        if interval.endswith('mo'):
            # Check if the last data point is from the current month and year
            if data.index[-1].month == today.month and data.index[-1].year == today.year:
                # Count the number of days in the current month up to today
                days_this_month_so_far = today.day

            # If the number of days in the last month is less than 20, drop the last month's data
            if days_this_month_so_far < 20:
                data = data.iloc[:-1].copy()

        elif interval.endswith('wk'):
            weekdays_till_today = (today.weekday() + 1) % 7
            # If the number of days in the last week is less than 3, drop the last week's data
            if weekdays_till_today < 3:  # 0: Monday, ..., 4: Friday
                data = data.iloc[:-1].copy()

        data.dropna(inplace=True)
        data = data[data['Volume'] != 0].copy()
        
        # Check if 'Adj Close' is in columns and drop it
        if 'Adj Close' in data.columns:
            data = data.drop('Adj Close', axis=1).copy()            
                    # Add a column to identify the symbol
        data['Symbol'] = symbol
        
                # Append the data to the all_data DataFrame
        all_data.append(data)
        
    # Concatenate all_data into a single DataFrame 
    all_data_df = pd.concat(all_data)
    
    # Reset the index of the final DataFrame
    all_data_df.reset_index(inplace=True)
            
    return all_data_df

def prepare_data(all_data, seq_len=None, target_col=None, symbol=None, scaler=StandardScaler(), interval=None, forward=0, 
                 train_size=0.8, united_transform=None, indicator_ver=None):
    
    test_data = all_data[all_data['Symbol'] == symbol]
    
    test_data = test_data.drop('Symbol', axis=1)

    if united_transform == True:
        # Scale test data
        test_data_unnormalized = test_data.copy()

        if indicator_ver == 'calculate_indicators_ver4':
            test_data_unnormalized = calculate_indicators_ver4(test_data_unnormalized, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver5.8':
            test_data_unnormalized = calculate_indicators_ver58(test_data_unnormalized, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver7':
            test_data_unnormalized = calculate_indicators_ver7(test_data_unnormalized, interval=interval, seq_len=seq_len)

        # If calculate_indicators returned None, return None
        if test_data_unnormalized is None:
            return None
        
        # Split the data 
        cut_point = int(len(test_data_unnormalized) * train_size)
        test_data_unnormalized = test_data_unnormalized.iloc[cut_point:].copy()

        test_data_unnormalized['Target'] = test_data_unnormalized[target_col]
        test_data_unnormalized.dropna(inplace=True)
        test_data_unnormalized = test_data_unnormalized.drop(target_col, axis=1)
        test_data_unnormalized['Date'] = pd.to_datetime(test_data_unnormalized['Date'])
        test_data_unnormalized.set_index('Date', inplace=True)
    
        # Scale test data
        test_data_indicator = test_data.copy()

        if indicator_ver == 'calculate_indicators_ver4':
            test_data_indicator = calculate_indicators_ver4(test_data_indicator, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver5.8':
            test_data_indicator = calculate_indicators_ver58(test_data_indicator, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver7':
            test_data_indicator = calculate_indicators_ver7(test_data_indicator, interval=interval, seq_len=seq_len)

        # If calculate_indicators returned None, return None
        if test_data_indicator is None:
            return None
        
        test_data_indicator = test_data_indicator.iloc[cut_point:]

        test_data_indicator['Date'] = pd.to_datetime(test_data_indicator['Date'])
        test_data_indicator.set_index('Date', inplace=True)
        test_data_indicator[test_data_indicator.columns] = scaler.fit_transform(test_data_indicator)
        test_data_indicator['Target'] = test_data_indicator[target_col].shift(forward)
        test_data_indicator.dropna(inplace=True)
        test_data_indicator = test_data_indicator.drop(target_col, axis=1)

    else:
        # Split the data 
        cut_point = int(len(test_data) * train_size)
        test_data_cut = test_data.copy()
        test_data_cut = test_data_cut.iloc[cut_point:]

        test_data_unnormalized = test_data_cut.copy()

        if indicator_ver == 'calculate_indicators_ver4':
            test_data_unnormalized = calculate_indicators_ver4(test_data_unnormalized, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver5.8':
            test_data_unnormalized = calculate_indicators_ver58(test_data_unnormalized, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver7':
            test_data_unnormalized = calculate_indicators_ver7(test_data_unnormalized, interval=interval, seq_len=seq_len)
       
       # If calculate_indicators returned None, return None
        if test_data_unnormalized is None:
            return None
              
        test_data_unnormalized['Target'] = test_data_unnormalized[target_col]
        test_data_unnormalized.dropna(inplace=True)
        test_data_unnormalized = test_data_unnormalized.drop(target_col, axis=1)
        test_data_unnormalized['Date'] = pd.to_datetime(test_data_unnormalized['Date'])
        test_data_unnormalized.set_index('Date', inplace=True)

        test_data_indicator = test_data_cut.copy()

        if indicator_ver == 'calculate_indicators_ver4':
            test_data_indicator = calculate_indicators_ver4(test_data_indicator, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver5.8':
            test_data_indicator = calculate_indicators_ver58(test_data_indicator, interval=interval, seq_len=seq_len)
        elif indicator_ver == 'calculate_indicators_ver7':
            test_data_indicator = calculate_indicators_ver7(test_data_indicator, interval=interval, seq_len=seq_len)

        # If calculate_indicators returned None, return None
        if test_data_indicator is None:
            return None
        
        test_data_indicator['Date'] = pd.to_datetime(test_data_indicator['Date'])
        test_data_indicator.set_index('Date', inplace=True)
        test_data_indicator[test_data_indicator.columns] = scaler.fit_transform(test_data_indicator)
        
        test_data_noshift = test_data_indicator.copy()
        test_data_indicator['Target'] = test_data_indicator[target_col].shift(forward)
        test_data_noshift['Target'] = test_data_noshift[target_col].shift(0)

        test_data_indicator.dropna(inplace=True)
        test_data_indicator = test_data_indicator.drop(target_col, axis=1)

        test_data_noshift.dropna(inplace=True)
        test_data_noshift = test_data_noshift.drop(target_col, axis=1)        

    return test_data_indicator, test_data_unnormalized, test_data_noshift, seq_len

def prepare_data_tensor(test_data, test_data_noshift, seq_len):
    # Create sequences
    X_test, y_test = create_sequences(test_data, seq_len)
    X_test_all_features, y_test_all_features  = create_sequences_all_features(test_data, seq_len)
    X_test_noshift, y_test_noshift = create_sequences(test_data_noshift, seq_len)

    # Convert to PyTorch tensors
    X_test = torch.Tensor(X_test)
    y_test = torch.Tensor(y_test)
    y_test_all_features = torch.Tensor(y_test_all_features)
    X_test_noshift = torch.Tensor(X_test_noshift)

    return X_test, y_test, y_test_all_features, X_test_noshift

class DecoderOnlyTransformerModel(nn.Module):
    def __init__(self, n_features, nhead=None, nhid=None, nlayers=None, dropout=None,
                 l1_regularization=0, l2_regularization=0,
                 activation_function=None):
        super(DecoderOnlyTransformerModel, self).__init__()

        self.pos_encoder = nn.Sequential(
            nn.Linear(n_features, nhid),
            activation_function,
            nn.Linear(nhid, nhid),
            activation_function
        )

        decoder_layers = TransformerDecoderLayer(nhid, nhead)
        self.transformer_decoder = TransformerDecoder(decoder_layers,nlayers)

        self.decoder = nn.Linear(nhid,1)

        self.l1_regularization = l1_regularization
        self.l2_regularization = l2_regularization

    def init_weights(self):
        # initrange = 0.1
        nn.init.xavier_uniform_(self.pos_encoder[0].weight)
        nn.init.xavier_uniform_(self.pos_encoder[2].weight)
        self.decoder.bias.data.zero_()
        nn.init.xavier_uniform_(self.decoder.weight)

    def regularization_loss(self):
        l1_loss = 0
        l2_loss = 0
        for param in self.parameters():
            l1_loss += torch.norm(param, 1)
            l2_loss += torch.norm(param, 2) ** 2
        return self.l1_regularization * l1_loss + self.l2_regularization * l2_loss

    def generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask

    def forward(self, x):
        x = self.pos_encoder(x)

        # Generating target mask
        tgt_mask = self.generate_square_subsequent_mask(x.size(0)).to(x.device)

        # Processing the entire input at once
        output = self.transformer_decoder(x, x, tgt_mask=tgt_mask)

        # Apply the final linear layer
        output = self.decoder(output)
        # Selecting the last element from each sequence
        output = output[:, -1, :]
        return output
    
class TransformerModel(nn.Module):

    def __init__(self, n_features, nhead=None, nhid=None, nlayers=None, dropout=None, l1_regularization=0, l2_regularization=0, activation_function = None):
        super(TransformerModel, self).__init__()

        self.activation_function = activation_function

        # Model architecture
        self.pos_encoder = nn.Sequential(
            nn.Linear(n_features, nhid),
            self.activation_function,
            nn.LayerNorm(nhid),
            nn.Dropout(dropout),
            nn.Linear(nhid, nhid),
            self.activation_function,
            nn.LayerNorm(nhid)
        )
        encoder_layers = TransformerEncoderLayer(nhid, nhead, nhid, dropout,)
        # encoder_layers.self_attn.batch_first = True
        self.transformer_encoder = TransformerEncoder(encoder_layers, nlayers)
        self.decoder = nn.Linear(nhid, 1)
        self.l1_regularization = l1_regularization
        self.l2_regularization = l2_regularization
        self.init_weights()

    def init_weights(self):
        for layer in self.pos_encoder:
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)

        self.decoder.bias.data.zero_()
        nn.init.xavier_uniform_(self.decoder.weight)

    def regularization_loss(self):
        l1_loss = 0
        l2_loss = 0
        for param in self.parameters():
            l1_loss += torch.norm(param, 1)
            l2_loss += torch.norm(param, 2) ** 2

        return self.l1_regularization * l1_loss + self.l2_regularization * l2_loss

    def forward(self, src):
        src = self.pos_encoder(src)
        output = self.transformer_encoder(src)
        output = self.decoder(output)
        output = output[:, -1:, :].squeeze(-1)
        return output
    
class ModelWrapper:
    def __init__(self, model_path):
        self.model = torch.load(model_path)  # Load the saved model

    def predict(self, x):
        with torch.no_grad():
            return self.model(x)  # Call the model's forward method
    
def inverse_transform_wrapper(data, orgshape, scaler):
    data_reshaped = data.reshape(-1, data.shape[-1])
    data_inv = scaler.inverse_transform(data_reshaped)
    data_inv_origshape = data_inv.reshape(orgshape)
    return data_inv_origshape

def stock_price_plots(data, y_pred_org=None, y_test_org=None, file_name=None, n_predict=1,  
                      save_directory=None, future_predictions=None, seq_len=2,
                      scaler=None, col_names=None, symbol=None, test_dates=None, test_length=None, freq=None):
    
    # Reshape tensors for metrics computation
    y_pred_org_reshape = y_pred_org.reshape(-1, y_pred_org.shape[-1])

    y_test_org_reshape = y_test_org.reshape(-1, y_test_org.shape[-1])
    y_test_org_df = pd.DataFrame(y_test_org_reshape, columns=[col_names[i] for i in range(y_test_org.shape[-1])])

    # Take last future prediction
    last_future_prediction = future_predictions

    data = data.copy()

    # Adjust data index for y_pred
    # adjusted_index = data.index[seq_len:]
    if test_dates == 0:
        adjusted_index = data.index
    else:
        adjusted_index = data.index[:-test_dates]   
   
    # Separate the numeric and string parts
    match = re.match(r'(\d+)?(\w+)', freq)
    if match:
        freq_number, freq_string = match.groups()
        freq_number = int(freq_number) if freq_number else 1
    else:
        freq_number = 1
        freq_string = freq

    frequency_mapping = {'M': 'months','MS': 'months','W': 'weeks','D': 'days',}
    freq_param = frequency_mapping.get(freq_string, freq_string)

    try:
        offset = pd.DateOffset(**{freq_param: freq_number})
    except KeyError:
        offset = None

    print(offset)

    if offset is not None:
        start_date = adjusted_index[-1] + offset
        future_dates = pd.date_range(start=start_date, periods=n_predict, freq=freq)
        print(f'future_dates: {future_dates}')
    else:
        print("Invalid frequency parameter.")
        
    combined_index = pd.concat([pd.Series(adjusted_index), pd.Series(future_dates)]).reset_index(drop=True)
    combined_index = pd.DatetimeIndex(combined_index)
    print('combined_index: {combined_index}')

    # Create a subplot with 6 rows
    fig = make_subplots(rows=6, cols=1, shared_xaxes=True,
                        vertical_spacing=0.03,
                        specs=[[{}], [{}], [{}], [{}], [{}], [{}]])

    # Adjust sizes of subplots
    fig['layout']['yaxis']['domain'] = [0.6, 1]
    fig['layout']['yaxis2']['domain'] = [0.5, 0.6]
    fig['layout']['yaxis3']['domain'] = [0.35, 0.5]
    fig['layout']['yaxis4']['domain'] = [0.25, 0.35]
    fig['layout']['yaxis5']['domain'] = [0.15, 0.25]
    fig['layout']['yaxis6']['domain'] = [0., 0.15]

    # Add line chart for Target (close) orice in the first row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['Target'], mode='lines', name='Target(Close)'), row=1, col=1)

    # # Add line chart for moving averages in the first row
    # fig.add_trace(go.Scatter(x=adjusted_index, y=data['MA_90'][seq_len:], mode='lines', name='MA 90'), row=1, col=1)
    # fig.add_trace(go.Scatter(x=adjusted_index, y=data['MA_200'][seq_len:], mode='lines', name='MA 200'), row=1, col=1)

    # Add line chart for ZLEMA and TEMA in the first row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['ZLEMA'], mode='lines', name='ZLEMA'), row=1, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['TEMA'], mode='lines', name='TEMA'), row=1, col=1)

    # Add Bollinger Bands in the first row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['Upper_Band'], mode='lines', name='Upper Band', line=dict(width=0.5, color='gray')), row=1, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['Lower_Band'], mode='lines', name='Lower Band', line=dict(width=0.5, color='gray')), row=1, col=1)
    # fig.add_trace(go.Scatter(x=adjusted_index, y=data['MA_10'][seq_len:], mode='lines', name='Middle Band', line=dict(width=0.5, color='gray')), row=1, col=1)

    # Add Volume in the second row
    fig.add_trace(go.Bar(x=adjusted_index, y=y_test_org_df['VPT'], name='VPT'), row=2, col=1)

    # # Add Volume in the second row as scatter plot with mode='markers'
    # volume_scatter = go.Scatter(x=adjusted_index, y=data['Volume'][seq_len:], mode='markers', name='Volume')
    # fig.add_trace(volume_scatter, row=2, col=1)

    # Add line chart for RSI in the third row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['RSI'], mode='lines', name='RSI'), row=3, col=1)

    # Add MACD and Signal Line in the fourth row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['MACD_Line'], mode='lines', name='MACD Line'), row=4, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['Signal_Line'], mode='lines', name='Signal Line'), row=4, col=1)

    # Add EMA Diff in the fourth row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['Return'], mode='lines', name='% Change'), row=5, col=1)

    # Add line chart for Buy and Sell Signal data in the fifth row
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['RSI_Signal'], mode='lines', name='RSI Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['MACD_Signal'], mode='lines', name='MACD Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['EMA_Signal'], mode='lines', name='EMA Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_test_org_df['BBand_Signal'], mode='lines', name='BBand Signal'), row=6, col=1)

    # Add title and labels
    fig.update_layout(title_text=f'Stock Price with Indicators and Predictions_{symbol}_{file_name}_{test_length}')
    fig.update_yaxes(title_text='Price', row=1, col=1)
    fig.update_yaxes(title_text='Volume', row=2, col=1)
    fig.update_yaxes(title_text='RSI', row=3, col=1)
    fig.update_yaxes(title_text='MACD', row=4, col=1)
    fig.update_yaxes(title_text='Return', row=5, col=1)
    fig.update_yaxes(title_text='Signals', row=6, col=1)

    y_combined = np.concatenate([y_pred_org_reshape, last_future_prediction])
    y_combined_df = pd.DataFrame(y_combined, columns=[col_names[i] for i in range(y_pred_org.shape[2])])

    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['Target'], mode='lines', name='Predicted Target'), row=1, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['VPT'], mode='lines', name='Predicted volume'), row=2, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['RSI'], mode='lines', name='Predicted RSI'), row=3, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['MACD_Line'], mode='lines', name='Predicted MACD Line'), row=4, col=1)
    fig.add_trace(go.Scatter(x=adjusted_index, y=y_combined_df['Signal_Line'], mode='lines', name='Signal Line'), row=4, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['Return'], mode='lines', name='Predicted Return'), row=5, col=1)

    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['RSI_Signal'], mode='lines', name='Predicted RSI Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['MACD_Signal'], mode='lines', name='Predicted MACD Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['EMA_Signal'], mode='lines', name='Predicted EMA Signal'), row=6, col=1)
    fig.add_trace(go.Scatter(x=combined_index, y=y_combined_df['BBand_Signal'], mode='lines', name='Predicted BBand Signal'), row=6, col=1)

    # Adjust overall plot size
    fig.update_layout(height=1500)

    # Add predicted data to each relevant subplot
    # fig.add_trace(go.Scatter(x=adjusted_index, y=y_predicted['Target'], mode='lines', name='Predicted Target'), row=1, col=1)
    # fig.add_trace(go.Scatter(x=adjusted_index, y=y_predicted['Volume'], mode='lines', name='Predicted Volume'), row=2, col=1)
    # fig.add_trace(go.Scatter(x=adjusted_index, y=y_predicted['RSI'], mode='lines', name='Predicted RSI'), row=3, col=1)
    # fig.add_trace(go.Scatter(x=adjusted_index, y=y_predicted['MACD_Line'], mode='lines', name='Predicted MACD Line'), row=4, col=1)

    if save_directory:
        save_path = os.path.join(save_directory, f"stock_price_plots_{symbol}_{file_name}_{test_dates}_{test_length}.html")
        pio.write_html(fig, save_path)

    # Show the figure
    fig.show()

def evaluate_and_plot_predictions(y_test, y_test_org, y_pred_prev, y_pred_prev_org, trial=None, n_predict=None, data=None, seq_len=None, use_target_col=True, metrics_use=True,
                                  plots_use=True, save_directory=None, future_predictions=None, col_names=None, symbol=None, test_dates=None, test_length=None, freq=None):
    torch.cuda.empty_cache()
    # Get n_features from y_test
    n_features = y_test.shape[-1]

    # Reshape tensors for metrics computation
    y_pred_for_metrics = y_pred_prev.reshape(-1, 1) if y_pred_prev.ndim == 1 else y_pred_prev.reshape(-1, y_pred_prev.shape[-1])    
    y_for_metrics = y_test.reshape(-1, 1) if y_test.ndim == 1 else y_test.reshape(-1, y_test.shape[-1])
    
    if metrics_use:

        evaluation_results = {}

        # # Calculate metrics for all features
        # mse_all_features = mean_squared_error(y_for_metrics, y_pred_for_metrics)
        # mae_all_features = mean_absolute_error(y_for_metrics, y_pred_for_metrics)
        # r2_all_features = r2_score(y_for_metrics, y_pred_for_metrics)

        # # Store results in dictionary
        # evaluation_results['all_features'] = {
        #     'mse': mse_all_features,
        #     'mae': mae_all_features,
        #     'r2': r2_all_features,
        # }

        # If there is only one feature, don't loop over features
        if y_for_metrics.shape[1] == 1:
            # Correctly handle the case where col_names might be a pandas Series or Index
            col_name = col_names[-1] if not col_names.empty else 'Feature_1'
            # Calculate metrics for this column
            mse = mean_squared_error(y_for_metrics, y_pred_for_metrics)
            mae = mean_absolute_error(y_for_metrics, y_pred_for_metrics)
            r2 = r2_score(y_for_metrics, y_pred_for_metrics)

            # Store results in dictionary
            evaluation_results[col_name] = {
            'mse': mse,
            'mae': mae,
            'r2': r2,
            }
        else:
            # Iterate over each column in y_true for multiple features
            for i, col_name in enumerate(col_names):
                # Calculate metrics for this column
                mse = mean_squared_error(y_for_metrics[:,i], y_pred_for_metrics[:,i])
                mae = mean_absolute_error(y_for_metrics[:,i], y_pred_for_metrics[:,i])
                r2 = r2_score(y_for_metrics[:,i], y_pred_for_metrics[:,i])

                # Store results in dictionary
                evaluation_results[col_name] = {
                    'mse': mse,
                    'mae': mae,
                    'r2': r2,
                } 
  
    if plots_use:
        
        # if test_dates == 0:
        #     date_index = data.index
        # elif test_dates >= 1:
        #     date_index = data.index[:-test_dates]
        date_index = data.index        
        # Ensure y_test_org and y_pred_prev_org are 2D for consistency in plotting
        y_test_org = y_test_org.reshape(-1, 1) if y_test_org.ndim == 1 else y_test_org.reshape(-1, y_test_org.shape[-1])
        y_pred_prev_org = y_pred_prev_org.reshape(-1, 1) if y_pred_prev_org.ndim == 1 else y_pred_prev_org.reshape(-1, y_pred_prev_org.shape[-1])

        # Separate the numeric and string parts
        match = re.match(r'(\d+)?(\w+)', freq)
        if match:
            freq_number, freq_string = match.groups()
            freq_number = int(freq_number) if freq_number else 1
        else:
            freq_number = 1
            freq_string = freq

        frequency_mapping = {'M': 'months','MS': 'months','W': 'weeks','D': 'days',}
        freq_param = frequency_mapping.get(freq_string, freq_string)

        try:
            offset = pd.DateOffset(**{freq_param: freq_number})
        except KeyError:
            offset = None

        print(offset)

        if offset is not None:
            start_date_index = date_index[-1] + offset
            future_dates = pd.date_range(start=start_date_index, periods=n_predict, freq=freq)
            print(f'future_dates: {future_dates}')
        else:
            print("Invalid frequency parameter.")

        # If use_target_col is True, only plot the target column, otherwise plot all feature columns
        if use_target_col:

            if future_predictions is not None:
                last_future_prediction = future_predictions.flatten()
                combined_predictions = np.concatenate([y_pred_prev_org.flatten(), last_future_prediction])                
            else:
                combined_predictions = y_pred_prev_org.flatten()

            combined_dates = pd.concat([pd.Series(date_index), pd.Series(future_dates)]).reset_index(drop=True)
            combined_dates = pd.DatetimeIndex(combined_dates)

            fig = go.Figure()
            # When plotting use_target_col_plot
            fig.add_trace(go.Scatter(x=combined_dates[:-n_predict], y=y_test_org[:, -1], mode='lines', name='Actual'))
            fig.add_trace(go.Scatter(x=combined_dates, y=combined_predictions, mode='lines', name='Predicted + Future'))

            fig.update_layout(title=f'Actual and Predicted Values for {col_names[-1]} ({symbol}_{trial}_{test_dates}_{test_length})', 
                              xaxis_title='Time Step', yaxis_title='Value', height=800, width=1500)

            if save_directory:
                save_path = os.path.join(save_directory, f"predictions_plot_target_{symbol}_{trial}_{test_length}.html")
                fig.write_html(save_path)
            fig.show()
        else:
            for j in range(n_features):
                last_future_prediction = future_predictions
                print("future_predictions:", last_future_prediction[:, j])

                combined_predictions = np.concatenate([y_pred_prev_org[:, j], last_future_prediction[:, j]])
                combined_dates = pd.concat([pd.Series(date_index), pd.Series(future_dates)]).reset_index(drop=True)
                combined_dates = pd.DatetimeIndex(combined_dates)

                fig = go.Figure()
                # When plotting use_target_col_plot
                fig.add_trace(go.Scatter(x=combined_dates[:-n_predict], y=y_test_org[:, j], mode='lines', name='Actual'))
                fig.add_trace(go.Scatter(x=combined_dates, y=combined_predictions, mode='lines', name='Predicted + Future'))

                fig.update_layout(title=f'Actual and Predicted Values for {col_names[j]} ({symbol}_{trial}_{test_length})', 
                                  xaxis_title='Time Step', yaxis_title='Value', height=800, width=1500)

                if save_directory:
                    save_path = os.path.join(save_directory, f"predictions_plot_var_{j + 1}_{symbol}_{trial}_{test_dates}_{test_length}.html")
                    fig.write_html(save_path)
                fig.show()
                
    # if plots_use:
    #     # Check if test_dates is not 0 and data is provided
    #     if test_dates != 0 and data is not None:
    #         date_index = data.index[:-test_dates]
    #     else:
    #         date_index = data.index

    #     # Reshape for consistency in plotting (this seems to be already correctly handled)
    #     y_test_org = y_test_org.reshape(-1, 1) if y_test_org.ndim == 1 else y_test_org
    #     y_pred_prev_org = y_pred_prev_org.reshape(-1, 1) if y_pred_prev_org.ndim == 1 else y_pred_prev_org

    #     # Handling for frequency and future dates
    #     match = re.match(r'(\d+)?(\w+)', freq)
    #     freq_number, freq_string = match.groups() if match else (1, freq)
    #     freq_number = int(freq_number) if freq_number else 1
    #     freq_param = {'M': 'months', 'MS': 'months', 'W': 'weeks', 'D': 'days'}.get(freq_string, freq_string)

    #     try:
    #         offset = pd.DateOffset(**{freq_param: freq_number})
    #     except KeyError:
    #         print("Invalid frequency parameter.")
    #         offset = None

    #     if offset is not None:
    #         start_date_index = date_index[-1] + offset
    #         future_dates = pd.date_range(start=start_date_index, periods=n_predict, freq=offset)
    #     else:
    #         future_dates = pd.date_range(start=date_index[-1], periods=n_predict, freq=freq)

    #     # Concatenate predictions for single feature
    #     if future_predictions is not None:
    #         last_future_prediction = future_predictions.flatten()
    #         combined_predictions = np.concatenate([y_pred_prev_org.flatten(), last_future_prediction])                
    #     else:
    #         combined_predictions = y_pred_prev_org.flatten()

    #     # Combine and normalize dates for plotting
    #     combined_dates = pd.concat([pd.Series(date_index), pd.Series(future_dates)]).reset_index(drop=True)
    #     combined_dates = pd.DatetimeIndex(combined_dates).normalize()

    #     # Plotting for single feature
    #     fig = go.Figure()
    #     fig.add_trace(go.Scatter(x=combined_dates[:len(y_test_org)], y=y_test_org.flatten(), mode='lines', name='Actual'))
    #     fig.add_trace(go.Scatter(x=combined_dates, y=combined_predictions, mode='lines', name='Predicted + Future'))

    #     fig.update_layout(title=f'Actual and Predicted Values for {col_names[0]} ({symbol}_{trial}_{test_dates}_{test_length})', 
    #                         xaxis_title='Time Step', yaxis_title='Value', height=800, width=1500)

    #     if save_directory:
    #         save_path = os.path.join(save_directory, f"predictions_plot_target_{symbol}_{trial}_{test_length}.html")
    #         fig.write_html(save_path)
    # fig.show()

    return evaluation_results

def calculate_metrics_all(y_true: np.ndarray , y_pred: np.ndarray):
    mse = mean_squared_error(y_true=y_true,y_pred=y_pred)
    mae = mean_absolute_error(y_true=y_true,y_pred=y_pred)
    r2 = r2_score(y_true=y_true,y_pred=y_pred)

    return mse, mae, r2

def calculate_metrics(y_true, y_pred, col_names=None):
    
    # Reshape tensors for metrics computation
    y_pred = y_pred.reshape(-1, 1) if y_pred.ndim == 1 else y_pred.reshape(-1, y_pred.shape[-1])    
    y_true = y_true.reshape(-1, 1) if y_true.ndim == 1 else y_true.reshape(-1, y_true.shape[-1])

    col_names = col_names.tolist()

    # # Ensure col_names is a list and has correct length
    # assert isinstance(col_names, list), "col_names must be a list"
    # assert len(col_names) == y_true.shape[-1], "col_names must have same length as number of columns in y_true"

    evaluation_results = {}
    # If there is only one feature, don't loop over features
    if y_true.shape[1] == 1:
        # Correctly handle the case where col_names might be a list
        col_name = col_names[-1] if col_names and len(col_names) > 0 else 'Feature_1'
        
        # Calculate metrics for this column
        mse = mean_squared_error(y_true, y_pred)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)

        evaluation_results[col_name] = {
            'mse': mse,
            'mae': mae,
            'r2': r2,
        }
    else:
        # Iterate over each column in y_true
        for i, col_name in enumerate(col_names):
            # Calculate metrics for this column
            mse = mean_squared_error(y_true[:,i], y_pred[:,i])
            mae = mean_absolute_error(y_true[:,i], y_pred[:,i])
            r2 = r2_score(y_true[:,i], y_pred[:,i])

            # Store results in dictionary
            evaluation_results[col_name] = {
                'mse': mse,
                'mae': mae,
                'r2': r2,
            }
    return evaluation_results
  
def predict_future(model, X_test, y_test_all_features=None, X_test_noshift=None, forward=None, scaler=None):
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
   
    y_pred_prev = model(X_test.to(device)).detach().cpu()

    # Create a copy of X_test to prevent modifying the original data
    X_test_copy = X_test.clone()

    if forward == -1:
        X_test_copy[-1, -1, :] = y_test_all_features[-1:, -1, :]

    elif forward == 0:
        # Convert the last sequence of y_pred_prev to the last sequence of X_test_copy
        X_test_copy[-1, -1, :] = y_test_all_features[-1, -1, :]

    # Predict using X_test_copy
    future_predictions = model(X_test_copy[-1, :, :].to(device)).detach().cpu().numpy()
   
    future_predictions_org = inverse_transform_wrapper(future_predictions, future_predictions.shape, scaler=scaler)
    y_pred_prev_org = inverse_transform_wrapper(y_pred_prev, y_pred_prev.shape, scaler=scaler)

    return future_predictions, future_predictions_org, y_pred_prev, y_pred_prev_org

def compute_future_difference(future_predictions=None, future_predictions_org=None, y_pred_prev=None, y_pred_prev_org=None, y_test=None, 
                              n_predict=None, col_label=None, scaler=None):
    
    n_features = y_test.shape[-1]
      
    y_test_org = inverse_transform_wrapper(y_test, y_test.shape, scaler=scaler)

    y_pred_prev = np.squeeze(y_pred_prev)
    y_pred_prev_org= np.squeeze(y_pred_prev_org)
    future_predictions = np.squeeze(future_predictions)
    future_predictions_org = np.squeeze(future_predictions_org)
    y_test = np.squeeze(y_test)
    y_test_org = np.squeeze(y_test_org)
    
    future_predictions_last = future_predictions[-(n_predict):]
    future_predictions_org_last = future_predictions_org[-(n_predict):]
    y_pred_prev_last = y_pred_prev[-(n_predict):]
    y_pred_prev_org_last = y_pred_prev_org[-(n_predict):]
    y_test_last = y_test[-(n_predict):]
    y_test_org_last = y_test_org[-(n_predict):]

    # Initialize an empty DataFrame to store all the data
    all_df = pd.DataFrame()
    all_org_df = pd.DataFrame()
    np.seterr(divide='ignore', invalid='ignore')
    # for i in range(n_features):
    #     # Compute the difference

    #     # Create DataFrames for future_predictions and y_test
    #     future_predictions_last_df = pd.DataFrame(future_predictions_last[:, i], columns=['Future_' + col_label[i]])
    #     future_predictions_org_last_df = pd.DataFrame(future_predictions_org_last[:, i], columns=['Future_' + col_label[i]])
    #     y_pred_prev_last_df = pd.DataFrame(y_pred_prev_last[:, i], columns=['Last_Predict_' + col_label[i]])
    #     predictions_org_last_df = pd.DataFrame(y_pred_prev_org_last[:, i], columns=['Last_Predict_' + col_label[i]])
    #     y_test_last_df = pd.DataFrame(y_test_last[:, i], columns=['Last_Actual_' + col_label[i]])
    #     y_test_org_last_df = pd.DataFrame(y_test_org_last[:, i], columns=['Last_Actual_' + col_label[i]])

    #     error = np.abs(np.subtract(y_pred_prev_last[:, i], y_test_last[:, i]))
    #     error_org = np.abs(np.subtract(y_pred_prev_org_last[:, i], y_test_org_last[:, i]))
    #     error_df = pd.DataFrame(error, columns=['Last_Predict_Error_' + col_label[i]])
    #     error_org_df = pd.DataFrame(error_org, columns=['Last_Predict_Error_' + col_label[i]])

    #     # Concatenate the DataFrames
    #     feature_org_df = pd.concat([y_test_org_last_df, predictions_org_last_df, error_org_df, future_predictions_org_last_df], axis=1)
    #     feature_df = pd.concat([y_test_last_df, y_pred_prev_last_df, error_df, future_predictions_last_df], axis=1)

    #     # Append feature_df to all_df
    #     all_org_df = pd.concat([all_org_df, feature_org_df], axis=1)
    #     all_df = pd.concat([all_df, feature_df], axis=1)

    for i in range(n_features):
        # Assuming col_label is a list of strings
        column_name = col_label[-1] if not col_label.empty and len(col_label) > i else f'Feature_{-1}'

        # Create DataFrames without indexing a second dimension
        future_predictions_last_df = pd.DataFrame(future_predictions_last, columns=['Future_' + column_name])
        future_predictions_org_last_df = pd.DataFrame(future_predictions_org_last, columns=['Future_' + column_name])
        y_pred_prev_last_df = pd.DataFrame(y_pred_prev_last, columns=['Last_Predict_' + column_name])
        predictions_org_last_df = pd.DataFrame(y_pred_prev_org_last, columns=['Last_Predict_' + column_name])
        y_test_last_df = pd.DataFrame(y_test_last, columns=['Last_Actual_' + column_name])
        y_test_org_last_df = pd.DataFrame(y_test_org_last, columns=['Last_Actual_' + column_name])

        error = np.abs(np.subtract(y_pred_prev_last, y_test_last))
        error_org = np.abs(np.subtract(y_pred_prev_org_last, y_test_org_last))
        error_df = pd.DataFrame(error, columns=['Last_Predict_Error_' + column_name])
        error_org_df = pd.DataFrame(error_org, columns=['Last_Predict_Error_' + column_name])

        # Concatenate the DataFrames
        feature_org_df = pd.concat([y_test_org_last_df, predictions_org_last_df, error_org_df, future_predictions_org_last_df], axis=1)
        feature_df = pd.concat([y_test_last_df, y_pred_prev_last_df, error_df, future_predictions_last_df], axis=1)

        # Append feature_df to all_df
        all_org_df = pd.concat([all_org_df, feature_org_df], axis=1)
        all_df = pd.concat([all_df, feature_df], axis=1)

    return all_df, all_org_df, future_predictions_last, future_predictions_org_last

def add_columns(df, item, k, file_name, l, t):
    # Create a DataFrame for the additional columns
    add_df = pd.DataFrame({'Symbol': [item]*len(df), 
                           'Symbol_ID': [k + 1]*len(df), 
                           'Model': [file_name]*len(df), 
                           'Model_ID': [l + 1]*len(df),
                            'test_dates': [t + 1]*len(df)})

    # Concatenate the additional columns with the original DataFrame
    return pd.concat([add_df, df], axis=1)

def select_best_model_for_future(data):

    data = data.copy()
    
    # Ascending order by "Symbol_ID", "test_dates", "Future_error_Return"
    data.sort_values(by=["Symbol_ID", "test_dates", "Last_Predict_Error_Target"], inplace=True)

    # Find the index of minimum "Future_error_Return" for each group of "Symbol_ID" and "test_dates"
    idx = data.groupby(["Symbol_ID", "test_dates"])["Last_Predict_Error_Target"].idxmin()

    # Select rows with the index
    data = data.loc[idx]

    # Copy a column "Last_Actual_Return" with a column name = "Future_Actual_Return"
    data["Future_Actual_Target"] = data["Last_Actual_Target"].values

    # Shift(-1) the data by a column "Future_Actual_Return"
    data["Future_Actual_Target"] = data["Future_Actual_Target"].shift(1)
    data['Future_Actual_Target'] = data['Future_Actual_Target'].fillna(0)

    # Delete rows of a column "test_dates" when "test_dates" = maximum of value
    if len(data["test_dates"].unique()) > 1:
        data = data[data["test_dates"] != data["test_dates"].min()]

    # Add a column "Last_Actual_Return_ABS" = absolute value of "Last_Actual_Return"
    data["Future_Actual_Target_ABS"] = abs(data["Future_Actual_Target"])
        
    # Add a column "Future_error_Return" = absolute value of "Future_Actual_Return" - "Future_Return"
    data["Future_Error_Target"] = abs(data["Future_Actual_Target"] - data["Future_Target"])

    # Add a column "Direction_Test"
    data["Direction_Test"] = np.where(
        ((data["Future_Actual_Target"] > 0) & (data["Future_Target"] > 0)) |
        ((data["Future_Actual_Target"] < 0) & (data["Future_Target"] < 0)) |
        (data["Future_Error_Target"] < 0.001),
        1, 0
    )

    # # Ascending order by "Symbol_ID", "test_dates", "Future_error_Return"
    # data.sort_values(by=["Symbol_ID", "test_dates", "Last_Predict_Error_Return"], inplace=True)

    # # Find the index of minimum "Future_error_Return" for each group of "Symbol_ID" and "test_dates"
    # idx = data.groupby(["Symbol_ID", "test_dates"])["Last_Predict_Error_Return"].idxmin()

    # # Select rows with the index
    # data = data.loc[idx]

    # Columns alignment
    cols = ["Symbol", "Symbol_ID", "Model", "Model_ID", "test_dates", "Future_Actual_Target", "Future_Actual_Target_ABS", "Last_Actual_Target", 
            "Last_Predict_Target", "Last_Predict_Error_Target", "Future_Target", "Future_Error_Target", "Direction_Test"]  + [col for col in data if col not in [
                "Symbol", "Symbol_ID", "Model", "Model_ID", "test_dates", "Future_Actual_Target", "Future_Actua_Target_ABS", "Last_Actual_Target", 
            "Last_Predict_Target", "Last_Predict_Error_Target", "Future_Target", "Future_Error_Target", "Direction_Test"]]
    data = data[cols]

    return data

def select_best_model_for_predict(data):

    data = data.copy()

    # Add a column "Last_Actual_Return_ABS" = absolute value of "Last_Actual_Return"
    data["Last_Actual_Target_ABS"] = abs(data["Last_Actual_Target"])

    # Add a column "Direction_Test"
    data["Direction_Test"] = np.where(
        ((data["Last_Actual_Target"] > 0) & (data["Last_Predict_Target"] > 0)) |
        ((data["Last_Actual_Target"] < 0) & (data["Last_Predict_Target"] < 0)) |
        (data["Last_Predict_Error_Target"] < 0.001),
        1, 0
    )

    # Ascending order by "Symbol_ID", "test_dates", "Predict_error_Return"
    data.sort_values(by=["Symbol_ID", "test_dates", "Last_Predict_Error_Target"], inplace=True)

    # Find the index of minimum "Future_error_Return" for each group of "Symbol_ID" and "test_dates"
    idx = data.groupby(["Symbol_ID", "test_dates"])["Last_Predict_Error_Target"].idxmin()

    # Select rows with the index
    data = data.loc[idx]

    # Columns alignment
    cols = ["Symbol", "Symbol_ID", "Model", "Model_ID", "test_dates", "Last_Actual_Target", "Last_Actual_Target_ABS", 
            "Last_Predict_Target", "Last_Predict_Error_Target", "Future_Target", "Direction_Test"]  + [col for col in data if col not in [
                "Symbol", "Symbol_ID", "Model", "Model_ID", "test_dates", "Last_Actual_Target", "Last_Actual_Target_ABS", 
            "Last_Predict_Target", "Last_Predict_Error_Target", "Future_Target", "Direction_Test"]]
    data = data[cols]

    return data

def select_best_metrics(data):

    data = data.copy()

    # Ascending order by "Symbol_ID", "test_dates", "Future_error_Return"
    data.sort_values(by=["Symbol_ID", "test_dates", "Target Test MAE"], inplace=True)

    # Find the index of minimum "Future_error_Return" for each group of "Symbol_ID" and "test_dates"
    idx = data.groupby(["Symbol_ID", "test_dates"])["Target Test MAE"].idxmin()

    # Select rows with the index
    data = data.loc[idx] 

    return data

def loaded_model(target_col=None, save_directory=None, load_directory=None, future_predictions=None, stock_price_plots_use=False, use_target_col=True, 
                  metrics_use=True, plots_use=False, symbol_list=None, start_date=None, interval=None,  freq=None, n_predict=1, n_last_sequence=None, 
                  n_last_sequence_long=None, test_dates=None, date_range=None, seq_len=2, forward=-1, train_size=None, united_transform=None, indicator_ver=None):

    if save_directory and not os.path.exists(save_directory):
        os.makedirs(save_directory)
    
    # Initialize dictionaries to store results for each symbol
    all_results_metrics_dict = {}
    all_future_predictions_dict = {}
    all_future_predictions_org_dict = {}        
   
    # delist=fdr.StockListing('KRX-DELISTING')
    
    # if stock_code_list is not None:
    #     iterable = stock_code_list
    # else:
    #     iterable = symbol_list
    
    all_data = fetch_data(symbol_list, start_date=start_date, interval=interval)
    
    # all_data_short = fetch_data(symbol_list, period=period_short, interval=interval)
        
    for k, symbol in enumerate(symbol_list):
        
        print(f'Symbol or Code: {symbol}')
        
        start = timeit.default_timer()
        
        # if (delist['Symbol'].isin([item])).any():
        #     print(f"Symbol {item} is delisted. Skipping...")
        #     continue
            
        # if iterable is symbol_list:
        #     symbol = item
        #     stock_code = None
        # elif iterable is stock_code_list:
        #     symbol = None
        #     stock_code = item
            
        result = prepare_data(all_data=all_data, symbol=symbol, seq_len=seq_len, target_col=target_col, interval=interval, forward=forward, 
                              train_size=0.8, united_transform=united_transform, indicator_ver=indicator_ver)
        
        # If prepare_data returned None, skip this iteration
        if result is None:
            print(f"Not enough data for {symbol}. Skipping...") 
            continue
        
        test_data, test_data_unnormalized, test_data_noshift, seq_len = result
        
        if test_data.shape[0] < n_last_sequence_long + seq_len + max(test_dates):
            print(f"Not enough data for {symbol}. Skipping...")            
            continue        
        
        X_test_original, y_test_original, y_test_all_features, X_test_noshift = prepare_data_tensor(test_data=test_data, test_data_noshift=test_data_noshift, seq_len=seq_len)
        
        X_test_copy = X_test_original.clone()
        y_test_copy = y_test_original.clone()
        y_test_all_features_copy = y_test_all_features.clone()
        X_test_noshift_copy = X_test_noshift.clone()

        all_future_predictions = []
        all_future_predictions_org = []
        all_results_metrics = []
        
        # Get a list of all files in the directory
        file_names = os.listdir(load_directory)
        
        # Iterate over each file name
        for l, file_name in enumerate(file_names):

            # Construct full path to each model file
            model_path = os.path.join(load_directory, file_name)
            # Load the model object
            loaded_model = torch.load(model_path)

            device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
            # Move the model to device
            loaded_model = loaded_model.to(device)
            loaded_model.eval()

            # Inverse transform the y_test to the original scale
            test_scaler = StandardScaler().fit(test_data_unnormalized)
            test_scaler_target_col = StandardScaler().fit(test_data_unnormalized['Target'].values.reshape(-1, 1))

            # Get the column names
            col_label = test_data_unnormalized.columns

            # min_date = min(date_range)
            # max_date = max(date_range)

            # for t in test_dates:
            #     if t == max_date:
            #         X_test = X_test_copy
            #         y_test = y_test_copy
            #         y_test_all_features = y_test_all_features_copy
            #         y_test_noshift = y_test_noshift_copy
            #     else:
            #         if min_date > 0:
            #             X_test = X_test_copy[:-(max_date - (t - min_date))]
            #             y_test = y_test_copy[:-(max_date - (t - min_date))]
            #             y_test_all_features = y_test_all_features_copy[:-(max_date - (t - min_date))]
            #             y_test_noshift = y_test_noshift_copy[:-(max_date - (t - min_date))]
            #         else:
            #             X_test = X_test_copy[:-(max_date - t)]
            #             y_test = y_test_copy[:-(max_date - t)]
            #             y_test_all_features = y_test_all_features_copy[:-(max_date - t)]
            #             y_test_noshift = y_test_noshift_copy[:-(max_date - t)]

            for t in test_dates:
                if t == 0:
                    X_test = X_test_copy
                    y_test = y_test_copy
                    y_test_all_features = y_test_all_features_copy
                    X_test_noshift = X_test_noshift_copy
                else:
                    X_test = X_test_copy[:-t]
                    y_test = y_test_copy[:-t]
                    y_test_all_features = y_test_all_features_copy[:-t]
                    X_test_noshift = X_test_noshift_copy[:-t]

                # Inverse transform the y_test to the original scale
                y_test_org = inverse_transform_wrapper(y_test, y_test.shape, scaler=test_scaler_target_col)
                
                # Generate future predictions
                future_predictions, future_predictions_org, y_pred_prev, y_pred_prev_org = predict_future(loaded_model, X_test, y_test_all_features=y_test_all_features, 
                                                                                                          X_test_noshift=X_test_noshift, scaler=test_scaler_target_col,forward=forward)
                
                future_difference, future_difference_org, future_predictions_last, future_predictions_org_last = compute_future_difference(future_predictions=future_predictions, future_predictions_org=future_predictions_org, 
                                                                                                                                           y_pred_prev=y_pred_prev, y_pred_prev_org=y_pred_prev_org, y_test=y_test, n_predict=n_predict, col_label=col_label, scaler=test_scaler_target_col)

                future_predictions = np.squeeze(future_predictions)
                future_predictions_org = np.squeeze(future_predictions_org)            
                
                # Add the additional columns
                future_difference = add_columns(future_difference, symbol, k, file_name, l, t)

                future_difference_org = add_columns(future_difference_org, symbol, k, file_name, l, t)

                # Append the new DataFrame to the list
                all_future_predictions.append(future_difference)

                all_future_predictions_org.append(future_difference_org)       

                # Concatenate all the future predictions into a single DataFrame
                all_future_predictions_df = pd.concat(all_future_predictions, axis=0)

                all_future_predictions_org_df = pd.concat(all_future_predictions_org, axis=0)  

                # Evaluation and Plot prediction results

                test_metrics_1 = evaluate_and_plot_predictions(y_test[-(n_last_sequence_long):], y_test_org[-(n_last_sequence_long):], y_pred_prev[-(n_last_sequence_long):], y_pred_prev_org[-(n_last_sequence_long):], 
                                                             trial=file_name, n_predict=n_predict, data=test_data_unnormalized[-(n_last_sequence_long):], seq_len=seq_len, use_target_col=use_target_col,
                                                            metrics_use=metrics_use, plots_use=plots_use, save_directory=save_directory, symbol=symbol, col_names=col_label, test_dates=t, test_length="long period", freq=freq,
                                                            future_predictions=future_predictions_org_last)
                
                # test_metrics_short = evaluate_and_plot_predictions(y_test[-(n_last_sequence):], y_test_org[-(n_last_sequence):], y_pred_prev[-(n_last_sequence):], y_pred_prev_org[-(n_last_sequence):], 
                #                                                    trial=file_name, n_predict=n_predict, data=test_data_unnormalized[-(n_last_sequence):], seq_len=seq_len, use_target_col=use_target_col,
                #                                                 metrics_use=metrics_use, plots_use=False, save_directory=save_directory, symbol=symbol, col_names=col_label, test_length="short period", freq=freq,
                #                                                 future_predictions=future_predictions_org_last)
                
                # future_metrics = evaluate_and_plot_predictions(y_test[1:], y_test_org[1:], future_predictions[:-1], future_predictions_org[:-1], trial=file_name, n_predict=n_predict, data=test_data_unnormalized, seq_len=seq_len, use_target_col=use_target_col,
                #                 metrics_use=metrics_use, plots_use=False, save_directory=save_directory, symbol=symbol, col_names=col_label, test_length="future_prediction", freq=freq,
                #                 future_predictions=future_predictions_org_last)

                # future_metrics = evaluate_and_plot_predictions(y_test[-(n_last_sequence_long)+1:], y_test_org[-(n_last_sequence_long)+1:], future_predictions[-(n_last_sequence_long):-1], future_predictions_org[-(n_last_sequence_long):-1],
                #                                             trial=file_name, n_predict=n_predict, data=test_data_unnormalized[-(n_last_sequence_long)+1:], seq_len=seq_len, use_target_col=use_target_col,
                #                                             metrics_use=metrics_use, plots_use=plots_use, save_directory=save_directory, symbol=symbol, col_names=col_label, test_dates=t, test_length="future prediction long", freq=freq,
                #                                             future_predictions=future_predictions_org_last) 

                # future_metrics_short = evaluate_and_plot_predictions(y_test[-(n_last_sequence)+1:], y_test_org[-(n_last_sequence)+1:], future_predictions[-(n_last_sequence):-1], future_predictions_org[-(n_last_sequence):-1],
                #                                                       trial=file_name, n_predict=n_predict, data=test_data_unnormalized[-(n_last_sequence)+1:], seq_len=seq_len, use_target_col=use_target_col,
                #                                                     metrics_use=metrics_use, plots_use=False, save_directory=save_directory, symbol=symbol, col_names=col_label, test_length="future prediction short", freq=freq,
                #                                                     future_predictions=future_predictions_org_last)

                test_metrics = calculate_metrics(y_test[-(n_last_sequence_long):], y_pred_prev[-(n_last_sequence_long):],col_names=col_label)

                test_metrics_org = calculate_metrics(y_test_org[-(n_last_sequence_long):], y_pred_prev_org[-(n_last_sequence_long):],col_names=col_label)

                # test_metrics_1 = calculate_metrics(y_test[-(n_last_sequence_long):-1], y_pred_prev[-(n_last_sequence_long):-1],col_names=col_label)
                
                # test_metrics_2 = calculate_metrics(y_test[-(n_last_sequence_long):-2], y_pred_prev[-(n_last_sequence_long):-2],col_names=col_label)
                
                # test_metrics_3 = calculate_metrics(y_test[-(n_last_sequence_long):-3], y_pred_prev[-(n_last_sequence_long):-3],col_names=col_label)   
                    
                results_metrics = {"Symbol": symbol, "Symbol_ID": k+1,  "Model": file_name, "Model_ID": l+1 , "test_dates": t+1}

                # Add metrics for each feature

                for col_name, metrics in test_metrics.items():
                    results_metrics[f"{col_name} Test MSE"] = np.round(metrics['mse'], 5)
                    results_metrics[f"{col_name} Test MAE"] = np.round(metrics['mae'], 5)
                    results_metrics[f"{col_name} Test R2"] = np.round(metrics['r2'], 5)

                for col_name, metrics in test_metrics_org.items():
                    results_metrics[f"{col_name} Test Org MSE"] = np.round(metrics['mse'], 5)
                    results_metrics[f"{col_name} Test Org MAE"] = np.round(metrics['mae'], 5)
                    results_metrics[f"{col_name} Test Org R2"] = np.round(metrics['r2'], 5)

                # for col_name, metrics in test_metrics_1.items():
                #     results_metrics[f"{col_name} Test_1 MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Test_1 MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Test_1 R2"] = np.round(metrics['r2'], 5)

                # for col_name, metrics in test_metrics_2.items():
                #     results_metrics[f"{col_name} Test_2 MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Test_2 MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Test_2 R2"] = np.round(metrics['r2'], 5)

                # for col_name, metrics in test_metrics_3.items():
                #     results_metrics[f"{col_name} Test_3 MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Test_3 MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Test_3 R2"] = np.round(metrics['r2'], 5)
                    
                # for col_name, metrics in test_metrics_short.items():
                #     results_metrics[f"{col_name} Test Short MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Test Short MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Test Short R2"] = np.round(metrics['r2'], 5)

                # for col_name, metrics in future_metrics.items():
                #     results_metrics[f"{col_name} Future MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Future MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Future R2"] = np.round(metrics['r2'], 5)

                # for col_name, metrics in future_metrics_short.items():
                #     results_metrics[f"{col_name} Future Short MSE"] = np.round(metrics['mse'], 5)
                #     results_metrics[f"{col_name} Future Short MAE"] = np.round(metrics['mae'], 5)
                #     results_metrics[f"{col_name} Future Short R2"] = np.round(metrics['r2'], 5)

                all_results_metrics.append(results_metrics)

                all_results_metrics_df = pd.concat([pd.DataFrame(data=d, index=[0]) for d in all_results_metrics], axis=0)            

                if stock_price_plots_use:

                    stock_price_plots(test_data_unnormalized[-(n_last_sequence_long):], y_pred_org=y_pred_prev_org[-(n_last_sequence_long):], y_test_org=y_test_org[-(n_last_sequence_long):],
                                    file_name=file_name, n_predict=n_predict, seq_len=seq_len, save_directory=save_directory, symbol=symbol, future_predictions=future_predictions_org_last, 
                                    scaler=test_scaler, col_names=col_label, test_length="long period", freq=freq)  


                    # stock_price_plots(test_data_unnormalized[-n_last_sequence:], y_pred_org=y_pred_prev_org[-n_last_sequence:], y_test_org=y_test_org[-n_last_sequence:],
                    #                   file_name=file_name, n_predict=n_predict, seq_len=seq_len, save_directory=save_directory, symbol=symbol, future_predictions=future_predictions_org_last, 
                    #                   scaler=test_scaler, col_names=col_label, test_length="short period", freq=freq) 
                
        end = timeit.default_timer()
        # Calculate and print duration
        duration = end - start
        print(f"Execution Time of Symbol_{symbol} is: {duration} seconds")
        # Save results for each symbol in the result dictionaries
        all_results_metrics_dict[symbol] = all_results_metrics_df
        all_future_predictions_dict[symbol] = all_future_predictions_df
        all_future_predictions_org_dict[symbol] = all_future_predictions_org_df

    return all_results_metrics_dict, all_future_predictions_dict, all_future_predictions_org_dict
```

```python
import csv
import datetime

# # Weekly Data
# symbol_list=[
#              "AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META", "BRK-B", "TSLA", "LLY", "V", "TSM", "UNH", "NVO", "JPM", "WMT", "XOM", "AVGO", "MA", "JNJ", "PG", "ORCL", 
#              "HD", "ADBE", "ASML", "CVX", "MRK", "TM", "COST", "KO", "ABBV", "BAC", "PEP", "FMX", "CRM", "SHEL", "ACN", "NFLX", "BABA", "MCD", "LIN", "AZN", "CSCO", "AMD", 
#              "NVS", "INTC", "TMO", "ABT", "SAP", "DIS", "TMUS", "CMCSA", "PFE", "NKE", "TTE", "INTU", "DHR", "WFC", "BHP", "VZ", "PDD", "HSBC", "HDB", "AMGN", "QCOM", "PM", 
#              "IBM", "TXN", "NOW", "UNP", "COP", "MS", "AMAT", "GE", "CAT", "SPGI", "BX", "HON", "UPS", "BA", "SBUX", "RY", "BUD", "UL", "LOW", "SNY", "AXP", "RTX", "NEE", 
#              "T", "UBER", "GS", "TD", "SYK", "LMT", "RIO", "BKNG", "DE", "ELV", "SONY", "ISRG", "BLK", "BMY", "PLD", "PBR", "TJX", "MUFG", "MDT", "BP", "SCHW", "MMC", 
#              "MDLZ", "PBR-A", "ADP", "EQNR", "PGR", "GILD", "AMT", "LRCX", "CB", "ETN", "ADI", "VRTX", "CVS", "SHOP", "C", "REGN", "MU", "CI", "UBS", "SNPS", "ZTS", "ABNB",
#              "BSX", "DEO", "IBN", "SO", "CME", "PANW", "FI", "CNI", "SLB", "MELI", "KLAC", "KKR", "EQIX", "CDNS", "NTES", "MO", "ITW", "INFY", "GSK", "EOG", "NOC", "ENB",
#              "BTI", "CNQ", "DUK", "RELX", "WM", "SHW", "BDX", "GD", "AON", "CP", "ANET", "VALE", "MCO", "VMW", "HCA", "SMFG", "FDX", "RACE", "SAN", "HUM", "ICE", "CL", "TRI",
#              "CSX", "PYPL", "STLA", "CHTR", "APD", "ITUB", "WDAY", "MCK", "TGT", "MAR", "CMG", "USB", "ORLY", "SCCO", "BMO", "MNST", "EPD", "CTAS", "ROP", "BN", "AMX",
#              "MPC", "PH", "TDG", "PXD", "APH", "OXY", "LULU", "SNOW", "BNS", "AJG", "MMM", "ECL", "MSI", "BBVA", "E", "DELL", "NXPI", "TT", "PNC", "FCX", "EMR", "RSG", "PSX",
#              "HMC", "APO", "CRWD", "BSBR", "TEAM", "ING", "PCAR", "AFL", "CPRT", "MRVL", "NSC", "AZO", "NGG", "WELL", "ADSK", "MET", "PSA", "SRE", "SPG", "CCI", "PCG", "AIG",
#              "MCHP", "KDP", "CARR", "JD", "ABEV", "EL", "HLT", "ODFL", "STZ", "HES", "DHI", "TAK", "PLTR", "WMB", "MFG", "SU", "PAYX", "TFC", "CRH", "VLO", "KHC", "MSCI", "ET",
#              "LNG", "KMB", "STM", "TEL", "DLR", "EW", "AEP", "ROST", "SGEN", "HSY", "COF", "DXCM", "F", "FTNT", "COR", "GWW", "ADM", "TRV", "EXC", "D", "CNC", "LI", "IDXX",
#              "NUE", "WDS", "BIDU", "OKE", "DASH", "GIS", "GM", "O", "LVS", "KMI", "IQV", "EA", "TRP", "DOW", "CM", "BCE", "YUM", "LEN", "BK", "SYY", "AME",
#              "MPLX", "DDOG", "AMP", "JCI", "LHX", "ALL", "WCN", "FAST", "SQ", "VRSK", "OTIS", "CTSH", "MFC", "ALC", "SPOT", "PRU", "BBD", "FERG", "CSGP", "BKR", "XEL", "A", "CTVA",
#              "HAL", "BIIB", "LYG", "CVE", "IT", "IBKR", "TTD", "URI", "FIS", "PPG", "QSR", "PEG", "CMI", "ORAN", "PUK", "ED", "ROK", "ACGL", "LYB", "ARES",
#              "NDAQ", "IMO", "KR", "DD", "ON", "VICI", "NEM", "SLF", "MRNA", "GPN", "CDW", "ARGX", "CHT", "MLM", "EXR", "ELP", "IR", "DVN", "CPNG", "NTR", "VMC", "VEEV", "BF-A",
#              "CQP", "HPQ", "GOLD", "MDB", "CCEP", "ZS", "PKX", "FANG", "PWR", "RCL", "TTWO", "ANSS", "BCS", "RYAAY", "WEC", "DG", "MPWR", "TU", "FICO", "SBAC", "AWK", "WST", "CAH",
#              "SPLK", "EIX", "WBD", "WTW", "DLTR", "VOD", "EFX", "XYL", "WIT", "EC", "GLW", "AEM", "AVB", "GIB", "BNTX", "DB", "CBRE", "FTV", "NET", "RBLX", "KEYS", "HUBS", "ZBH",
#              "FNV", "COIN", "WY", "TEF", "HIG", "MTD", "CHD", "DAL", "TCOM", "GRMN", "TLK", "APTV", "RCI", "ICLR", "TSCO", "NWG", "RMD", "TROW", "TW", "SE", "RJF", "FCNCA", "EBAY",
#              "STT", "VRSN", "BR", "DTE", "MOH", "FE", "HWM", "ALNY", "PINS", "DFS", "EQR", "LIFW", "ETR", "GMAB", "BRO", "WPM", "HEI-A", "HEI", "IX", "WAB", "INVH", "LYV", "MTB",
#              "ULTA", "NVR", "HPE", "AEE", "STE", "ES", "FTS", "BGNE", "SYM", "CTRA", "SNAP", "TS", "SIRI", "MT", "ZTO", "NOK", "CCL", "GPC", "DOV", "BEKE", "ROL", "PPL", "PHG",
#              "OWL", "ZM", "CCJ", "IFF", "UMC", "YUMC", "DRI", "EBR", "PHM", "TRGP", "AIU", "FITB", "JBHT", "TDY", "ASX", "MKL", "STLD", "PTC", "IRM", "TECK", "ARE", "EXPE", "BAX",
#              "WRB", "VTR", "WBA", "LH", "K", "HRL", "KOF", "HOLX", "RKT", "MKC", "FDS", "TYL", "UWMC", "PBA", "CNP", "EXPD", "DKNG", "EG", "CLX", "VIV", "TSN", "FLT", "PFG",
#              "ATO", "AKAM", "COO", "CMS", "J", "JBL", "LPLA", "BAH", "AXON", "BEP", "FSLR", "EQT", "CHKP", "BMRN", "WLK", "BALL", "VRT", "WMG", "ERIC", "BLDR", "NTAP", "RPRX", "SUI",
#              "CINF", "DECK", "STX", "KB", "HBAN", "WAT", "BSY", "RS", "MGA", "ALGN", "HUBB", "ENTG", "NTRS", "BG", "CF", "OMC", "AVY", "DGX", "WSO", "RF", "ILMN", "TXT", "DT",
#              "SMCI", "AMH", "L", "IEX", "WDC", "SWKS", "FWONA", "EPAM", "ERIE", "SNA", "TME", "BBY", "MRO", "RTO", "LUV", "LII", "LDOS", "ALB", "SHG", "AER", "MAA", "MGM",
#              "SUZ", "XPEV", "JHX", "ESS", "PKG", "TER", "FOXA", "CE", "LW", "SWK", "ELS", "MANH", "SSNC", "APP", "SQM", "POOL", "GGG", "CAG", "AMCR", "CBOE", "MAS", "LOGI",
#              "CSL", "CNHI", "NDSN", "NIO", "RPM", "DPZ", "AVTR", "GEN", "GDDY", "UAL", "VST", "ROKU", "XP", "TAP", "LNT", "CFG", "BIP", "IHG", "GRAB", "ENPH", "TPL", "NICE",
#              "NMR", "AGR", "CPB", "MOS", "LBRDK", "PODD", "WPC", "BEN", "GLPI", "NWSA", "INCY", "FNF", "ACI", "ACM", "LKQ", "HST", "OVV", "SYF", "RYAN", "UDR", "H", "KIM",
#              "IP", "AZPN", "LEGN", "GFI", "EVRG", "MORN", "ZI", "OC", "FLEX", "HTHT", "IPG", "OKTA", "EDU", "SJM", "AES", "CG", "KEY", "FMS", "PSTG", "AOS", "RBA", "ARCC", "JKHY",
#              "SNN", "TWLO", "LECO", "GL", "GFL", "REG", "U", "ZBRA", "EXAS", "RDY", "WSM", "VTRS", "APA", "TRU", "RVTY", "SAIA", "BCH", "CNA", "NRG", "RNR", "NI", "UTHR", "CHK",
#              "TRMB", "NBIX", "WES", "CDAY", "CASY", "USFD", "RGA", "PNR", "AEG", "OTEX", "PAG", "PATH", "KMX", "UHAL", "YPF", "CCK", "TFX", "XPO", "FIVE", "PAYC",
#              "CX", "TFII", "REXR", "BAP", "EME", "LAMR", "CZR", "TEVA", "DKS", "WYNN", "DOX", "LCID", "CRL", "EMN", "COTY", "WRK", "TECH", "PFGC", "FFIV", "CHRW", "EQH", "FND",
#              "DINO", "NTNX", "Z", "WPP", "CPT", "WMS", "WTRG", "HII", "UHS", "ALLE", "BLD", "AFG", "TOL", "QRVO", "SBS", "PEAK", "BJ", "BRKR", "QGEN", "CHDN", "HSIC", "DBX", "TTC",
#              "CLH", "ITT", "CUBE", "SCI", "BSAC", "BIO", "CHE", "SNX", "NLY", "KEP", "FTI", "ESLT", "NVT", "GGB", "EWBC", "TTEK", "AGCO", "BXP", "DOCU", "DVA", "DUOL", "MEDP", "ETSY",
#              "ONON", "SKM", "CLF", "CHWY", "PEN", "MTCH", "JNPR", "MKTX", "MTN", "RGEN", "PARA", "RHI", "TIMB", "AIZ", "MNSO", "BURL", "ALV", "DLB", "ATR", "PCTY", "PSO", "FBIN",
#              "KNX", "FUTU", "VIPS", "PNW", "UNM", "ALLY", "KNSL", "MNDY", "LSCC", "RRC", "CW", "WWD", "ORI", "PCOR", "SSL", "SKX", "AAL", "BWA", "GWRE", "RL", "CYBR",
#              "STN", "FCN", "EGP", "LNW", "LEA", "GLOB", "OHI", "SRPT", "MUSA", "CNM", "COLD", "OLED", "JAZZ", "X", "SEIC", "AR", "GRFS", "PR", "RRX", "FRT", "WCC", "BERY",
#              "RGLD", "WSC", "YMM", "AFRM", "SWN", "HESM", "EDR", "WEX", "WF", "NOV", "LAD", "TXRH", "JLL", "VOYA", "WBS", "JEF", "PRI", "NYT", "ESTC", "HOOD", "DCI", "ARMK",
#              "NFE", "HLI", "AU", "BZ", "TX", "CACI", "TREX", "CTLT", "CCCS", "RMBS", "GNTX", "NNN", "APPF", "MSTR", "IEP", "BWXT", "KBR", "ROIV", "OGE", "UI", "RBC", "CSAN",
#              "GNRC", "IONS", "FIX", "TPR", "DSGX", "PAC", "APG", "BBWI", "GPK", "ASR", "ACHC", "INGR", "UFPI", "FHN", "AXTA", "GXO", "FSV", "TPX", "CAR", "ARW",
#              "WING", "MAT", "MTDR", "VFC", "EXEL", "KGC", "FMC", "CAE", "WFRD", "ONTO", "DAR", "MUR", "LBTYK", "SSD", "CIEN", "MIDD", "PSN", "NYCB", "CIB", "CIG", "HRB", "COKE",
#              "BILL", "MSA", "CHRD", "ICL", "TKO", "GMED", "BRX", "FLR", "QLYS", "WTS", "SOFI", "SWAV", "WFG", "STAG", "EHC", "AMKR", "CIVI", "WH", "KRTX", "STWD", "KT"]
DOW_symbol_list=[
             "AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META", "BRK-B", "TSLA", "LLY", "V", "TSM", "UNH", "NVO", "JPM", "WMT", "XOM", "AVGO", "MA", "JNJ", "PG", "ORCL", 
             "HD", "ADBE", "ASML", "CVX", "MRK", "TM", "COST", "KO", "ABBV", "BAC", "PEP", "FMX", "CRM", "SHEL", "ACN", "NFLX", "BABA", "MCD", "LIN", "AZN", "CSCO", "AMD", 
             "NVS", "INTC", "TMO", "ABT", "SAP", "DIS", "TMUS", "CMCSA", "PFE", "NKE", "TTE", "INTU", "DHR", "WFC", "BHP", "VZ", "PDD", "HSBC", "HDB", "AMGN", "QCOM", "PM", 
             "IBM", "TXN", "NOW", "UNP", "COP", "MS", "AMAT", "GE", "CAT", "SPGI", "BX", "HON", "UPS", "BA", "SBUX", "RY", "BUD", "UL", "LOW", "SNY", "AXP", "RTX", "NEE", 
             "T", "UBER", "GS", "TD", "SYK", "LMT", "RIO", "BKNG", "DE", "ELV", "SONY", "ISRG", "BLK", "BMY", "PLD", "PBR", "TJX", "MUFG", "MDT", "BP", "SCHW", "MMC", 
             "MDLZ", "PBR-A", "ADP", "EQNR", "PGR", "GILD", "AMT", "LRCX", "CB", "ETN", "ADI", "VRTX", "CVS", "SHOP", "C", "REGN", "MU", "CI", "UBS", "SNPS", "ZTS", "ABNB",
             "BSX", "DEO", "IBN", "SO", "CME", "PANW", "FI", "CNI", "SLB", "MELI", "KLAC", "KKR", "EQIX", "CDNS", "NTES", "MO", "ITW", "INFY", "GSK", "EOG", "NOC", "ENB",
             "BTI", "CNQ", "DUK", "RELX", "WM", "SHW", "BDX", "GD", "AON", "CP", "ANET", "VALE", "MCO", "VMW", "HCA", "SMFG", "FDX", "RACE", "SAN", "HUM", "ICE", "CL", "TRI",
             "CSX", "PYPL", "STLA", "CHTR", "APD", "ITUB", "WDAY", "MCK", "TGT", "MAR", "CMG", "USB", "ORLY", "SCCO", "BMO", "MNST", "EPD", "CTAS", "ROP", "BN", "AMX",
             "MPC", "PH", "TDG", "PXD", "APH", "OXY", "LULU", "SNOW", "BNS", "AJG", "MMM", "ECL", "MSI", "BBVA", "E", "DELL", "NXPI", "TT", "PNC", "FCX", "EMR", "RSG", "PSX",
             "HMC", "APO", "CRWD", "BSBR", "TEAM", "ING", "PCAR", "AFL", "CPRT", "MRVL", "NSC", "AZO", "NGG", "WELL", "ADSK", "MET", "PSA", "SRE", "SPG", "CCI", "PCG", "AIG",
             "MCHP", "KDP", "CARR", "JD", "ABEV", "EL", "HLT", "ODFL", "STZ", "HES", "DHI", "TAK", "PLTR", "WMB", "MFG", "SU", "PAYX", "TFC", "CRH", "VLO", "KHC", "MSCI", "ET",
             "LNG", "KMB", "STM", "TEL", "DLR", "EW", "AEP", "ROST", "SGEN", "HSY", "COF", "DXCM", "F", "FTNT", "COR", "GWW", "ADM", "TRV", "EXC", "D", "CNC", "LI", "IDXX",
             "NUE", "WDS", "BIDU", "OKE", "DASH", "GIS", "GM", "O", "LVS", "KVUE", "KMI", "IQV", "NU", "EA", "TRP", "DOW", "CM", "BCE", "YUM", "LEN", "BK", "SYY", "AME",
             "MPLX", "DDOG", "AMP", "JCI", "LHX", "ALL", "WCN", "FAST", "SQ", "VRSK", "OTIS", "CTSH", "MFC", "ALC", "SPOT", "PRU", "BBD", "FERG", "CSGP", "BKR", "XEL", "A", "CTVA",
             "HAL", "BIIB", "LYG", "MBLY", "CVE", "IT", "GEHC", "IBKR", "TTD", "URI", "FIS", "PPG", "QSR", "PEG", "CMI", "ORAN", "PUK", "ED", "ROK", "ACGL", "LYB", "GFS", "ARES",
             "NDAQ", "IMO", "KR", "DD", "ON", "VICI", "NEM", "SLF", "MRNA", "GPN", "CDW", "ARGX", "CHT", "MLM", "EXR", "ELP", "IR", "DVN", "CPNG", "NTR", "VMC", "VEEV", "BF-A",
             "CQP", "HPQ", "GOLD", "MDB", "CCEP", "ZS", "PKX", "FANG", "PWR", "RCL", "TTWO", "ANSS", "BCS", "RYAAY", "WEC", "DG", "MPWR", "TU", "FICO", "SBAC", "AWK", "WST", "CAH",
             "SPLK", "EIX", "WBD", "WTW", "DLTR", "VOD", "EFX", "XYL", "WIT", "EC", "GLW", "AEM", "AVB", "GIB", "BNTX", "DB", "CBRE", "FTV", "NET", "RBLX", "KEYS", "HUBS", "ZBH",
             "FNV", "COIN", "WY", "TEF", "HIG", "MTD", "CHD", "DAL", "TCOM", "GRMN", "TLK", "APTV", "RCI", "ICLR", "TSCO", "NWG", "RMD", "TROW", "TW", "SE", "RJF", "FCNCA", "EBAY",
             "STT", "VRSN", "BR", "DTE", "MOH", "FE", "HWM", "ALNY", "PINS", "DFS", "EQR", "LIFW", "ETR", "GMAB", "BRO", "WPM", "HEI-A", "HEI", "IX", "WAB", "INVH", "LYV", "MTB",
             "ULTA", "NVR", "HPE", "AEE", "STE", "ES", "FTS", "BGNE", "SYM", "CTRA", "SNAP", "TS", "SIRI", "MT", "ZTO", "NOK", "CCL", "GPC", "DOV", "BEKE", "ROL", "PPL", "PHG",
             "OWL", "ZM", "CCJ", "IFF", "UMC", "YUMC", "DRI", "EBR", "PHM", "TRGP", "AIU", "FITB", "JBHT", "TDY", "ASX", "MKL", "STLD", "PTC", "IRM", "TECK", "ARE", "EXPE", "BAX",
             "WRB", "VTR", "WBA", "LH", "K", "HRL", "KOF", "HOLX", "RKT", "MKC", "FDS", "TYL", "UWMC", "PBA", "CNP", "VLTO", "EXPD", "DKNG", "EG", "CLX", "VIV", "TSN", "FLT", "PFG",
             "ATO", "AKAM", "COO", "CMS", "J", "JBL", "LPLA", "BAH", "AXON", "BEP", "FSLR", "EQT", "CHKP", "BMRN", "WLK", "BALL", "VRT", "WMG", "ERIC", "BLDR", "NTAP", "RPRX", "SUI",
             "CINF", "DECK", "STX", "KB", "HBAN", "WAT", "BSY", "RS", "MGA", "ALGN", "HUBB", "ENTG", "NTRS", "RIVN", "BG", "CF", "OMC", "AVY", "DGX", "WSO", "RF", "ILMN", "TXT", "DT",
             "SMCI", "AMH", "L", "IEX", "WDC", "SWKS", "FWONA", "EPAM", "ERIE", "SNA", "TME", "BBY", "MRO", "RTO", "LUV", "LII", "LDOS", "ALB", "SHG", "AER", "IOT", "MAA", "MGM",
             "SUZ", "XPEV", "JHX", "ESS", "PKG", "TER", "FOXA", "CE", "LW", "SWK", "ELS", "MANH", "SSNC", "APP", "SQM", "POOL", "GGG", "CAG", "AMCR", "VFS", "CBOE", "MAS", "LOGI",
             "CSL", "CNHI", "NDSN", "NIO", "RPM", "BAM", "DPZ", "AVTR", "GEN", "GDDY", "UAL", "VST", "ROKU", "XP", "TAP", "LNT", "CFG", "BIP", "IHG", "GRAB", "ENPH", "TPL", "NICE",
             "CRBG", "NMR", "AGR", "CPB", "MOS", "LBRDK", "PODD", "WPC", "BEN", "GLPI", "NWSA", "INCY", "FNF", "ACI", "ACM", "LKQ", "HST", "OVV", "SYF", "RYAN", "UDR", "H", "KIM",
             "IP", "AZPN", "LEGN", "GFI", "EVRG", "MORN", "ZI", "OC", "FLEX", "HTHT", "IPG", "OKTA", "EDU", "SJM", "AES", "CG", "KEY", "FMS", "PSTG", "AOS", "RBA", "ARCC", "JKHY",
             "SNN", "TWLO", "LECO", "GL", "GFL", "REG", "U", "ZBRA", "EXAS", "RDY", "WSM", "VTRS", "APA", "TRU", "RVTY", "SAIA", "BCH", "CNA", "NRG", "RNR", "NI", "UTHR", "CHK",
             "TRMB", "NBIX", "WES", "CDAY", "CASY", "TPG", "USFD", "RGA", "PNR", "AEG", "OTEX", "UHAL-B", "PAG", "PATH", "KMX", "UHAL", "YPF", "CCK", "TFX", "XPO", "FIVE", "PAYC",
             "CX", "TFII", "REXR", "BAP", "EME", "LAMR", "CZR", "TEVA", "DKS", "WYNN", "DOX", "LCID", "CRL", "EMN", "COTY", "WRK", "TECH", "PFGC", "FFIV", "CHRW", "EQH", "FND",
             "DINO", "NTNX", "Z", "WPP", "CPT", "WMS", "WTRG", "HII", "UHS", "ALLE", "BLD", "AFG", "TOL", "QRVO", "SBS", "PEAK", "BJ", "BRKR", "QGEN", "CHDN", "HSIC", "DBX", "TTC",
             "CLH", "ITT", "CUBE", "SCI", "BSAC", "BIO", "CHE", "SNX", "NLY", "KEP", "FTI", "ESLT", "NVT", "GGB", "EWBC", "TTEK", "AGCO", "BXP", "DOCU", "DVA", "DUOL", "MEDP", "ETSY",
             "ONON", "SKM", "CLF", "CHWY", "PEN", "MTCH", "JNPR", "MKTX", "MTN", "RGEN", "PARA", "RHI", "TIMB", "AIZ", "MNSO", "BURL", "ALV", "DLB", "ATR", "PCTY", "PSO", "FBIN",
             "KNX", "FUTU", "VIPS", "PNW", "UNM", "ALLY", "KNSL", "MNDY", "LSCC", "RRC", "CW", "WWD", "ORI", "PCOR", "SSL", "SKX", "AAL", "BWA", "BIRK", "GWRE", "RL", "STVN", "CYBR",
             "STN", "FCN", "EGP", "LNW", "LEA", "GLOB", "OHI", "SRPT", "MUSA", "CNM", "COLD", "OLED", "JAZZ", "X", "SEIC", "AR", "GRFS", "PR", "RRX", "TOST", "FRT", "WCC", "BERY",
             "RGLD", "WSC", "GTLB", "YMM", "AFRM", "SWN", "HESM", "EDR", "WEX", "WF", "NOV", "LAD", "TXRH", "JLL", "VOYA", "WBS", "JEF", "PRI", "NYT", "ESTC", "HOOD", "DCI", "ARMK",
             "NFE", "HLI", "AU", "BZ", "TX", "CACI", "TREX", "CTLT", "CCCS", "RMBS", "GNTX", "NNN", "APPF", "MSTR", "IEP", "INFA", "BWXT", "KBR", "ROIV", "OGE", "UI", "RBC", "CSAN",
             "GNRC", "IONS", "CART", "FIX", "TPR", "DSGX", "PAC", "APG", "BBWI", "GPK", "ASR", "ACHC", "INGR", "UFPI", "FHN", "AXTA", "GXO", "FSV", "KVYO", "TPX", "CAR", "ARW",
             "WING", "MAT", "MTDR", "VFC", "EXEL", "KGC", "FMC", "CAE", "WFRD", "ONTO", "DAR", "MUR", "LBTYK", "SSD", "CIEN", "MIDD", "PSN", "NYCB", "CIB", "CIG", "HRB", "COKE",
             "BILL", "MSA", "CHRD", "ICL", "TKO", "GMED", "BRX", "FLR", "QLYS", "WTS", "SOFI", "SWAV", "WFG", "STAG", "EHC", "AMKR", "CIVI", "WH", "KRTX", "STWD", "KT"
]
#Futures
Futures_symbol_list=['ES=F','YM=F','NQ=F','ZB=F','ZN=F','ZF=F','ZT=F','GC=F','HG=F','SI=F','CL=F','NG=F','BZ=F','ZC=F','ZO=F','KE=F','ZR=F','ZM=F','ZL=F','ZS=F','GF=F','HE=F','HO=F','LE=F','CC=F','KC=F','CT=F','SB=F']
#Index
Index_symbol_list=['^N225','^KS11','^BSESN','^MXX','^IXIC','^GSPTSE','^DJI','^FCHI','^GDAXI','^FTSE','^IBEX','^N100','^GSPC','^RUT','^NYA','^STI']
#ETF
ETF_symbol_list=[
    "SPY", "IVV", "VOO", "VTI", "QQQ", "VEA", "VTV", "IEFA", "BND", "VUG", "AGG", "IWF", "VWO", "IEMG", "IJH", "VIG", "IJR", "VXUS", "GLD", "VO", "VGT", "BNDX", "IWM", "XLK", "IWD",
 "VYM", "SCHD", "EFA", "ITOT", "VB", "TLT", "BIL", "XLE", "RSP", "VCIT", "XLV", "VCSH", "MUB", "VEU", "IVW", "BSV", "SCHX", "QUAL", "SCHF", "IXUS", "VNQ", "JEPI", "IWB", "LQD", "VT",
 "XLF", "DIA", "VV", "IEF", "VTEB", "SHY", "USMV", "IWR", "MBB", "IAU", "IVE", "VBR", "DGRO", "JPST", "GOVT", "SCHB", "VGSH", "SHV", "IUSB", "DFAC", "IGSB", "SPLG", "TIP", "SCHG", "SDY", 
 "SPYG", "USFR", "VGIT", "SGOV", "VGK", "MDY", "DVY", "ACWI", "EFV", "VHT", "SPYV", "VMBS", "EEM", "XLP", "XLY", "SPDW", "COWZ", "TQQQ", "BIV", "VXF", "VOE", "QQQM", "MGK", "VONG", "XLI"
 ]

Special_symbol_list=[
             '^GSPC','^DJI','^RUT','^NYA','^STI','^IXIC','^FTSE','^FCHI','^GDAXI','^IBEX','^N100','^N225','^KS11','^BSESN','^MXX','^GSPTSE',
             'ES=F','YM=F','NQ=F','ZB=F','ZN=F','ZF=F','ZT=F','GC=F','HG=F','SI=F','CL=F','NG=F','BZ=F','ZC=F','ZO=F','KE=F','ZM=F','ZL=F','ZS=F','GF=F','HE=F','HO=F','LE=F','CC=F','KC=F','CT=F','SB=F',
             "SPY", "IVV", "VOO", "VTI", "QQQ", "VEA", "VTV", "IEFA", "BND", "VUG", "AGG", "IWF", "VWO", "IEMG", "IJH", "VIG", "IJR", "VXUS", "GLD", "VO", "VGT", "BNDX", "IWM", "XLK", "IWD",             
             ]

Test_symbol_list=[
            #  '^GSPC','^DJI','^RUT','^NYA','^STI','^IXIC',
                # "SPYG", "USFR", "VGIT",
'^STI','^IXIC',
             ]

# stock_code_list = None
# stock_code_list = ["015760", "034020", "088980", "055550", "032640", "323410", "000270", "105560", "047040", "003490", "034220", "005940", "138930", "030200", "086790",]
# symbol_list = ["005930.KS", "015760.KS", "034020.KS", "088980.KS", "055550.KS", "032640.KS", "323410.KS", "000270.KS", "105560.KS", "047040.KS", "003490.KS", "034220.KS", "005940.KS", "138930.KS", "030200.KS", "086790.KS",]
KOSPI_symbol_list = [
"005935.KS", "005930.KS", "373220.KS", "000660.KS", "207940.KS", "005387.KS", "005389.KS", "005380.KS", "005385.KS", "068270.KS", "000270.KS", "035420.KS", "051915.KS", 
"005490.KS", "051910.KS", "006405.KS", "006400.KS", "035720.KS", "003670.KS", "028260.KS", "02826K.KS", "105560.KS", "012330.KS", "055550.KS", "086520.KQ", "066575.KS", 
"066570.KS", "011200.KS", "323410.KS", "018260.KS", "03473K.KS", "096770.KS", "096775.KS", "086790.KS", "015760.KS", "003555.KS", "003550.KS", "032830.KS", "138040.KS", 
"009155.KS", "009150.KS", "000810.KS", "017670.KS", "329180.KS", "000815.KS", "352820.KS", "033780.KS", "010130.KS", "034020.KS", "316140.KS", "259960.KS", "047050.KS", 
"024110.KS", "034730.KS", "003495.KS", "003490.KS", "030200.KS", "090430.KS", "090435.KS", "009540.KS", "010950.KS", "010955.KS", "326030.KS", "377300.KS", "012450.KS", 
"042660.KS", "402340.KS", "086280.KS", "009830.KS", "454910.KS", "009835.KS", "066970.KQ", "247540.KQ", "010140.KS", "028300.KQ", "361610.KS", "001570.KS", "011170.KS", 
"042700.KS", "161390.KS", "18064K.KS", "047810.KS", "011070.KS", "180640.KS", "302440.KS", "051905.KS", "307950.KS", "051900.KS", "028050.KS", "000105.KS", "034220.KS", 
"005830.KS", "251270.KS", "000100.KS", "267250.KS", "241560.KS", "271560.KS", "005070.KS", "068760.KQ", "128940.KS", "036570.KS", "00680K.KS", "097950.KS", "088980.KS", 
"004020.KS", "032640.KS", "097955.KS", "006800.KS", "006805.KS", "005945.KS", "021240.KS", "000720.KS", "000725.KS", "078930.KS", "078935.KS", "018880.KS", "272210.KS", 
"029780.KS", "267260.KS", "035900.KQ", "016360.KS", "035250.KS", "010620.KS", "058470.KQ", "011785.KS", "011780.KS", "383220.KS", "071050.KS", "071055.KS", "064350.KS", 
"011790.KS", "005940.KS", "112610.KS", "003410.KS", "079550.KS", "008930.KS", "000120.KS", "001045.KS", "052690.KS", "00104K.KS", "001040.KS", "282330.KS", "007070.KS", 
"001450.KS", "081660.KS", "039490.KS", "014680.KS", "008775.KS", "138930.KS", "008770.KS", "006260.KS", "293490.KQ", "004370.KS", "010060.KS", "005290.KQ", "328130.KQ", 
"002790.KS", "002795.KS", "000990.KS", "041510.KQ", "000880.KS", "00088K.KS", "175330.KS", "000885.KS", "010120.KS", "036460.KS", "009420.KS", "023530.KS", "039030.KQ", 
"088350.KS", "028670.KS", "111770.KS", "012750.KS", "030000.KS", "139480.KS", "00499K.KS", "004990.KS", "007660.KS", "204320.KS", "020150.KS", "047040.KS", "298040.KS", 
"026960.KS", "095340.KQ", "000250.KQ", "065350.KQ", "298050.KS", "002380.KS", "240810.KQ", "086900.KQ", "036930.KQ", "003230.KS", "011210.KS", "37550K.KS", "004170.KS", 
"073240.KS", "375500.KS", "042670.KS", "005850.KS", "003620.KS", "051600.KS", "348370.KQ", "006110.KS", "336260.KS", "000080.KS", "000240.KS", "33626K.KS", "000087.KS", 
"185750.KS", "017800.KS", "248070.KS", "298020.KS", "137310.KS", "35320K.KS", "002710.KS", "353200.KS", "007310.KS", "005420.KS", "253450.KQ", "000150.KS", "139130.KS", 
"035760.KQ", "000155.KS", "018670.KS", "005305.KS", "108320.KS", "005300.KS", "006280.KS", "069620.KS", "004000.KS", "067310.KQ", "196170.KQ", "336370.KS", "064760.KQ", 
"006360.KS", "192820.KS", "078600.KQ", "161890.KS", "38380K.KS", "121600.KQ", "365340.KQ", "383800.KS", "120115.KS", "120110.KS", "001440.KS", "004800.KS", "009520.KQ", 
"000210.KS", "140860.KQ", "237690.KQ", "440110.KQ", "100090.KS", "214450.KQ", "280360.KS", "285130.KS", "000215.KS", "006040.KS", "145720.KS", "101000.KQ", "298380.KQ", 
"003535.KS", "034230.KQ", "33637K.KS", "025900.KQ", "096530.KQ", "067160.KQ", "069960.KS", "28513K.KS", "395400.KS", "001740.KS", "067630.KQ", "166090.KQ", "089590.KS", 
"003690.KS", "213420.KQ", "093370.KS", "085660.KQ", "103140.KS", "000157.KS", "224060.KQ", "278280.KQ", "003280.KS", "195870.KS", "056190.KQ", "014820.KS", "014825.KS", 
"294870.KS", "089860.KS", "001120.KS", "084370.KQ", "112040.KQ", "009900.KS", "271940.KS", "048410.KQ", "003545.KS", "003547.KS", "012510.KS", "267270.KS", "192650.KS", 
"003540.KS", "039130.KS", "222800.KQ", "009970.KS", "036540.KQ", "001067.KS", "064550.KQ", "003030.KS", "000670.KS", "001800.KS", "006650.KS", "020560.KS", "001060.KS", 
"192080.KS", "005880.KS", "086450.KQ", "003530.KS", "074600.KQ", "214320.KS", "101490.KQ", "300720.KS", "348950.KS", "178920.KS", "009240.KS", "003090.KS", "001065.KS", 
"099190.KQ", "002840.KS", "058610.KQ", "287410.KQ", "114090.KS", "010780.KS", "004490.KS", "060280.KQ", "105630.KS", "001430.KS", "122870.KQ", "003850.KS", "069260.KS", 
"016380.KS", "365550.KS", "002355.KS", "456040.KS", "023590.KS", "137400.KQ", "181710.KS", "002350.KS", "006125.KS", "025540.KS", "000400.KS", "006120.KS", "330590.KS", 
"032350.KS", "007340.KS", "950160.KQ", "215200.KQ", "008730.KS", "005257.KS", "005250.KS", "352480.KQ", "015750.KQ", "145020.KQ", "036830.KQ", "210980.KS", "001680.KS", 
"082740.KS", "001685.KS", "003380.KQ", "009450.KS", "299900.KQ", "087010.KQ", "025320.KQ", "082640.KS", "079160.KS", "399720.KQ", "268280.KS", "272450.KS", "033240.KS", 
"179900.KQ", "161000.KS", "286940.KS", "019175.KS", "019170.KS", "344820.KS", "290650.KQ", "054780.KQ", "025980.KQ", "281820.KS", "229640.KS", "085620.KS", "091120.KQ", 
"049770.KS", "090460.KS", "053800.KQ", "080220.KQ", "094360.KQ", "031430.KS", "091810.KS", "000640.KS", "141080.KQ", "084850.KQ", "450140.KS", "018290.KQ", "039200.KQ", 
"007700.KS", "030520.KQ", "293940.KS", "170900.KS", "041190.KQ", "046890.KQ", "009415.KS", "237880.KQ", "950210.KS", "064960.KS", "079370.KQ", "017960.KS", "030190.KS", 
"460860.KS", "182360.KQ", "082270.KQ", "030610.KS", "031980.KQ", "095660.KQ", "45014K.KS", "005440.KS", "183300.KQ", "001530.KS", "403550.KS", "003475.KS", "032500.KQ", 
"189300.KQ", "078340.KQ", "381970.KS", "215000.KQ", "034120.KS", "000070.KS", "068240.KQ", "000075.KS", "003000.KS", "032190.KQ", "200130.KQ", "003470.KS", "093320.KQ", 
"003240.KS", "002240.KS", "005610.KS", "299030.KQ", "348340.KQ", "007390.KQ", "097520.KS", "241590.KS", "192400.KS", "095500.KQ", "001720.KS", "900140.KS", "001725.KS", 
"069080.KQ", "263750.KQ", "057050.KS", "144510.KQ", "007815.KS", "005180.KS", "131290.KQ", "394280.KQ", "200880.KS", "284740.KS", "060250.KQ", "00781K.KS", "007810.KS", 
"108860.KQ", "043150.KQ", "000370.KS", "052020.KQ", "003920.KS", "033640.KQ", "018250.KS", "049070.KQ", "357120.KS", "001470.KS", "003925.KS", "249420.KS", "377740.KS", 
"054450.KQ", "145995.KS", "068400.KS", "145990.KS", "034310.KS", "056080.KQ", "001390.KS", "020000.KS", "094170.KQ", "084690.KS", "118990.KQ", "091700.KQ", "319400.KQ", 
"010690.KS", "218410.KQ", "036810.KQ", "103590.KS", "014830.KS", "039440.KQ", "206640.KQ", "101160.KQ", "203690.KQ", "241710.KQ", "058650.KS", "084695.KS", "006730.KQ", 
"011930.KS", "089030.KQ", "095700.KQ", "050890.KQ", "047920.KQ", "151860.KQ", "000490.KS", "314930.KQ", "215600.KQ", "183190.KS", "263720.KQ", "003520.KS", "448730.KS", 
"232140.KQ", "004430.KS", "053610.KQ", "001820.KS", "041830.KQ", "082920.KQ", "033270.KS", "453340.KS", "108675.KS", "007575.KS", "123890.KS", "017810.KS", "194480.KQ", 
"110990.KQ", "108670.KS", "102710.KQ", "042600.KQ", "340570.KQ", "033100.KQ", "071970.KS", "304100.KQ", "001940.KS", "025860.KS", "104830.KQ", "298690.KS", "093050.KS", 
"005690.KS", "023160.KQ", "084680.KS", "064290.KQ", "297890.KQ", "094480.KQ", "088260.KS", "005090.KS", "108490.KQ", "095610.KQ", "253590.KQ", "225570.KQ", "002310.KS", 
"089010.KQ", "131970.KQ", "027410.KS", "451800.KS", "044490.KQ", "178320.KQ", "306200.KS", "092040.KQ", "011810.KS", "243070.KQ", "005810.KS", "100790.KQ", "083310.KQ", 
"330860.KQ", "000430.KS", "013030.KQ", "034950.KQ", "014620.KQ", "317330.KQ", "012630.KS", "092220.KS", "025770.KQ", "017940.KS", "033500.KQ", "039840.KQ", "027360.KQ", 
"054950.KQ", "126560.KS", "090360.KQ", "060150.KQ", "003300.KS", "136480.KQ", "089980.KQ", "003570.KS", "029960.KQ", "002320.KS", "119610.KQ", "035600.KQ", "102940.KQ", 
"007690.KS", "038500.KQ", "140410.KQ", "089970.KQ", "058820.KQ", "077360.KQ", "060980.KS", "322510.KQ", "001200.KS", "003220.KS", "322000.KS", "265520.KQ", "097230.KS", 
"004690.KS", "267980.KQ", "058430.KS", "200670.KQ", "001880.KS", "061970.KQ", "008490.KS", "036030.KQ", "139050.KQ", "060720.KQ", "071320.KS", "064260.KQ", "000680.KS", 
"298000.KS", "012030.KS", "002030.KS", "047310.KQ", "008355.KS", "013890.KS", "950140.KQ", "008350.KS", "307750.KQ", "001790.KS", "950130.KQ", "079810.KQ", 
"102460.KS", "084650.KQ", "005950.KS", "024720.KS", "002960.KS", "051500.KQ", "006390.KS", "072710.KS", "042000.KQ", "000020.KS", "206400.KQ", "001780.KS", "078160.KQ", 
"095720.KS", "382840.KQ", "084110.KQ", "243840.KQ", "086390.KQ", "009290.KS", "389260.KQ", "356860.KQ", "013120.KQ", "001630.KS", "138490.KS", "001465.KS", "080160.KQ", 
"029530.KS", "001460.KS", "059090.KQ", "048530.KQ", "272290.KQ", "004560.KS", "007460.KS", "108230.KQ", "126700.KQ", "036620.KQ", "297090.KQ", "026890.KS", "077970.KS", 
"005010.KS", "031330.KQ", "000545.KS", "003800.KQ", "037270.KS", "053210.KS", "036420.KS", "016800.KS", "001500.KS", "035890.KQ", "452260.KS", "100120.KQ", "018470.KS", 
"005500.KS", "037560.KS", "126720.KS", "045100.KQ", "272550.KS", "002900.KS", "117580.KS", "075580.KS", "053030.KQ", "010100.KS", "418420.KQ", "001525.KS", "051360.KQ", 
"122900.KS", "200710.KQ", "005390.KS", "007570.KS", "099320.KQ", "206650.KQ", "002150.KS", "214370.KQ", "001510.KS", "041960.KQ", "033160.KQ", "322310.KQ", "001230.KS", 
"115390.KS", "000480.KS", "011500.KS", "003960.KS", "063080.KQ", "089600.KQ", "049950.KQ", "051370.KQ", "093640.KQ", "092070.KQ", "040300.KQ", "114190.KQ", "006220.KS", 
"211270.KQ", "271980.KS", "213500.KS", "104700.KS", "048870.KQ", "462520.KS", "054210.KQ", "036530.KS", "060370.KQ", "030530.KQ", "315640.KQ", "123860.KQ", "083650.KQ", 
"171090.KQ", "143160.KQ", "293480.KS", "101390.KQ", "370090.KQ", "130660.KS", "096760.KS", "042420.KQ", "005160.KQ", "032620.KQ", "036200.KQ", "035150.KS", "194370.KS", 
"041020.KQ", "034830.KS", "074610.KS", "104460.KQ", "045390.KQ", "001515.KS", "388720.KQ", "078350.KQ", "017890.KQ", "037710.KS", "079940.KQ", "278650.KQ", "017390.KS", 
"038290.KQ", "034810.KQ", "023410.KQ", "206560.KQ", "083450.KQ", "321820.KQ", "029460.KS", "004360.KS", "016580.KS", "018310.KQ", "004365.KS", "348150.KQ", "030210.KS", 
"052400.KQ", "002025.KS", "071840.KS", "900290.KQ", "090710.KQ", "002020.KS", "092730.KQ", "267850.KS", "095570.KS", "060590.KQ", "084010.KS", "047560.KQ", "214430.KQ", 
"131370.KQ", "004380.KS", "036630.KQ", "051980.KQ", "000050.KS", "086980.KQ", "068290.KS", "044450.KS", "004255.KS", "006840.KS", "011760.KS", "122640.KQ", "020120.KS", 
"125210.KQ", "001250.KS", "088390.KQ", "011280.KS", "078070.KQ", "073490.KQ", "00806K.KS", "009160.KS", "004250.KS", "008060.KS", "226320.KS", "003075.KS", "003070.KS", 
"027970.KS", "002810.KS", "230240.KQ", "005725.KS", "102260.KS", "004710.KS", "149950.KQ", "081000.KS", "053300.KQ", "321550.KQ", "011000.KS", "094860.KQ", "294570.KQ", 
"078150.KQ", "004980.KS", "005720.KS", "143240.KQ", "004985.KS", "019680.KS", "102120.KQ", "063570.KQ", "179290.KQ", "073570.KQ", "065510.KQ", "001130.KS", "086890.KQ", 
"063170.KQ", "083790.KQ", "067000.KQ", "060310.KQ", "377190.KS", "214680.KQ", "011330.KS", "019685.KS", "091580.KQ", "053690.KS", "098460.KQ", "036800.KQ", "204620.KQ", 
"025000.KS", "046440.KQ", "007160.KS", "251630.KQ", "083930.KQ", "339770.KS", "067280.KQ", "078020.KQ", "035810.KQ", "092230.KS", "043610.KQ", "036890.KQ", "080420.KQ", 
"460850.KS", "032820.KQ", "291230.KQ", "078520.KS", "009580.KS", "043370.KQ", "035080.KQ", "234080.KS", "010580.KS", "382800.KQ", "016590.KS", "000540.KS", "003060.KS", 
"092200.KS", "084730.KQ", "084990.KQ", "001527.KS", "950170.KQ", "089890.KQ", "036560.KQ", "192440.KQ", "317400.KS", "000145.KS", "003580.KS", "223310.KQ", "013990.KQ", 
"036010.KQ", "005680.KS", "002390.KS", "000140.KS", "005860.KQ", "001520.KS", "115960.KQ", "072770.KQ", "019210.KQ", "139990.KS", "023760.KQ", "214390.KS", "027710.KQ", 
"003200.KS", "048550.KQ", "024850.KQ", "105840.KS", "092870.KQ", "092790.KS", "107590.KS", "122990.KQ", "092190.KQ", "293780.KQ", "300080.KQ", "015890.KS", "002100.KS", 
"119860.KQ", "016450.KS", "002995.KS", "225530.KQ", "100130.KQ", "002990.KS", "090350.KS", "377030.KQ", "001360.KS", "011040.KQ", "002620.KS", "015590.KS", "004700.KS", 
"024060.KQ", "030960.KQ", "028100.KS", "090355.KS", "045970.KQ", "104480.KQ", "009680.KS", "014580.KS", "334890.KS", "094820.KQ", "038540.KQ", "083420.KS", "196450.KQ", 
"080580.KQ", "060570.KQ", "042370.KQ", "015860.KS", "362320.KQ", "290550.KQ", "314130.KQ", "220260.KQ", "001020.KS", "025950.KQ", "039570.KS", "111380.KS", "100250.KS", 
"008700.KS", "007860.KS", "024900.KS", "214180.KQ", "068790.KQ", "008260.KS", "238090.KQ", "023910.KQ", "010820.KS", "365270.KQ", "092130.KQ", "035510.KS", "950190.KQ", 
"230360.KQ", "044340.KQ", "011700.KS", "376190.KQ", "085670.KQ", "123040.KQ", "033530.KS", "006910.KQ", "014280.KS", "355390.KQ", "282690.KS", "046210.KQ", "013310.KQ", 
"121440.KQ", "161580.KQ", "006620.KQ", "006060.KS", "002785.KS", "003160.KS", "196490.KQ", "002787.KS", "067080.KQ", "419530.KQ", "004090.KS", "003720.KS"
]

# # freq_int = '1wk'
# start_date = '2001-01-01'
# start_date_short = '2011-01-01'
# target_col = 'Close'

# n_predict = 1
# n_last_sequence = 100
# forward= 0
# seq_len= 2
# interval='1wk'
# freq='W'

# def compute_average_metrics_by_Model(dict):
#     dict_by_Model = {}
#     for key, df in dict.items():
#         if isinstance(df, pd.DataFrame):
#             average_df = df.groupby('Model').mean()
#             dict_by_Model[key] = average_df
#     return dict_by_Model

# def compute_average_test_dates_by_Model(dict):
#     # Concatenate all DataFrames
#     all_df = pd.concat([dict])
#     # Select numeric columns and compute average
#     all_df_numeric = all_df.select_dtypes(include=['number'])
#     average_df_by_Model = all_df_numeric.groupby('Model_ID').mean()
#     # Reset index
#     average_df_by_Model = average_df_by_Model.reset_index()
#     # Get the first 'Model' for each 'Model_ID'
#     model_df = all_df.groupby('Model_ID')['Model'].first().reset_index()
#     # Merge with 'Model' columns
#     average_df_by_Model = pd.merge(average_df_by_Model, model_df, on='Model_ID', how='left')
#     # Remove 'test_dates' column
#     del average_df_by_Model['test_dates']
#     # Make 'Model' column the first column
#     cols = ['Model']  + [col for col in average_df_by_Model if col != 'Model']
#     average_df_by_Model = average_df_by_Model[cols]
#     return average_df_by_Model

# def compute_average_test_dates_by_Symbol(dict):
#     dict_by_Symbol = {}
#     for key, df in dict.items():
#         if isinstance(df, pd.DataFrame):
#             # Select numeric columns and compute average
#             df_numeric = df.select_dtypes(include=['number'])
#             average_df = df_numeric.groupby('Symbol_ID').mean()
#             # Reset index
#             average_df = average_df.reset_index()
#             # Get the first 'Symbol' for each 'Symbol_ID'
#             symbol_df = df.groupby('Symbol_ID')['Symbol'].first().reset_index()
#             # Merge with 'Symbol' columns
#             average_df = pd.merge(average_df, symbol_df, on='Symbol_ID', how='left')
#             # Remove 'test_dates' column
#             del average_df['test_dates']
#             # Make 'Symbol' column the first column
#             cols = ['Symbol']  + [col for col in average_df if col != 'Symbol']
#             average_df = average_df[cols]
#             dict_by_Symbol[key] = average_df
#     return dict_by_Symbol

# def compute_average_metrics_by_Model(dict):
#     # Concatenate all DataFrames
#     all_df = pd.concat(dict.values())
#     # Select numeric columns and compute average
#     all_df_numeric = all_df.select_dtypes(include=['number'])
#     average_df_by_Model = all_df_numeric.groupby(['Model_ID', 'test_dates']).mean()
#     # Reset index
#     average_df_by_Model = average_df_by_Model.reset_index()
#     # Get the first 'Model' for each 'Model_ID'
#     model_df = all_df.groupby(['Model_ID', 'test_dates'])['Model'].first().reset_index()
#     # Merge with 'Model' columns
#     average_df_by_Model = pd.merge(average_df_by_Model, model_df, on=['Model_ID', 'test_dates'], how='left')
#     # Remove 'Model_ID' column
#     del average_df_by_Model['Symbol_ID']
#     # Make 'Model' column the first column
#     cols = ['Model', 'test_dates']  + [col for col in average_df_by_Model if col not in ['Model', 'test_dates']]
#     average_df_by_Model = average_df_by_Model[cols]
#     return average_df_by_Model

# def compute_average_metrics_by_Symbol(dict):
#     dict_by_Symbol = {}
#     for key, df in dict.items():
#         if isinstance(df, pd.DataFrame):
#             # Select numeric columns and compute average
#             df_numeric = df.select_dtypes(include=['number'])
#             average_df = df_numeric.groupby(['Symbol_ID', 'test_dates']).mean()
#             # Reset index
#             average_df = average_df.reset_index()
#             # Get the first 'Symbol' for each 'Symbol_ID'
#             symbol_df = df.groupby(['Symbol_ID', 'test_dates'])['Symbol'].first().reset_index()
#             # Merge with 'Symbol' columns
#             average_df = pd.merge(average_df, symbol_df, on=['Symbol_ID', 'test_dates'], how='left')
#             # Remove 'Symbol_ID' column
#             del average_df['Model_ID']
#             # Make 'Symbol' column the first column
#             cols = ['Symbol', 'test_dates']  + [col for col in average_df if col not in ['Symbol', 'test_dates']]
#             average_df = average_df[cols]
#             dict_by_Symbol[key] = average_df
#     return dict_by_Symbol

# def compute_average_by_level(dict, level='Model_ID'):
#     # Initialize an empty dictionary to store the average DataFrames
#     average_dict = {}
    
#     # Iterate over the items in the input dictionary
#     for key, df in dict.items():
            
#         if not isinstance(value, pd.DataFrame):
#             # Assuming the value can be converted to a DataFrame
#             average_metrics_dict_by_Model[key] = pd.DataFrame(value)

#         # if isinstance(df, pd.DataFrame):
#             # Select numeric columns and compute average
#             df_numeric = df.select_dtypes(include=['number'])
#             average_df = df_numeric.groupby(level).mean()
            
#             # Reset index
#             average_df = average_df.reset_index()
            
#             # Get the first 'level' for each 'level'
#             level_df = df.groupby(level)[level].first().reset_index()
            
#             # Merge with 'level' columns
#             average_df = pd.merge(average_df, level_df, on=level, how='left')
            
#             # Remove 'level' column
#             del average_df[level]
            
#             # Make 'level' column the first column
#             cols = [level]  + [col for col in average_df if col != level]
#             average_df = average_df[cols]
            
#             # Store the average DataFrame in the output dictionary
#             average_dict[key] = average_df
    
#     return average_dict

def compute_average_test_dates(df, group_by):

    if group_by not in ['Model', 'Symbol']:
        raise ValueError("The 'group_by' parameter should be either 'Model' or 'Symbol'.")

    # Select numeric columns and compute average
    df_numeric = df.select_dtypes(include=['number'])
    average_df = df_numeric.groupby(f'{group_by}_ID').mean()

    # Reset index
    average_df = average_df.reset_index()

    # Get the first 'group_by' for each 'group_by_ID'
    group_df = df.groupby(f'{group_by}_ID')[group_by].first().reset_index()

    # Merge with 'group_by' columns
    average_df = pd.merge(average_df, group_df, on=f'{group_by}_ID', how='left')

    # Remove 'test_dates' column
    if 'test_dates' in average_df.columns:
        del average_df['test_dates']

    # Make 'group_by' column the first column
    cols = [group_by]  + [col for col in average_df if col != group_by]
    average_df = average_df[cols]

    return average_df

def compute_average_metrics(df, group_by):

    if group_by not in ['Model', 'Symbol']:
        raise ValueError("The 'group_by' parameter should be either 'Model' or 'Symbol'.")

    # Select numeric columns and compute average
    df_numeric = df.select_dtypes(include=['number'])
    average_df = df_numeric.groupby([f'{group_by}_ID', 'test_dates']).mean()

    # Reset index
    average_df = average_df.reset_index()

    # Get the first 'group_by' for each 'group_by_ID'
    group_df = df.groupby([f'{group_by}_ID', 'test_dates'])[group_by].first().reset_index()

    # Merge with 'group_by' columns
    average_df = pd.merge(average_df, group_df, on=[f'{group_by}_ID', 'test_dates'], how='left')

    # Remove 'other_group_ID' column
    if group_by == 'Model':
        if 'Symbol_ID' in average_df.columns:
            del average_df['Symbol_ID']
    elif group_by == 'Symbol':
        if 'Model_ID' in average_df.columns:
            del average_df['Model_ID']

    # Make 'group_by' column the first column
    cols = [group_by, 'test_dates']  + [col for col in average_df if col not in [group_by, 'test_dates']]
    average_df = average_df[cols]

    return average_df


# def get_parameters(select_interval):
#     if  select_interval == '1wk':
#         n_predict = 1
#         n_last_sequence = 10
#         forward= 0
#         seq_len= 2
#         interval='1wk'
#         freq='W'
#         period='10y'
#         period_short='5y'
#         # start_date = '2001-01-01'
#         # start_date_short = '2021-01-01'
#         target_col = 'Close'
#     elif select_interval == '1mo':
#         n_predict = 1
#         n_last_sequence = 10
#         forward= 0
#         seq_len= 2
#         interval='1mo'
#         freq='M'
#         period='max'
#         period_short='10y'
#         target_col = 'Close'
#     elif select_interval == '1d':
#         n_predict = 1
#         n_last_sequence = 10
#         forward= 0
#         seq_len= 2
#         interval='1d'
#         freq='D'
#         period='5y'
#         period_short='2y'
#         target_col = 'Close'
#     return n_predict, n_last_sequence, forward, seq_len, interval, freq, period, period_short, target_col

def get_parameters(select_interval):
    if  select_interval == '1wk':
        n_predict = 1
        n_last_sequence = 10
        n_last_sequence_long = 30
        forward= 0
        seq_len= 2
        interval='1wk'
        freq='W'
        # period='10y'
        # period_short='5y'
        start_date = '2001-01-01'
        target_col = 'Return'
    elif select_interval == '1mo':
        n_predict = 1
        n_last_sequence = 10
        n_last_sequence_long = 20
        forward= 0
        seq_len= 2
        interval='1mo'
        freq='M'
        start_date = '2001-01-01'
        # period_short='10y'
        target_col = 'Return'
    elif select_interval == '1d':
        n_predict = 1
        n_last_sequence = 5
        n_last_sequence_long = 2
        forward= 0
        seq_len= 5
        interval='1d'
        freq='D'
        start_date = '2001-01-01'
        target_col = 'Return'
    return n_predict, n_last_sequence, n_last_sequence_long, forward, seq_len, interval, freq, start_date, target_col

def get_symbol_list(select_list):
    if select_list == 'World':
        symbol_list = World_list
    elif select_list == 'US':
        symbol_list = US_list
    elif select_list == 'DOW':
        symbol_list = DOW_symbol_list
    elif select_list == 'KOSPI':
        symbol_list = KOSPI_symbol_list
    elif select_list == 'Special':
        symbol_list = Special_symbol_list
    elif select_list == 'ETF':
        symbol_list = ETF_list
    elif select_list == 'Futures':
        symbol_list = Futures_symbol_list
    elif select_list == 'Index':
        symbol_list = Index_symbol_list
    elif select_list == 'Test':
        symbol_list = Test_symbol_list
    return symbol_list

def generate_plots(get_plots=0, target_only=None):
    # all plots for only target
    if get_plots == 1 and target_only==True:
        stock_price_plots_use = False
        plots_use = True
        use_target_col = True
    
    # stock price plots only
    elif get_plots == 2 and target_only==True:
        stock_price_plots_use = True
        plots_use = False
        use_target_col = True        
       
    # prediction plot for only target
    elif get_plots == 3 and target_only==True:
        stock_price_plots_use = False
        plots_use = True
        use_target_col = True
        
    # prediction plots for all features    
    elif get_plots == 4 and target_only==False:
        stock_price_plots_use = False
        plots_use = True
        use_target_col = False
    
    # all plots with all features    
    elif get_plots == 5 and target_only==False:
        stock_price_plots_use = True
        plots_use = True
        use_target_col = False
    
    # no plots
    else:
        stock_price_plots_use = False
        plots_use = False
        use_target_col = True

    return stock_price_plots_use, plots_use, use_target_col

get_plots=1
target_only=True
interval='1d'
target_col='Return'
select_lists='Test'
train_size=0.8
united_transform=False
indicator_ver="calculate_indicators_ver4"
# test_size = 0.8
# test_dates = [0, 1, 2, 3, 4, 5, 6, 8, 9]

symbol_list  = get_symbol_list(select_lists)
stock_price_plots_use, plots_use, use_target_col = generate_plots(get_plots=get_plots, target_only=target_only)
n_predict, n_last_sequence, n_last_sequence_long, forward, seq_len, interval, freq, start_date, target_col = get_parameters(select_interval=interval)
    
# load_directory 
load_directory = "/home/predict_price/stock_price/save_model/daily/ver4/regular_large/ver4_target_regular_split_0.8t+World_list_start_2001/model"
save_directory = "/home/predict_price/stock_price/model_load/regular/ver4/ver4_test_3"
# save_directory = f"/home/predict_price/stock_price/Loaded_Model/test/test_{symbol_list}"
# output_file_path = f"/home/predict_price/stock_price/Loaded_Model/test/test_{symbol_list}/test_{symbol_list}"
output_file_path = "/home/predict_price/stock_price/model_load/regular/ver4/ver4_test_3/ver4_test_3"

all_results_metrics_list = []
all_future_predictions_list = []
all_future_predictions_org_list = []

date_range = range(2)

# for i in date_range:
#     test_dates = [i]
test_dates = range(5)

all_results_metrics_dict, all_future_predictions_dict, all_future_predictions_org_dict = loaded_model(
    symbol_list=symbol_list, start_date=start_date, interval=interval, freq=freq, target_col=target_col, 
    save_directory=save_directory, load_directory=load_directory, metrics_use=True, plots_use=plots_use, stock_price_plots_use=stock_price_plots_use, 
    use_target_col=use_target_col, future_predictions=None, n_predict=n_predict, n_last_sequence=n_last_sequence, test_dates=test_dates, date_range=date_range,
    n_last_sequence_long=n_last_sequence_long, seq_len=seq_len, forward=forward, train_size=train_size, united_transform=united_transform, indicator_ver=indicator_ver)
    
# Append dataframes to lists
all_results_metrics_list.extend(list(all_results_metrics_dict.values()))
all_future_predictions_list.extend(list(all_future_predictions_dict.values()))
all_future_predictions_org_list.extend(list(all_future_predictions_org_dict.values()))

# After the loop, concatenate lists of dataframes
all_results_metrics_df = pd.concat(all_results_metrics_list, ignore_index=True)
all_future_predictions_df = pd.concat(all_future_predictions_list, ignore_index=True)
all_future_predictions_org_df = pd.concat(all_future_predictions_org_list, ignore_index=True)

# Save the DataFrames to CSV files
all_results_metrics_df.to_csv(f'{output_file_path}_all_results_metrics.csv', index=True)
all_future_predictions_df.to_csv(f'{output_file_path}_all_future_predictions.csv', index=True)
all_future_predictions_org_df.to_csv(f'{output_file_path}_all_future_predictions_org.csv', index=True)

all_future_df = select_best_model_for_future(all_future_predictions_df)
all_future_org_df = select_best_model_for_future(all_future_predictions_org_df)

all_predict_df = select_best_model_for_predict(all_future_predictions_df)
all_predict_org_df = select_best_model_for_predict(all_future_predictions_org_df)

# Save the DataFrames to CSV files
all_future_df.to_csv(f'{output_file_path}_all_future.csv', index=True)
all_future_org_df.to_csv(f'{output_file_path}_all_future_org.csv', index=True)

all_predict_df.to_csv(f'{output_file_path}_all_predict.csv', index=True)
all_predict_org_df.to_csv(f'{output_file_path}_all_predict_org.csv', index=True)

    # # Save short_future_predictions
    # pd.concat(short_all_future_predictions_dict.values()).to_csv(f'{output_file_path}_short_all_future_predictions.csv', index=True)

    # Save the overall_future_metrics DataFrame to a CSV file
    # pd.concat(all_future_metrics_dict.values()).to_csv(f'{output_file_path}_all_future_metrics.csv', index=True)

# average_metrics_by_Model = compute_average_metrics(all_results_metrics_df, 'Model')

# # Save average_results_metrics
# average_metrics_by_Model.to_csv(f'{output_file_path}_average_metrics_by_model.csv', index=True)

# average_metrics_by_Symbol = compute_average_metrics(all_results_metrics_df, 'Symbol')

# # Save average_results_metrics
# average_metrics_by_Symbol.to_csv(f'{output_file_path}_average_metrics_by_symbol.csv', index=True)

average_future_predictions_by_Model = compute_average_metrics(all_future_predictions_df, 'Model')

average_future_predictions_org_by_Model = compute_average_metrics(all_future_predictions_org_df, 'Model')

# Save average_future_predictions
average_future_predictions_by_Model.to_csv(f'{output_file_path}_average_future_predictions_by_model.csv', index=True)
average_future_predictions_org_by_Model.to_csv(f'{output_file_path}_average_future_predictions_org_by_model.csv', index=True)

average_future_predictions_by_Symbol = compute_average_metrics(all_future_predictions_df, 'Symbol')
average_future_predictions_org_by_Symbol = compute_average_metrics(all_future_predictions_org_df, 'Symbol')

# Save average_future_predictions
average_future_predictions_by_Symbol.to_csv(f'{output_file_path}_average_future_predictions_by_symbol.csv', index=True)
average_future_predictions_org_by_Symbol.to_csv(f'{output_file_path}_average_future_predictions_org_by_symbol.csv', index=True)

# average_metrics_test_dates_by_Model = compute_average_test_dates(average_metrics_by_Model, 'Model')
# average_metrics_test_dates_by_Model.to_csv(f'{output_file_path}_average_metrics_test_dates_by_Model.csv', index=True)
# average_metrics_test_dates_by_Symbol = compute_average_test_dates(average_metrics_by_Symbol, 'Symbol')
# average_metrics_test_dates_by_Symbol.to_csv(f'{output_file_path}_average_metrics_test_dates_by_Symbol.csv', index=True)

# average_future_predictions_test_dates_by_Model = compute_average_test_dates(average_future_predictions_by_Model, 'Model')
# average_future_predictions_org_test_dates_by_Model = compute_average_test_dates(average_future_predictions_org_by_Model, 'Model')

# average_future_predictions_test_dates_by_Model.to_csv(f'{output_file_path}_average_future_predictions_test_dates_by_Model.csv', index=True)
# average_future_predictions_org_test_dates_by_Model.to_csv(f'{output_file_path}_average_future_predictions_org_test_dates_by_Model.csv', index=True)

# average_future_predictions_test_dates_by_Symbol = compute_average_test_dates(average_future_predictions_by_Symbol, 'Symbol')
# average_future_predictions_org_test_dates_by_Symbol = compute_average_test_dates(average_future_predictions_org_by_Symbol, 'Symbol')

# average_future_predictions_test_dates_by_Symbol.to_csv(f'{output_file_path}_average_future_predictions_test_dates_by_Symbol.csv', index=True)
# average_future_predictions_org_test_dates_by_Symbol.to_csv(f'{output_file_path}_average_future_predictions_org_test_dates_by_Symbol.csv', index=True)


```
