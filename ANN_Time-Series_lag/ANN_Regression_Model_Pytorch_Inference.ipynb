```python
import pandas as pd
import numpy as np
import yfinance as yf
import torch
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error


def calculate_indicators_ver3(data, interval=None, seq_len=None):
    # Adjust parameters based on frequency
    if interval == '1wk':  # Weekly data
        period = 6
        rsi_period = 9
        short_ema_period = 8
        long_ema_period = 16
        signal_period = 5
        vol_period = 14
        bband_window_size = 12
    elif interval == '1mo':  # Monthly data
        period = 4
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 12
        signal_period = 4
        vol_period = 10
        bband_window_size = 8
    elif interval == '1d':  # Monthly data
        period = 7
        rsi_period = 14
        short_ema_period = 12
        long_ema_period = 26
        signal_period = 9
        vol_period = 26
        bband_window_size = 20
    elif interval == None:  # not specified
        period = 5
        rsi_period = 7
        short_ema_period = 6
        long_ema_period = 13
        signal_period = 4
        vol_period = 13
        bband_window_size = 10

    if data.shape[0] < long_ema_period + seq_len:
        return None
    
    # Check if 'Close' column exists in the data
    if 'Close' in data.columns:
        # Calculate daily returns
        data['Return'] = data['Close'].pct_change()
        # Calculate daily Direction
        data["Direction_Signal"] = np.where((data['Return']<-0.001), 1, np.where((data['Return']>0.001), 3, 2))
        # Calculate VPT (Volume-Price Trend)
        data['VPT'] = (data['Volume'] * data['Return']).cumsum()
        # Price Range
        data['Price Range'] = data['High'] - data['Low']
        # Candle Body Size
        data['Body Size'] = data['Open'] - data['Close']
        # Upper Shadow Size
        data['Upper Shadow'] = data.apply(lambda row: row.High - max(row.Open, row.Close), axis=1)
        # Lower Shadow Size
        data['Lower Shadow'] = data.apply(lambda row: min(row.Open, row.Close) - row.Low, axis=1)

        # Calculate ZLEMA
        lag = (period - 1) // 2  # Ensure integer division
        ema = data['Close'].ewm(span=10, adjust=False).mean()
        data['ZLEMA'] = ema + (ema - ema.shift(lag))

        # Define tolerance as a percentage of the close price
        tolerance = 0.015

        # Initialize column with 0s
        data['ZLEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * ZLEMA
        data.loc[data['Close'] > data['ZLEMA'] * (1 + tolerance), 'ZLEMA_Signal'] = 3

        # Sell signal when Close is below (1 - tolerance) * ZLEMA
        data.loc[data['Close'] < data['ZLEMA'] * (1 - tolerance), 'ZLEMA_Signal'] = 1

        # Calculate the first exponential moving average (EMA1).
        ema1 = data['Close'].ewm(span=period, min_periods=period).mean()
        # Calculate the second exponential moving average (EMA2).
        ema2 = ema1.ewm(span=period, min_periods=period).mean()
        # Calculate the third exponential moving average (EMA3).
        ema3 = ema2.ewm(span=period, min_periods=period).mean()
        # Calculate the TEMA.
        tema = 3 * ema1 - 3 * ema2 + ema3

        # Convert the TEMA to a Pandas Series.
        data['TEMA'] = tema

        # Initialize column with 0s
        data['TEMA_Signal'] = 2

        # Buy signal when Close is above (1 + tolerance) * TEMA
        data.loc[data ['Close']> data ['TEMA']* (1 + tolerance), 'TEMA_Signal']= 3

        # Sell signal when Close is below (1 - tolerance) * TEMA
        data.loc[data ['Close']< data ['TEMA']*(1-tolerance), 'TEMA_Signal']= 1

        # Calculate RSI and its signal
        delta = data['Close'].diff()
        up, down = delta.copy(), delta.copy()

        up[up < 0] = 0
        down[down > 0] = 0

        average_gain = up.rolling(window=rsi_period).mean()
        average_loss = abs(down.rolling(window=rsi_period).mean())

        epsilon = 1e-10  # A small constant to prevent division by zero
        rs = average_gain / (average_loss + epsilon)

        data['RSI'] =100 - (100 / (1 + rs))
        data["RSI_Signal"] =2
        data.loc[data.RSI < 30,"RSI_Signal"] =3
        data.loc[data.RSI > 70,"RSI_Signal"] =1

        # # Calculate MACD Line: (12-day EMA - 26-day EMA)
        EMA_short = data['Close'].ewm(span=short_ema_period).mean()
        EMA_long = data['Close'].ewm(span=long_ema_period).mean()
        data['MACD_Line'] = EMA_short - EMA_long

        # Calculate Signal Line: a n-day MA of MACD Line
        data['Signal_Line'] = data["MACD_Line"].ewm(span=signal_period).mean()

        # Generate MACD signals based on crossovers
        data["MACD_Signal"] = 2 # Neutral
        data.loc[(data.MACD_Line > data.Signal_Line) & (data.MACD_Line.shift() < data.Signal_Line.shift()),"MACD_Signal"] = 3 # Buy
        data.loc[(data.MACD_Line < data.Signal_Line) & (data.MACD_Line.shift() > data.Signal_Line.shift()),"MACD_Signal"] = 1 # Sell

        # Generate EMA signals based on crossovers
        data["EMA_Short"]=EMA_short
        data["EMA_Long"]=EMA_long

        # Define a small percentage as a threshold
        ema_diff_threshold = 0.01 # 1% difference

        # Calculate absolute percent difference between short and long EMA
        data['EMA_Diff'] = (EMA_short - EMA_long) / EMA_long

        data["EMA_Signal"] = np.where((data["EMA_Short"] > data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 3, 
                             np.where((data["EMA_Short"] < data["EMA_Long"]) & (abs(data["EMA_Diff"]) > ema_diff_threshold), 1, 2))

        # Calculate Volatility as rolling standard deviation of log returns
        data["Log_Return"] =(np.log(data.Close).diff())
        data["Volatility"] =(data.Log_Return.rolling(window=vol_period, center=False).std())

        # Generate Volatility signals based on high or low volatility
        high_vol_threshold = data["Volatility"].quantile(0.85)
        low_vol_threshold = data["Volatility"].quantile(0.15)
        
        data['Volatility_Signal'] = 2
        data.loc[data.Volatility > high_vol_threshold, 'Volatility_Signal'] = 3
        data.loc[data.Volatility < low_vol_threshold, 'Volatility_Signal'] = 1

        # Calculate Bollinger Bands
        data["Middle_Band"] = data["Close"].rolling(window=bband_window_size, center=False).mean()
        data["Std_Dev"] = data["Close"].rolling(window=bband_window_size, center=False).std()
        data["Upper_Band"] = data["Middle_Band"] + data["Std_Dev"] * 2
        data["Lower_Band"] = data["Middle_Band"] - data["Std_Dev"] * 2

        # Identify shifting points
        data["BBand_Signal"] = 2  # neutral

        # Create a copy of the "Close" column
        close_shifted = data["Close"].copy()

        # Check for crossing Upper Band
        cross_upper_band = (close_shifted.shift(1) <= data["Upper_Band"].shift(1)) & (data["Close"] > data["Upper_Band"])
        data.loc[cross_upper_band, "BBand_Signal"] = 3  # buy

        # Check for crossing Lower Band
        cross_lower_band = (close_shifted.shift(1) >= data["Lower_Band"].shift(1)) & (data["Close"] < data["Lower_Band"])
        data.loc[cross_lower_band, "BBand_Signal"] = 1  # sell

        # Drop unnecessary columns
        data = data.drop(columns=["Middle_Band", "Std_Dev", 'Log_Return', 'Open', 'Low', 'High', 'Volume'])

    else:
        print("'Close' column is not present in the input dataframe.")

    return data

def fetch_data(symbol_list, start_date, end_date=None, interval='1d', target_only=False):
    # Initialize an empty DataFrame
    all_data = []

    # Keep 'today' as a datetime object for datetime operations
    today_datetime = datetime.now()
    # Convert to string only when needed for string operations
    today_str = today_datetime.strftime('%Y-%m-%d')

    if end_date is None:
        end_date = today_str

    for symbol in symbol_list:
        try:
            data = yf.download(symbol, start=start_date, end=end_date, interval=interval)
        except KeyError:
            print(f"Data for {symbol} not found.")
            continue

        data = data.copy()

        # Convert the index to datetime
        data.index = pd.to_datetime(data.index)

        # If the 'Date' column is expected to be the index after fetching data
        if 'Date' not in data.columns and data.index.name == 'Date':
            data.reset_index(inplace=True)
        elif 'Date' not in data.columns:
            print("Date column is still missing. Please check the DataFrame source.")

        if interval.endswith('mo'):
            # Check if the last data point is from the current month and year
            if data.index[-1].month == today_datetime.month and data.index[-1].year == today_datetime.year:
                # Count the number of days in the current month up to today
                days_this_month_so_far = today_datetime.day

            # If the number of days in the last month is less than 20, drop the last month's data
            if days_this_month_so_far < 20:
                data = data.iloc[:-1].copy()

        elif interval.endswith('wk'):
            weekdays_till_today = (today_datetime.weekday() + 1) % 7
            # If the number of days in the last week is less than 3, drop the last week's data
            if weekdays_till_today < 3:  # 0: Monday, ..., 6: Sunday
                data = data.iloc[:-1].copy()

        data.dropna(inplace=True)
        data = data[data['Volume'] != 0].copy()

        # If "Close" only = true, drop all columns except "Close"
        if target_only:
            data.drop(['Open', 'High', 'Low', 'Volume', 'Adj Close'], axis=1, inplace=True)
        else:
            data.drop(['Adj Close'], axis=1, inplace=True)

        # Add a column to identify the symbol
        data['Symbol'] = symbol
        
        # Append the data to the all_data list
        all_data.append(data)

    # Concatenate all_data into a single DataFrame 
    all_data_df = pd.concat(all_data)

    return all_data_df

def normalize_data(features, target):
    """ Normalize the dataframe using mean and standard deviation, assumes features are all numeric. """
    # Exclude datetime and other non-numeric columns from normalization
    numeric_features = features.select_dtypes(include=[np.number])
    
    feature_mean = numeric_features.mean()
    feature_std = numeric_features.std()

    # Apply normalization only to numeric features
    normalized_features = (numeric_features - feature_mean) / feature_std

    # Normalize the target if it's numeric
    target_mean = target.mean() if pd.api.types.is_numeric_dtype(target) else None
    target_std = target.std() if pd.api.types.is_numeric_dtype(target) else None
    normalized_target = (target - target_mean) / target_std if pd.api.types.is_numeric_dtype(target) else target

    return normalized_features, normalized_target, target_mean, target_std

def inverse_normalize(prediction, target_mean, target_std):
    """ Revert the normalization for a prediction. """
    return prediction * target_std + target_mean

def create_lagged_rows(data, num_lags, target_col, time_column='Date'):
    """
    Create lagged rows for all columns except the specified target column,
    and include the current time from the specified time column.
    Optionally use scale parameters if provided.
    """
    # Create a list to hold the lagged data rows
    lagged_data_rows = []
    
    # Determine which columns are features based on exclusion of the target and time columns
    feature_columns = [col for col in data.columns if col != target_col and col != time_column]

    for i in range(num_lags, len(data)):
        row = {}
        
        # Create lags for each feature
        for feature in feature_columns:
            for lag in range(0, num_lags + 1):
                row[f'{feature}_lag_{lag}'] = data[feature].iloc[i - lag]
        
        # Create lags for the target column
        for lag in range(1, num_lags + 1):
            row[f'{target_col}_lag_{lag}'] = data[target_col].iloc[i - lag]
        
        # Include the target column as 'Target' for lag_0
        row['Target'] = data[target_col].iloc[i]
        
        lagged_data_rows.append(row)
    
    return pd.DataFrame(lagged_data_rows)

def generate_future_prediction_data(latest_data, last_used_data, num_lags, target_feature):
    future_data = {}
    
    # Set up lag_0 with latest data for all features except the target column
    for feature in latest_data.index:
        if feature != 'Date' and feature != target_feature:  
            future_data[f'{feature}_lag_0'] = latest_data[feature]
    
    # Set up lag_1 with latest data for all features including the target column
    for feature in latest_data.index:
        if feature != 'Date':  
            future_data[f'{feature}_lag_1'] = latest_data[feature]
    
    # Set up lag_2 to num_lags data from the last_used_data
    for lag in range(2, num_lags + 1):
        for feature in latest_data.index:
            if feature != 'Date':  
                # Retrieve the last used data for the previous lag, if it exists, otherwise default to None
                future_data[f'{feature}_lag_{lag}'] = last_used_data.get(f'{feature}_lag_{lag - 1}', None)
    
    # Order the future_data to ensure that the sequence is Open_lag_1, Open_lag_2, etc.
    ordered_future_data = {}
    for feature in latest_data.index:
        if feature != 'Date':
            for lag in range(0, num_lags + 1):
                if not (lag == 0 and feature == target_feature):  # Exclude target feature for lag 0
                    ordered_future_data[f'{feature}_lag_{lag}'] = future_data.get(f'{feature}_lag_{lag}')

    return ordered_future_data

def impute_nulls(df):
    """
    Impute null values in a Pandas DataFrame based on the data type of each column.
    - For float columns, impute with the mean.
    - For integer columns, impute with the median.
    - For object columns, impute with the mode.
    - For datetime columns, impute with the most recent or most frequent date.
    - For timedelta columns, impute with the mode.
    - For bool columns, impute with the mode.
    - For category columns, impute with the mode.
    - For complex columns, impute with the mean.
    """
    # Get data types of all columns
    dtypes = df.dtypes

    # Iterate over all columns
    for col in df.columns:
        # Check if column contains null values
        if df[col].isnull().sum() > 0:
            # Get data type of column
            dtype = dtypes[col]
            # Impute null values based on data type
            if dtype in ['float64', 'float32', 'float16', 'complex64', 'complex128']:
                df[col] = df[col].fillna(df[col].mean())
            elif dtype in ['int64', 'int32', 'int16', 'int8']:
                df[col] = df[col].fillna(df[col].median())
            elif dtype in ['object', 'timedelta64', 'bool', 'category']:
                mode_value = df[col].mode()
                if not mode_value.empty:
                    df[col] = df[col].fillna(mode_value[0])
                else:
                    # Handle case where all values are NaN or no mode found
                    # For example, use a common default value or skip the column
                    df[col] = df[col].fillna('Unknown')  # Modify this based on your context
            elif dtype == 'datetime64':
                # For datetime, choosing to backfill, but can also forward fill or use a specific value
                df[col] = df[col].fillna(method='bfill')
    return df

def prepare_time_series_data(data=None, group_col=None, target_col=None, time_column=None, num_lags=None, symbol=None, interval=None, indicator_use=False):

    test_data = data[data[group_col] == symbol]
    
    test_data = test_data.drop(group_col, axis=1)

    if indicator_use:            
        test_data = calculate_indicators_ver3(test_data, interval)        
        # Handling NaNs before proceeding
        test_data.dropna(inplace=True)

    if target_col in test_data.columns:
        target = test_data[target_col]
        numeric_features = test_data.drop(columns=[target_col])

    normalized_features, normalized_target, target_mean, target_std = normalize_data(numeric_features, target)

    # Reconstruct normalized data including datetime columns for proper indexing
    normalized_data = normalized_features.copy()
    normalized_data[target_col] = normalized_target
    # normalized_data[time_column] = test_data[time_column]  # Preserve datetime column for sorting and future use

    lagged_data = create_lagged_rows(normalized_data, num_lags, target_col, time_column)

    # Create a new DataFrame by copying lagged_data
    lagged_data_shifted = lagged_data.copy()

    # Shift the 'Target' column in the new DataFrame
    lagged_data_shifted['Target'] = lagged_data_shifted['Target'].shift(-1)

    # Generate future data for predictions using the latest data point
    if len(lagged_data) > num_lags:
        latest_data = normalized_data.iloc[-1]
        # print(latest_data)
        last_used_data = lagged_data.iloc[-1]  # Last used lagged data point
        # print(last_used_data)
        future_prediction_data = generate_future_prediction_data(latest_data, last_used_data, num_lags, target_col)
        future_data = future_prediction_data
        future_data_df = pd.DataFrame([future_data])
        future_data_impute = impute_nulls(future_data_df)
        # lagged_data = lagged_data.iloc[:-num_lags]  # Use all but the last 'num_lags' entries for training

    X_test = lagged_data_shifted.dropna().drop(columns=['Target'])
    y_test = lagged_data_shifted.dropna()['Target']

    return X_test, y_test, test_data, target_mean, target_std, future_data_impute
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import random
import timeit
import os
import yfinance as yf
import copy
import plotly.graph_objs as go
import plotly.io as pio
import re
import joblib
from datetime import datetime
import calendar
import warnings
import gc
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")

# def predict_for_each_sequence(model, sequences, target_mean, target_std):
#     predictions = []
#     for seq in sequences:
#         # Check if the sequence is a NumPy array, if not, log an error or convert it
#         if not isinstance(seq, np.ndarray):
#             print("Error: Expected a NumPy array, but got:", type(seq))
#             continue  # Skip this sequence or handle it according to your needs
        
#         # Check if seq is a 1D array and reshape if necessary
#         if seq.ndim == 1:
#             seq = seq.reshape(1, -1)  # Reshape from (n_features,) to (1, n_features)
        
#         # Now predict using the model
#         output = model.predict(seq)
#         predictions.append(output)
#     predictions = np.array(predictions)
#     predictions_org = inverse_normalize(predictions, target_mean, target_std)

#     return predictions, predictions_org

# def predict_for_each_sequence(model, sequences, target_mean, target_std):
#     predictions = []
#     for _, seq in sequences.iterrows():  # Use iterrows to iterate over DataFrame rows
#         # Convert row to NumPy array and ensure it's 2D
#         seq_array = seq.values.reshape(1, -1)

#         # Now predict using the model
#         output = model.predict(seq_array)
#         predictions.append(output)

#     predictions = np.array(predictions).flatten()  # Flatten since we get an array of arrays
#     predictions_org = inverse_normalize(predictions, target_mean, target_std)

#     return predictions, predictions_org

class ANNModel(nn.Module):
    """
    Artificial Neural Network model class.

    This class defines an ANN architecture with multiple hidden layers 
    and activation functions between each layer. Dropout is also applied
    between hidden layers for regularization. L1 and L2 regularization 
    are supported on the weights.
    """
    def __init__(self, n_features, num_hidden_layers=2, neurons_per_layer=64, activation_function=torch.nn.ReLU(), 
                 dropout_rate=0.1, l1_regularization=0, l2_regularization=0):
        super(ANNModel, self).__init__()

        self.activation_function = activation_function
        self.l1_regularization = l1_regularization
        self.l2_regularization = l2_regularization

        # Build the ANN architecture
        layers = [nn.Linear(n_features, neurons_per_layer), activation_function]
        for _ in range(num_hidden_layers - 1):
            layers += [
                nn.Linear(neurons_per_layer, neurons_per_layer),
                activation_function,
                nn.Dropout(dropout_rate)
            ]
        
        # Final layer without activation
        layers.append(nn.Linear(neurons_per_layer, 1))
        
        # Wrap the layers into nn.Sequential
        self.layers = nn.Sequential(*layers)
        
        # Initialize weights
        self.init_weights()

    def init_weights(self):
        """Initialize weights of layers using Xavier initialization."""

        for layer in self.layers:
            if isinstance(layer, nn.Linear):
                # Describe what this line is doing
                nn.init.xavier_uniform_(layer.weight) 
                # Describe what this line is doing  
                nn.init.zeros_(layer.bias)

    def regularization_loss(self):
        """Calculate regularization loss from model weights."""

        l1_loss = 0
        # Describe what this variable represents
        l2_loss = 0  

        for param in self.parameters():
            # Describe what this line is doing
            l1_loss += torch.norm(param, 1)  
            # Describe what this line is doing
            l2_loss += torch.norm(param, 2) ** 2

        return self.l1_regularization * l1_loss + self.l2_regularization * l2_loss

    def forward(self, x):
        """Forward pass for the ANN model."""

        for layer in self.layers:
            # Describe what this line is doing
            x = layer(x) if not isinstance(layer, nn.Dropout) else layer(x)

            if any(isinstance(layer, activ) for activ in [torch.nn.ReLU, torch.nn.Tanh, torch.nn.ELU, torch.nn.LeakyReLU, torch.nn.Sigmoid]):
                # Describe what this line is doing
                x = self.activation_function(x)

        return x

def get_device():
    return torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_ann_model(model_path, device):
    model = torch.load(model_path)
    model.to(device)  # Move model to the appropriate device
    model.eval()  # Set the model to evaluation mode
    return model

def predict_with_ann(model, input_tensor, device):
    # Ensure the model is in evaluation mode
    model.eval()

    # Convert input data to PyTorch tensor if it's not already
    if not isinstance(input_tensor, torch.Tensor):
        input_tensor = torch.Tensor(input_tensor).to(device)
    
    # Add a batch dimension if necessary
    if input_tensor.dim() == 1:
        input_tensor = input_tensor.unsqueeze(0)
    
    with torch.no_grad():  # No need to track gradients during inference
        output = model(input_tensor)
    
    # Move output to CPU and convert to numpy array for further processing
    return output.cpu().numpy()
    
def plot_predictions_vs_actuals(actuals, predictions):
    plt.figure(figsize=(20, 12))

    # Actual prices plotted normally
    actual_times = np.arange(len(actuals))
    plt.plot(actual_times, actuals, label='Actual Prices', color='blue', marker='o')

    # Predictions plotted with an offset to align with the actual times
    prediction_times = np.arange(len(predictions))
    plt.plot(prediction_times, predictions, label='Predicted Prices', color='red', linestyle='--', marker='x')

    plt.title('Actual vs Predicted Prices')
    plt.xlabel('Time Steps')
    plt.ylabel('Prices')
    plt.legend()
    plt.show()

def evaluate_and_plot_recursive_forecasting_with_future(model, X_test=None, y_test=None, data=None, plots_use=False, n_days=None, freq=None, n_predict=None, save_directory=None, 
                                                       target_col=None, symbol=None, trial=None, future_data=None, target_mean=None, target_std=None, pca=None, device=None):
    device = device or get_device()  # Ensure device is set, otherwise get the default device
    model.to(device)  # Make sure the model is on the correct device

    all_predictions = []
    all_predictions_org = []
    all_actuals = []

    for i in range(n_days + 1):
        start_idx = -(n_days + 1 - i)

        if start_idx < -1:
            end_idx = start_idx + 1
            actual_prices = y_test[start_idx:end_idx]
            input_sequence = X_test[start_idx:end_idx]    
        else:
            actual_prices = y_test[start_idx:]
            input_sequence = X_test[start_idx:]

        # Ensure input_sequence is a torch.Tensor and on the correct device
        if not isinstance(input_sequence, torch.Tensor):
            input_sequence = torch.tensor(input_sequence.values, dtype=torch.float32).to(device)

        # Make predictions using the ANN model
        predictions = model(input_sequence)

        # Inverse normalization applied to predictions detached from the computation graph
        predictions_org = inverse_normalize(predictions.detach().cpu().numpy(), target_mean, target_std)

        all_predictions.extend(predictions.detach().cpu().numpy().tolist())  # Converting tensor to list
        all_predictions_org.extend(predictions_org)
        all_actuals.extend(actual_prices)


    # Convert lists to numpy arrays for metric computation
    all_actuals = np.array(all_actuals)
    all_predictions = np.array(all_predictions)
    all_predictions_org = np.array(all_predictions_org)

    all_actuals_squeezed = np.squeeze(all_actuals)
    all_predictions_squeezed = np.squeeze(all_predictions)
    all_predictions_org_squeezed = np.squeeze(all_predictions_org)

    # Compute returns for actual prices
    all_return_actuals = (all_actuals_squeezed[1:] - all_actuals_squeezed[:-1]) / all_actuals_squeezed[:-1]

    # Compute returns for predicted prices
    all_predictions_return = (all_predictions_squeezed[1:] - all_predictions_squeezed[:-1]) / np.abs(all_predictions_squeezed[:-1])
    # all_predictions_return = (all_predictions_squeezed[1:] - all_actuals_squeezed[:-1]) / all_actuals_squeezed[:-1]

    all_actuals_org = inverse_normalize(all_actuals, target_mean, target_std)
    # all_predictions_org = inverse_normalize(all_predictions, target_mean, target_std)

    all_actuals_org_squeezed = np.squeeze(all_actuals_org)

    all_return_actuals_org = (all_actuals_org_squeezed[1:] - all_actuals_org_squeezed[:-1]) / all_actuals_org_squeezed[:-1]
    all_predictions_return_org = (all_predictions_org_squeezed[1:] - all_predictions_org_squeezed[:-1]) / np.abs(all_predictions_org_squeezed[:-1])
    # all_predictions_return_org = (all_predictions_org_squeezed[1:] - all_actuals_org_squeezed[:-1]) / all_actuals_org_squeezed[:-1]

    # Adjusting the sample sizes
    all_actuals = all_actuals[1:]
    all_predictions = all_predictions[1:]
    # all_return_actuals = all_return_actuals[1:]

    all_actuals_org = all_actuals_org[1:]
    all_predictions_org = all_predictions_org[1:]
    # all_return_actuals_org = all_return_actuals_org[1:]
    
    # Assuming future_data is a Pandas DataFrame and needs to be converted
    if isinstance(future_data, pd.DataFrame):
        # Convert DataFrame to tensor
        future_data_tensor = torch.tensor(future_data.values, dtype=torch.float32).to(device)
    else:
        # Assuming future_data is already a tensor but still check if it's on the right device
        future_data_tensor = future_data.to(device)

    # Pass the tensor to the model
    future_predictions = model(future_data_tensor)

    # Process the output if necessary, e.g., detaching and converting to numpy
    future_predictions = future_predictions.detach().cpu().numpy()

    # future_predictions = recursive_forecasting(model, input_tensor, n_steps)
    future_predictions = np.array(future_predictions)
    future_predictions_org = inverse_normalize(future_predictions, target_mean, target_std)

    future_predictions_squeezed = np.squeeze(future_predictions)
    future_predictions_org_squeezed = np.squeeze(future_predictions_org)

    future_return = (future_predictions_squeezed - all_predictions_squeezed[-1]) / np.abs(all_predictions_squeezed[-1])
    future_return_org = (future_predictions_org_squeezed - all_predictions_org_squeezed[-1]) / np.abs(all_predictions_org_squeezed[-1])
    
    mse_score = mean_squared_error(all_actuals, all_predictions)
    mae_score = mean_absolute_error(all_actuals, all_predictions)
    r2 = r2_score(all_actuals, all_predictions)

    mse_return_score = mean_squared_error(all_return_actuals, all_predictions_return)
    mae_return_score = mean_absolute_error(all_return_actuals, all_predictions_return)
    r2_return = r2_score(all_return_actuals, all_predictions_return)

    mse_score_org = mean_squared_error(all_actuals_org, all_predictions_org) 
    mae_score_org = mean_absolute_error(all_actuals_org, all_predictions_org)
    r2_org = r2_score(all_actuals_org, all_predictions_org)

    mse_return_score_org = mean_squared_error(all_return_actuals_org, all_predictions_return_org)
    mae_return_score_org = mean_absolute_error(all_return_actuals_org, all_predictions_return_org)
    r2_return_org = r2_score(all_return_actuals_org, all_predictions_return_org)
    
    metrics = {"MSE": mse_score, "MAE": mae_score, "R2": r2, "Future_Prediction": future_predictions[-1],
               "MSE_return": mse_return_score, "MAE_return": mae_return_score, "R2_return": r2_return, "Future_Return": future_return,
               "MSE_org": mse_score_org, "MAE_org": mae_score_org, "R2_org": r2_org, "Future_Prediction_org": future_predictions_org[-1],
               "MSE_return_org": mse_return_score_org, "MAE_return_org": mae_return_score_org, "R2_return_org": r2_return_org, "Future_Return_org": future_return_org}    
   
    if plots_use:

        # Ensure that the number of predictions matches the number of actuals
        assert len(all_predictions) == len(all_actuals), "The number of predictions and actual prices should match."

        date_index = data.index[-(n_days+1):]
        print(date_index)

        # Plotting the results
        plot_predictions_vs_actuals(all_actuals, all_predictions)
        plot_predictions_vs_actuals(all_return_actuals, all_predictions_return)
        plot_predictions_vs_actuals(all_return_actuals_org, all_predictions_return_org)      

        # Ensure y_test_org and y_pred_prev_org are 2D for consistency in plotting
        all_actuals_org = all_actuals_org.reshape(-1, 1) if all_actuals_org.ndim == 1 else all_actuals_org.reshape(-1, all_actuals_org.shape[-1])
        all_predictions_org = all_predictions_org.reshape(-1, 1) if all_predictions_org.ndim == 1 else all_predictions_org.reshape(-1, all_predictions_org.shape[-1])   
        # Separate the numeric and string parts
        match = re.match(r'(\d+)?(\w+)', freq)
        if match:
            freq_number, freq_string = match.groups()
            freq_number = int(freq_number) if freq_number else 1
        else:
            freq_number = 1
            freq_string = freq

        frequency_mapping = {'M': 'months','MS': 'months','W': 'weeks','D': 'days',}
        freq_param = frequency_mapping.get(freq_string, freq_string)

        try:
            offset = pd.DateOffset(**{freq_param: freq_number})
        except KeyError:
            offset = None

        print(offset)

        if offset is not None:
            start_date_index = date_index[-1] + offset
            future_dates = pd.date_range(start=start_date_index, periods=n_predict, freq=freq)
            print(f'future_dates: {future_dates}')
        else:
            print("Invalid frequency parameter.")

        last_future_prediction = future_predictions_org.flatten()
        combined_predictions = np.concatenate([all_predictions_org.flatten(), last_future_prediction])                

        combined_dates = pd.concat([pd.Series(date_index), pd.Series(future_dates)]).reset_index(drop=True)
        combined_dates = pd.DatetimeIndex(combined_dates)

        fig = go.Figure()
        # When plotting use_target_col_plot
        fig.add_trace(go.Scatter(x=combined_dates[:-n_predict], y=all_actuals_org[:, -1], mode='lines', name='Actual'))
        fig.add_trace(go.Scatter(x=combined_dates, y=combined_predictions, mode='lines', name='Predicted + Future'))

        fig.update_layout(title=f'Actual and Predicted Values for Target ({symbol}_{trial})', 
                            xaxis_title='Time Step', yaxis_title='Value', height=800, width=1500)

        if save_directory:
            save_path = os.path.join(save_directory, f"predictions_plot_target_{symbol}_{trial}.html")
            fig.write_html(save_path)

        fig.show()

    if plots_use:

        # Ensure that the number of predictions matches the number of actuals
        assert len(all_predictions_return) == len(all_return_actuals), "The number of predictions and actual prices should match."
     
        date_index = data.index[-(n_days+1):]

        # Ensure y_test_org and y_pred_prev_org are 2D for consistency in plotting
        all_return_actuals = all_return_actuals.reshape(-1, 1) if all_return_actuals.ndim == 1 else all_return_actuals.reshape(-1, all_return_actuals.shape[-1])
        all_predictions_return = all_predictions_return.reshape(-1, 1) if all_predictions_return.ndim == 1 else all_predictions_return.reshape(-1, all_predictions_return.shape[-1])
        # Separate the numeric and string parts
        match = re.match(r'(\d+)?(\w+)', freq)
        if match:
            freq_number, freq_string = match.groups()
            freq_number = int(freq_number) if freq_number else 1
        else:
            freq_number = 1
            freq_string = freq

        frequency_mapping = {'M': 'months','MS': 'months','W': 'weeks','D': 'days',}
        freq_param = frequency_mapping.get(freq_string, freq_string)

        try:
            offset = pd.DateOffset(**{freq_param: freq_number})
        except KeyError:
            offset = None

        print(offset)

        if offset is not None:
            start_date_index = date_index[-1] + offset
            future_dates = pd.date_range(start=start_date_index, periods=n_predict, freq=freq)
            print(f'future_dates: {future_dates}')
        else:
            print("Invalid frequency parameter.")

        last_future_return = future_return.flatten()
        combined_predictions_return = np.concatenate([all_predictions_return.flatten(), last_future_return])                

        combined_dates = pd.concat([pd.Series(date_index), pd.Series(future_dates)]).reset_index(drop=True)
        combined_dates = pd.DatetimeIndex(combined_dates)

        fig = go.Figure()
        # When plotting use_target_col_plot
        fig.add_trace(go.Scatter(x=combined_dates[:-n_predict], y=all_return_actuals[:, -1], mode='lines', name='Actual'))
        fig.add_trace(go.Scatter(x=combined_dates, y=combined_predictions_return, mode='lines', name='Predicted + Future'))

        fig.update_layout(title=f'Actual and Predicted Return Values for Target ({symbol}_{trial})', 
                            xaxis_title='Time Step', yaxis_title='Value', height=800, width=1500)

        if save_directory:
            save_path = os.path.join(save_directory, f"predictions_return_plot_target_{symbol}_{trial}.html")
            fig.write_html(save_path)

        fig.show()

    return metrics, future_predictions, future_predictions_org, future_return, future_return_org, all_predictions, all_predictions_org, all_predictions_return, all_predictions_return_org, all_actuals, all_actuals_org, all_return_actuals, all_return_actuals_org

def compute_future_difference(future_predictions=None, future_predictions_org=None, future_return=None, future_return_org=None, all_predictions=None, all_predictions_org=None, all_predictions_return=None, 
                              all_predictions_return_org=None, all_actuals=None, all_actuals_org=None, all_return_actuals=None, all_return_actuals_org=None, n_predict=None, col_label='Target'):
    
    n_features = all_actuals.shape[-1]
      
    future_predictions = np.squeeze(future_predictions)
    future_predictions_org = np.squeeze(future_predictions_org)

    all_return_actuals_ABS = np.abs(all_return_actuals)
    all_return_actuals_org_ABS = np.abs(all_return_actuals_org)

    all_predictions_return_ABS = np.abs(all_predictions_return)
    all_predictions_return_org_ABS = np.abs(all_predictions_return_org)   

    all_actuals_squeezed = np.squeeze(all_actuals)
    all_actuals_org_squeezed = np.squeeze(all_actuals_org)
    all_predictions_squeezed = np.squeeze(all_predictions)
    all_predictions_org_squeezed = np.squeeze(all_predictions_org)
    all_return_actuals_squeezed = np.squeeze(all_return_actuals)
    all_return_actuals_org_squeezed = np.squeeze(all_return_actuals_org)
    all_predictions_return_squeezed = np.squeeze(all_predictions_return)
    all_predictions_return_org_squeezed = np.squeeze(all_predictions_return_org)

    future_predictions_last = future_predictions.reshape(-1, 1)
    future_predictions_org_last = future_predictions_org.reshape(-1, 1)
    future_return_last = future_return.reshape(-1, 1)
    future_return_org_last = future_return_org.reshape(-1, 1)

    all_predictions_return_last = all_predictions_return[-(n_predict):]
    all_predictions_return_org_last = all_predictions_return_org[-(n_predict):]
    all_predictions_return_last_ABS = all_predictions_return_ABS[-(n_predict):]
    all_predictions_return_org_last_ABS = all_predictions_return_org_ABS[-(n_predict):]

    all_predictions_last = all_predictions[-(n_predict):]
    all_predictions_org_last = all_predictions_org[-(n_predict):]
    all_actuals_last = all_actuals[-(n_predict):]
    all_actuals_org_last = all_actuals_org[-(n_predict):]
    all_return_actuals_last = all_return_actuals[-(n_predict):]
    all_return_actuals_org_last = all_return_actuals_org[-(n_predict):]
    all_return_actuals_last_ABS = all_return_actuals_ABS[-(n_predict):]
    all_return_actuals_org_last_ABS = all_return_actuals_org_ABS[-(n_predict):]

    # Initialize an empty DataFrame to store all the data
    all_df = pd.DataFrame()
    all_org_df = pd.DataFrame()
    all_residual_df = pd.DataFrame()
    np.seterr(divide='ignore', invalid='ignore')


    # Assuming col_label is a list of strings
    column_name = 'Target'

    # Create DataFrames without indexing a second dimension
    future_predictions_last_df = pd.DataFrame(future_predictions_last, columns=['Future_' + column_name])
    future_predictions_org_last_df = pd.DataFrame(future_predictions_org_last, columns=['Future_org_' + column_name])
    future_return_df = pd.DataFrame(future_return_last, columns=['Future_Return_' + column_name])
    future_return_org_df = pd.DataFrame(future_return_org_last, columns=['Future_Return_org_' + column_name])

    all_predictions_return_last_df = pd.DataFrame(all_predictions_return_last, columns=['Last_Predict_Return_' + column_name])
    all_predictions_return_org_last_df = pd.DataFrame(all_predictions_return_org_last, columns=['Last_Predict_Return_org_' + column_name])
    all_predictions_return_last_ABS_df = pd.DataFrame(all_predictions_return_last_ABS, columns=['Last_Predict_Return_ABS_' + column_name])
    all_predictions_return_org_last_ABS_df = pd.DataFrame(all_predictions_return_org_last_ABS, columns=['Last_Predict_Return_org_ABS_' + column_name])

    all_predictions_last_df = pd.DataFrame(all_predictions_last, columns=['Last_Predict_' + column_name])
    all_predictions_org_last_df = pd.DataFrame(all_predictions_org_last, columns=['Last_Predict_org_' + column_name])
    all_actuals_last_df = pd.DataFrame(all_actuals_last, columns=['Last_Actual_' + column_name])
    all_actuals_org_last_df = pd.DataFrame(all_actuals_org_last, columns=['Last_Actual_org_' + column_name])

    all_return_actuals_last_df = pd.DataFrame(all_return_actuals_last, columns=['Last_Return_Actual_' + column_name])
    all_return_actuals_org_last_df = pd.DataFrame(all_return_actuals_org_last, columns=['Last_Return_Actual_org_' + column_name])
    all_return_actuals_last_ABS_df = pd.DataFrame(all_return_actuals_last_ABS, columns=['Last_Return_Actual_ABS_' + column_name])
    all_return_actuals_org_last_ABS_df = pd.DataFrame(all_return_actuals_org_last_ABS, columns=['Last_Return_Actual_org_ABS_' + column_name])

    error = np.abs(np.subtract(all_predictions_last, all_actuals_last))
    error_org = np.abs(np.subtract(all_predictions_org_last, all_actuals_org_last))

    error_return = np.subtract(all_predictions_return_last, all_return_actuals_last)
    error_return_ABS = np.abs(error_return)
    error_return_org = np.subtract(all_predictions_return_org_last, all_return_actuals_org_last)
    error_return_org_ABS = np.abs(error_return_org)

    error_df = pd.DataFrame(error, columns=['Last_Predict_Error_' + column_name])
    error_org_df = pd.DataFrame(error_org, columns=['Last_Predict_Error_' + column_name])

    error_return_df = pd.DataFrame(error_return, columns=['Last_Predict_Return_Error_' + column_name])
    error_return_org_df = pd.DataFrame(error_return_org, columns=['Last_Predict_Return_org_Error_' + column_name])
    error_return_ABS_df = pd.DataFrame(error_return_ABS, columns=['Last_Predict_Return_Error_ABS_' + column_name])
    error_return_org_ABS_df = pd.DataFrame(error_return_org_ABS, columns=['Last_Predict_Return_org_Error_ABS_' + column_name])       

    residual_org = np.subtract(all_actuals_org_squeezed, all_predictions_org_squeezed) / all_actuals_org_squeezed
    # residual_org_ABS = np.abs(residual_org)

    residual_return_org = np.subtract(all_return_actuals_org_squeezed, all_predictions_return_org)
    # residual_return_org_ABS = np.abs(residual_return_org)


    direction_test_org = np.where(
        ((all_return_actuals_org_squeezed > 0) & (all_predictions_return_org_squeezed > 0)) |
        ((all_return_actuals_org_squeezed < 0) & (all_predictions_return_org_squeezed < 0)) ,
        1, 0)

    direction_test_org_standard = np.where(
        ((all_return_actuals_org_squeezed > 0) & (all_predictions_return_squeezed > 0)) |
        ((all_return_actuals_org_squeezed < 0) & (all_predictions_return_squeezed < 0)) ,
        1, 0)

    direction_test_standard = np.where(
        ((all_return_actuals_squeezed > 0) & (all_predictions_return_squeezed > 0)) |
        ((all_return_actuals_squeezed < 0) & (all_predictions_return_squeezed < 0)) ,
        1, 0)
    
    direction_test_standard_org = np.where(
        ((all_return_actuals_squeezed > 0) & (all_predictions_return_org_squeezed > 0)) |
        ((all_return_actuals_squeezed < 0) & (all_predictions_return_org_squeezed < 0)) ,
        1, 0)

    direction_test_last_org = np.where(
        ((all_return_actuals_org_last > 0) & (all_predictions_return_org_last > 0)) |
        ((all_return_actuals_org_last < 0) & (all_predictions_return_org_last < 0)) ,
        1, 0)
    
    direction_test_last_org_standard = np.where(
        ((all_return_actuals_org_last > 0) & (all_predictions_return_last > 0)) |
        ((all_return_actuals_org_last < 0) & (all_predictions_return_last < 0)) ,
        1, 0)

    direction_test_last_standard = np.where(
        ((all_return_actuals_last > 0) & (all_predictions_return_last > 0)) |
        ((all_return_actuals_last < 0) & (all_predictions_return_last < 0)) ,
        1, 0)
    
    direction_test_last_standard_org = np.where(
        ((all_return_actuals_last > 0) & (all_predictions_return_org_last > 0)) |
        ((all_return_actuals_last < 0) & (all_predictions_return_org_last < 0)) ,
        1, 0) 

        
    all_actuals_df = pd.DataFrame(all_actuals, columns=['Actual_'+ column_name])
    all_actuals_org_df = pd.DataFrame(all_actuals_org, columns=['Actual_org_'+ column_name])
    all_predictions_df = pd.DataFrame(all_predictions, columns=['Predict_'+ column_name])
    all_predictions_org_df = pd.DataFrame(all_predictions_org, columns=['Predict_org_'+ column_name])

    all_return_actuals_df = pd.DataFrame(all_return_actuals, columns=['Actual_Return_'+ column_name])
    # all_return_actuals_ABS_df = pd.DataFrame(all_return_actuals_ABS, columns=['Actual_Return_ABS_'+ column_name])

    all_return_actuals_org_df = pd.DataFrame(all_return_actuals_org, columns=['Actual_Return_org_'+ column_name])
    # all_return_actuals_org_ABS_df = pd.DataFrame(all_return_actuals_org_ABS, columns=['Actual_Return_org_ABS_'+ column_name])
    
    all_predictions_return_df = pd.DataFrame(all_predictions_return, columns=['Predict_Return_'+ column_name])

    all_predictions_return_org_df = pd.DataFrame(all_predictions_return_org, columns=['Predict_Return_org_'+ column_name])
    # all_predictions_return_ABS_org_df = pd.DataFrame(all_predictions_return_org_ABS, columns=['Predict_Return_org_ABS'+ column_name])

    residual_org_df = pd.DataFrame(residual_org, columns=['Predict_Return_org_Residual_'+ column_name])
    # residual_org_ABS_df = pd.DataFrame(residual_org_ABS, columns=['Predict_Return_org_Residual_ABS_'+ column_name])

    residual_return_org_df = pd.DataFrame(residual_return_org, columns=['Predict_Return_org_Residual_'+ column_name])
    # residual_return_org_ABS_df = pd.DataFrame(residual_return_org_ABS, columns=['Predict_Return_org_Residual_ABS_'+ column_name])

    direction_test_org_df = pd.DataFrame(direction_test_org, columns=['Direction_Test_org_'+ column_name])
    direction_test_org_standard_df = pd.DataFrame(direction_test_org_standard, columns=['Direction_Test_org_standard_'+ column_name])
    direction_test_standard_df = pd.DataFrame(direction_test_standard, columns=['Direction_Test_standard_'+ column_name])
    direction_test_standard_org_df = pd.DataFrame(direction_test_standard_org, columns=['Direction_Test_standard_org_'+ column_name])
    direction_test_last_org_df = pd.DataFrame(direction_test_last_org, columns=['Direction_Test_last_org_'+ column_name])
    direction_test_last_org_standard_df = pd.DataFrame(direction_test_last_org_standard, columns=['Direction_Test_last_org_standard_'+ column_name])
    direction_test_last_standard_df = pd.DataFrame(direction_test_last_standard, columns=['Direction_Test_last_standard_'+ column_name])
    direction_test_last_standard_org_df = pd.DataFrame(direction_test_last_standard_org, columns=['Direction_Test_last_standard_org_'+ column_name])

    # Concatenate the DataFrames
    feature_org_df = pd.concat([all_return_actuals_org_last_df, all_return_actuals_org_last_ABS_df, all_actuals_org_last_df, all_predictions_org_last_df, error_org_df, error_return_org_df, 
                                error_return_org_ABS_df, all_predictions_return_org_last_df, all_predictions_return_org_last_ABS_df, 
                                future_predictions_org_last_df, future_return_org_df, direction_test_last_org_df, direction_test_last_org_standard_df,
                                    direction_test_last_standard_df, direction_test_last_standard_org_df], axis=1)
    feature_df = pd.concat([all_return_actuals_last_df, all_return_actuals_last_ABS_df, all_actuals_last_df, all_predictions_last_df, 
                            all_predictions_return_last_df, error_df, error_return_df, all_predictions_return_last_ABS_df, 
                            error_return_ABS_df, future_predictions_last_df, future_return_df], axis=1)
    residuals_df = pd.concat([all_return_actuals_df, all_return_actuals_org_df, all_predictions_return_df, all_predictions_return_org_df, all_actuals_df, all_actuals_org_df, 
                                all_predictions_df, all_predictions_org_df, residual_org_df, residual_return_org_df, direction_test_org_df, direction_test_org_standard_df, direction_test_standard_df, direction_test_standard_org_df], axis=1)
    # Append feature_df to all_df
    all_org_df = pd.concat([all_org_df, feature_org_df], axis=1)
    all_df = pd.concat([all_df, feature_df], axis=1)
    all_residual_df = pd.concat([all_residual_df, residuals_df], axis=1)
    return all_df, all_org_df, all_residual_df, future_predictions_last, future_predictions_org_last

def add_columns(df, item, k, file_name, l):
    # Create a DataFrame for the additional columns
    add_df = pd.DataFrame({'Symbol': [item]*len(df), 
                           'Symbol_ID': [k + 1]*len(df), 
                           'Model': [file_name]*len(df), 
                           'Model_ID': [l + 1]*len(df)
                            })

    # Concatenate the additional columns with the original DataFrame
    return pd.concat([add_df, df], axis=1)

def loaded_model(target_col=None, group_col=None, time_column=None, save_directory=None, load_directory=None, output_file_path=None, plots_use=False, 
                 symbol_list=None, start_date=None, end_date=None, interval=None, freq=None, n_predict=1, n_days=None, pca=None, num_lags=5, train_frac=0.7, indicator_use=False):

    device = get_device()
    
    if save_directory and not os.path.exists(save_directory):
        os.makedirs(save_directory)
    
    # Initialize dictionaries to store results for each symbol
    all_results_metrics_dict = {}
    all_residual_differences_dict = {}
    all_future_predictions_dict = {}
    all_future_predictions_org_dict = {}        
    
    data = fetch_data(symbol_list, start_date=start_date, end_date=end_date, interval=interval)

    X_train = pd.read_csv(os.path.join(output_file_path, 'X_train.csv'))

    if pca is not None:
        pca_transformer = PCA(n_components=pca)
        X_train = pca_transformer.fit_transform(X_train)

    for k, symbol in enumerate(symbol_list):
        
        print(f'Symbol or Code: {symbol}')
        
        start = timeit.default_timer()
        
        result = prepare_time_series_data(data=data, group_col=group_col, target_col=target_col, time_column=time_column, num_lags=num_lags, symbol=symbol, 
                             interval=interval, indicator_use=indicator_use)
        
        # If prepare_data returned None, skip this iteration
        if result is None:
            print(f"Not enough data for {symbol}. Skipping...") 
            continue
        
        X_test, y_test, test_data, target_mean, target_std, future_data = result
 
        if pca is not None:
            X_test = pca_transformer.transform(X_test)
            future_data = pca_transformer.transform(future_data)

        all_future_predictions = []
        all_future_predictions_org = []
        all_results_metrics = []
        all_residual_differences = []
        
        # Get a list of all files in the directory
        file_names = os.listdir(load_directory)
        
        # Iterate over each file name
        for l, file_name in enumerate(file_names):

            # Construct full path to each model file
            model_path = os.path.join(load_directory, file_name)
            # Load the saved model
            loaded_model = load_ann_model(model_path, device)

            # if sequence_predict:
            #     predictions, predictions_org = predict_for_each_sequence(loaded_model, X_test, target_mean, target_std)
            #     future_predictions, future_predictions_org = predict_for_each_sequence(loaded_model, future_data, target_mean, target_std)               
            # else:
            #     predictions = loaded_model.predict(X_test)
            #     predictions_org = inverse_normalize(predictions, target_mean, target_std)
            #     future_predictions = loaded_model.predict(future_data)
            #     future_predictions_org = inverse_normalize(future_predictions, target_mean, target_std)
           
            # Evaluate the model and plot the results
            test_data_copy = test_data.copy()
            test_data_copy.set_index('Date', inplace=True)

            metrics, future_predictions, future_predictions_org, future_return, future_return_org, all_predictions, all_predictions_org, all_predictions_return, all_predictions_return_org, all_actuals, all_actuals_org, all_return_actuals, all_return_actuals_org = evaluate_and_plot_recursive_forecasting_with_future(
                        loaded_model, X_test=X_test, y_test=y_test, data=test_data_copy, n_days=n_days, plots_use=plots_use, freq=freq, n_predict=n_predict, save_directory=save_directory,  
                        symbol=symbol, trial=file_name, target_col=target_col, future_data=future_data, target_mean=target_mean, target_std=target_std, pca=pca)
            
            metrics.update({"Symbol": symbol,"Symbol_ID": k + 1,"Model": file_name, "Model_ID": l + 1})

            future_difference, future_difference_org, residual_difference, future_predictions_last, future_predictions_org_last = compute_future_difference(future_predictions=future_predictions, future_predictions_org=future_predictions_org, 
                                                            future_return=future_return, future_return_org=future_return_org, all_predictions=all_predictions, all_predictions_org=all_predictions_org, all_predictions_return=all_predictions_return, all_predictions_return_org=all_predictions_return_org,
                                                            all_actuals=all_actuals, all_actuals_org=all_actuals_org, all_return_actuals=all_return_actuals, 
                                                            all_return_actuals_org=all_return_actuals_org, n_predict=n_predict, col_label='Target')
            future_predictions = np.squeeze(future_predictions)
            future_predictions_org = np.squeeze(future_predictions_org)            
            
            # Add the additional columns
            future_difference = add_columns(future_difference, symbol, k, file_name, l)
            future_difference_org = add_columns(future_difference_org, symbol, k, file_name, l)
            residual_difference = add_columns(residual_difference, symbol, k, file_name, l)

            # Append the new DataFrame to the list
            all_future_predictions.append(future_difference)
            all_future_predictions_org.append(future_difference_org)
            all_residual_differences.append(residual_difference) 

            # Concatenate all the future predictions into a single DataFrame
            all_future_predictions_df = pd.concat(all_future_predictions, axis=0)
            all_future_predictions_org_df = pd.concat(all_future_predictions_org, axis=0)
            all_residual_differences_df = pd.concat(all_residual_differences, axis=0)  

            all_results_metrics.append(metrics)

            all_results_metrics_df = pd.concat([pd.DataFrame(data=d, index=[0]) for d in all_results_metrics], axis=0)
            
        end = timeit.default_timer()
        # Calculate and print duration
        duration = end - start
        print(f"Execution Time of Symbol_{symbol} is: {duration} seconds")
        # Save results for each symbol in the result dictionaries
        all_results_metrics_dict[symbol] = all_results_metrics_df
        all_future_predictions_dict[symbol] = all_future_predictions_df
        all_future_predictions_org_dict[symbol] = all_future_predictions_org_df
        all_residual_differences_dict[symbol] = all_residual_differences_df

    return all_results_metrics_dict, all_future_predictions_dict, all_future_predictions_org_dict, all_residual_differences_dict
```
